### Declaração e Atribuição de Variáveis

Em Python, não é necessário declarar o tipo da variável explicitamente. O tipo é inferido automaticamente pelo valor atribuído.

## 6. Operadores Aritiméticos

### Operadores Aritméticos em Python

Python suporta os seguintes operadores aritméticos:

* **+** (adição)
* **-** (subtração)
* ***** (multiplicação)
* **/** (divisão com resultado decimal)
* **//** (divisão inteira - retorna apenas a parte inteira)
* **%** (módulo - retorna o resto da divisão)
* ****** (potenciação/exponenciação)

````python
# Declaração de variáveis para os exemplos
a = 10
b = 5
c = 11

# Adição
print(f"Adição: {a} + {b} = {a + b}")

# Subtração
print(f"Subtração: {a} - {b} = {a - b}")
print(f"Subtração inversa: {b} - {a} = {b - a}")

# Divisão (resultado decimal)
print(f"Divisão: {a} / {b} = {a / b}")
print(f"Divisão inversa: {b} / {a} = {b / a}")
````

### Divisão Inteira (//)

Retorna apenas a parte inteira da divisão, descartando a parte decimal.

### Operador Módulo (%)

Retorna o **resto** da divisão entre dois números. Útil para verificar se um número é par/ímpar ou divisível por outro.

````python
# Variáveis para os exemplos
a = 10
b = 5
c = 11

# Divisão inteira (//) - retorna apenas a parte inteira
print(f"Divisão inteira: {a} // {b} = {a // b}")
print(f"Divisão inteira inversa: {b} // {a} = {b // a}")

# Operador módulo (%) - retorna o resto da divisão
print(f"Módulo: {a} % {b} = {a % b}")  # 10 dividido por 5 tem resto 0
print(f"Módulo: {c} % {b} = {c % b}")  # 11 dividido por 5 tem resto 1
print(f"Módulo: {a} % {c} = {a % c}")  # 10 dividido por 11 tem resto 10
````

### Potenciação (**)

Eleva um número a uma potência.

### Precedência de Operadores

Python segue a ordem matemática padrão:
1. **Parênteses** ()
2. **Potenciação** **
3. **Multiplicação, Divisão, Módulo** *, /, //, %
4. **Adição e Subtração** +, -

````python
# Variáveis para os exemplos
a = 10
b = 5
c = 11

# Potenciação (exponenciação)
print(f"Potência: {a} ** 2 = {a ** 2}")  # 10 ao quadrado
print(f"Potência: {a} ** {b} = {a ** b}")  # 10 elevado a 5

# Precedência de operadores - multiplicação antes da adição
resultado1 = a + b * c
print(f"Sem parênteses: {a} + {b} * {c} = {resultado1}")  # 5 * 11 = 55, depois 10 + 55 = 65

# Parênteses alteram a precedência
resultado2 = (a + b) * c
print(f"Com parênteses: ({a} + {b}) * {c} = {resultado2}")  # 10 + 5 = 15, depois 15 * 11 = 165
````

````python
# SUBTRAÇÃO: Diferença entre os dois valores
z = x - y
print(f"Subtração: {x} - {y} = {z}")
````

````python
# MULTIPLICAÇÃO: Produto dos dois valores
z = x * y
print(f"Multiplicação: {x} × {y} = {z}")
````

````python
# DIVISÃO: Quociente dos dois valores (resultado decimal)
if y != 0:
    z = x / y
    print(f"Divisão: {x} ÷ {y} = {z}")
else:
    print(f"Divisão: {x} ÷ {y} = Erro! Divisão por zero não é permitida.")
````

````python
# MÓDULO: Resto da divisão entre os dois valores
if y != 0:
  z = x % y
  print(f"Módulo (resto): {x} % {y} = {z}")
else:
  print("Não é possível calcular o módulo de um número por zero.")
````

````python
# POTENCIAÇÃO: Primeiro número elevado ao segundo
z = x ** y
print(f"Potência: {x} elevado a {y} = {z}")
````

### Versão Compacta (Todas as Operações em Uma Única Célula)

Caso prefira uma versão mais compacta, todas as operações podem ser agrupadas em uma única célula:

````python
# Entrada de dados do usuário
try:
    x = int(input("Digite o primeiro número: "))
    y = int(input("Digite o segundo número: "))

    print(f"\nResultados para x = {x} e y = {y}:")
    print("=" * 50)

    # Realizando todas as operações aritméticas
    print(f"Soma: {x} + {y} = {x + y}")
    print(f"Subtração: {x} - {y} = {x - y}")
    print(f"Multiplicação: {x} × {y} = {x * y}")

    # Divisão e módulo requerem verificação para evitar divisão por zero
    if y != 0:
        print(f"Divisão: {x} ÷ {y} = {x / y}")
        print(f"Módulo (resto): {x} % {y} = {x % y}")
    else:
        print(f"Divisão: {x} ÷ {y} = Erro! Divisão por zero não é permitida.")
        print(f"Módulo (resto): {x} % {y} = Erro! Operação por zero não é permitida.")

    print(f"Potência: {x}^{y} = {x ** y}")
except ValueError:
    print("Erro: Por favor, digite apenas números inteiros válidos.")
````

## 8. Operadores de Comparação ou Relacionais

### Operadores de Comparação (Relacionais)

Operadores de comparação são usados para comparar dois valores e **sempre retornam um valor booleano** (`True` ou `False`).

**Lista de operadores:**

* **==** (igual a)
* **!=** (diferente de)
* **>** (maior que)
* **<** (menor que)
* **>=** (maior ou igual a)
* **<=** (menor ou igual a)

````python
# Declaração de variáveis para os exemplos
a = 5
b = 5

# Operador de igualdade (==) - verifica se os valores são iguais
print(f"a = {a}")
print(f"b = {b}")
print(f"a == 5? {a == 5}")  # True, pois a é 5
print(f"b == 6? {b == 6}")  # False, pois b é 5, não 6
print(f"a == b? {a == b}")  # True, pois ambos valem 5

print("\n" + "=" * 50 + "\n")

# Operador de diferença (!=) - verifica se os valores são diferentes
print(f"a != b? {a != b}")  # False, pois são iguais
print(f"b != (a + 1)? {b != a + 1}")  # True, pois 5 != 6
````

### Operadores de Magnitude (Maior/Menor)

Comparam o tamanho relativo entre dois valores:

* **>** verifica se o valor à esquerda é **maior** que o da direita
* **<** verifica se o valor à esquerda é **menor** que o da direita
* **>=** verifica se o valor à esquerda é **maior ou igual** ao da direita
* **<=** verifica se o valor à esquerda é **menor ou igual** ao da direita

````python
# Variáveis para os exemplos
a = 5
b = 5
c = 10

# Operadores de maior e menor
print(f"a = {a}, b = {b}, c = {c}")
print(f"a > b? {a > b}")  # False, pois são iguais
print(f"c > a? {c > a}")  # True, pois 10 > 5
print(f"a < c? {a < c}")  # True, pois 5 < 10

print("\n" + "=" * 50 + "\n")

# Operadores de maior ou igual / menor ou igual
print(f"a >= b? {a >= b}")  # True, pois são iguais (5 >= 5)
print(f"a * 2 >= c? {a * 2 >= c}")  # True, pois 10 >= 10
print(f"a <= c? {a <= c}")  # True, pois 5 <= 10
````

### Armazenando Resultados em Variáveis

Os resultados das comparações (valores booleanos) podem ser armazenados em variáveis para uso posterior em lógica condicional.

````python
# Declaração de variáveis
n1 = 10
n2 = 20

# Armazenando o resultado de comparações em variáveis
iguais = n1 == n2
diferentes = n1 != n2
maior = n1 > n2

print(f"n1 = {n1}, n2 = {n2}")
print(f"São iguais? {iguais}")  # False
print(f"São diferentes? {diferentes}")  # True
print(f"n1 é maior que n2? {maior}")  # False
````

### Comparador Interativo de Números

Vamos criar um programa que solicita dois números ao usuário e exibe o resultado de várias comparações entre eles.

````python
# Inicialização das variáveis de resultado
x = y = z = False
n1 = n2 = 0

# Solicitando entrada do usuário com tratamento de erro
try:
    n1 = int(input("Digite um número: "))
    n2 = int(input("Digite outro número: "))
    
    print("\n" + "=" * 50)
    print(f"Comparando {n1} e {n2}")
    print("=" * 50 + "\n")
    
    # Comparação de igualdade
    x = n1 == n2
    print(f"São iguais? {x}\n")
    
    # Comparação de diferença
    y = n1 != n2
    print(f"São diferentes? {y}\n")
    
    # Comparação de magnitude (maior que)
    z = n1 > n2
    print(f"{n1} é maior que {n2}? {z}\n")
    
    # Comparações adicionais
    print(f"{n1} é menor que {n2}? {n1 < n2}\n")
    print(f"{n1} é maior ou igual a {n2}? {n1 >= n2}\n")
    print(f"{n1} é menor ou igual a {n2}? {n1 <= n2}")
    
except ValueError:
    print("Erro: Por favor, digite apenas números inteiros válidos.")
````

## 9. Operadores Lógicos AND, OR, NOT

### Operadores Lógicos em Python

Operadores lógicos são usados para combinar expressões condicionais e **sempre retornam valores booleanos** (`True` ou `False`).

**Os três operadores lógicos:**

* **and** - retorna `True` apenas se **ambas** as condições forem verdadeiras
* **or** - retorna `True` se **pelo menos uma** das condições for verdadeira
* **not** - **inverte** o valor booleano (True vira False, False vira True)

````python
# Exemplo: Verificar se uma pessoa pode participar de um evento
# Requisitos: ter pelo menos 18 anos E ter pelo menos 1.75m de altura

idade = 15
altura = 1.75

# Operador AND: ambas as condições precisam ser True
resultado = (idade >= 18) and (altura >= 1.75)

# Construindo e exibindo a mensagem
msg = "Pode participar do evento? " + str(resultado)
print(msg)
print(f"\nExplicação: idade ({idade}) >= 18? {idade >= 18}")
print(f"Explicação: altura ({altura}) >= 1.75? {altura >= 1.75}")
print(f"AND: Ambas verdadeiras? {resultado}")
````

### Operador AND - Tabela Verdade

O operador `and` só retorna `True` quando **ambas** as condições são verdadeiras:

| Condição A | Condição B | A and B |
| --- | --- | --- |
| True | True | **True** |
| True | False | False |
| False | True | False |
| False | False | False |

### Operador OR (OU)

O operador `or` retorna `True` se **pelo menos uma** das condições for verdadeira. É usado quando queremos verificar se qualquer uma de várias condições é satisfeita.

````python
# Exemplo: Sistema de alarme
# O alarme dispara se a porta OU a janela estiverem abertas
# "a" = aberto, "f" = fechado

porta = "a"
janela = "f"

# Operador OR: pelo menos uma condição precisa ser True
alarme = (porta == "a") or (janela == "a")

# Construindo e exibindo a mensagem
msg = "Alarme ativado? " + str(alarme)
print(msg)
print(f"\nExplicação: porta aberta? {porta == 'a'}")
print(f"Explicação: janela aberta? {janela == 'a'}")
print(f"OR: Pelo menos uma aberta? {alarme}")
````

### Operador OR - Tabela Verdade

O operador `or` retorna `True` quando **pelo menos uma** condição é verdadeira:

| Condição A | Condição B | A or B |
| --- | --- | --- |
| True | True | **True** |
| True | False | **True** |
| False | True | **True** |
| False | False | False |

### Operador NOT (NÃO)

O operador `not` **inverte** o valor booleano:
* `not True` retorna `False`
* `not False` retorna `True`

É usado para negar uma condição ou inverter o estado de uma variável booleana.

````python
# Exemplo: Inverter o estado de uma variável booleana
# Inicialmente a pessoa não tem dinheiro, mas depois recebe

tem_dinheiro = False
print(f"Estado inicial: tem dinheiro? {tem_dinheiro}")

# Operador NOT: inverte o valor booleano
tem_dinheiro = not tem_dinheiro

# Exibindo o novo estado
msg = "Tem dinheiro? " + str(tem_dinheiro)
print(msg)
print(f"\nExplicação: NOT False = {not False}")
print(f"Explicação: NOT True = {not True}")
````

### Operador NOT - Tabela Verdade

O operador `not` simplesmente **inverte** o valor:

| Condição A | not A |
| --- | --- |
| True | **False** |
| False | **True** |

### Combinando Múltiplos Operadores Lógicos

Podemos combinar AND, OR e NOT em expressões mais complexas. Use parênteses para controlar a ordem de avaliação.

**Ordem de precedência:**
1. **not** (maior precedência)
2. **and**
3. **or** (menor precedência)

````python
# Exemplo complexo: Acesso a um sistema
# Regra: (usuário autenticado E tem permissão) OU (é administrador)

usuario_autenticado = True
tem_permissao = False
e_administrador = False

# Combinando operadores lógicos
acesso_permitido = (usuario_autenticado and tem_permissao) or e_administrador

print("=" * 50)
print("Sistema de Controle de Acesso")
print("=" * 50)
print(f"Usuário autenticado? {usuario_autenticado}")
print(f"Tem permissão? {tem_permissao}")
print(f"É administrador? {e_administrador}")
print(f"\nAcesso permitido? {acesso_permitido}")

# Testando com NOT
print(f"\nAcesso NEGADO? {not acesso_permitido}")
````

**Critérios:**
* Ter renda maior ou igual a R$ 2000
* TER score de crédito acima de 600 OU ter um avalista
* NÃO ter restrições no CPF

````python
# Dados do solicitante
renda = 2500
score_credito = 550
tem_avalista = True
tem_restricoes = False

print("=" * 50)
print("Sistema de Análise de Crédito")
print("=" * 50)
print(f"Renda: R$ {renda}")
print(f"Score de crédito: {score_credito}")
print(f"Tem avalista? {tem_avalista}")
print(f"Tem restrições no CPF? {tem_restricoes}")
print("\n" + "=" * 50)

# Verificando cada critério
renda_suficiente = renda >= 2000
score_ok = score_credito > 600
score_ou_avalista = score_ok or tem_avalista
sem_restricoes = not tem_restricoes

# Decisão final: TODAS as condições devem ser verdadeiras
emprestimo_aprovado = renda_suficiente and score_ou_avalista and sem_restricoes

# Exibindo análise detalhada
print("Análise dos Critérios:")
print(f"✓ Renda suficiente (>= R$ 2000)? {renda_suficiente}")
print(f"✓ Score > 600 OU tem avalista? {score_ou_avalista}")
print(f"  - Score > 600? {score_ok}")
print(f"  - Tem avalista? {tem_avalista}")
print(f"✓ Sem restrições no CPF? {sem_restricoes}")
print("\n" + "=" * 50)
print(f"EMPRÉSTIMO APROVADO? {emprestimo_aprovado}")
print("=" * 50)
````

## 10. Desvios Condicionais - if, elif, else

### Estruturas Condicionais: if, elif, else

As estruturas condicionais permitem que o programa tome **decisões** e execute diferentes blocos de código baseados em condições.

**Sintaxe básica:**

```python
if condição:
    # código executado se a condição for True
elif outra_condição:

else:
    # código executado se todas as condições anteriores forem False
```

**Regras importantes:**

* O `elif` e `else` são opcionais
* Pode haver múltiplos `elif`, mas apenas um `else`
* Os dois-pontos (`:`) são obrigatórios após cada condição

### Sistema de Aprovação de Aluno

Vamos criar um programa que determina se um aluno foi **aprovado**, está em **recuperação** ou foi **reprovado** baseado em:

**Critérios de aprovação:**
* **Aprovado**: média >= 7 **E** menos de 10 faltas
* **Recuperação**: média >= 5 **E** menos de 10 faltas
* **Reprovado**: qualquer outra situação (média < 5 OU faltas >= 10)

````python
# Inicialização das variáveis
n1 = n2 = 0.0
media = 0.0

# Solicitando entrada do usuário
try:
    n1 = float(input("Digite a primeira nota: "))
    n2 = float(input("Digite a segunda nota: "))
    faltas = int(input("Digite o número de faltas: "))
    
    # Calculando a média aritmética das duas notas
    media = (n1 + n2) / 2
    
    print(f"\nNotas: {n1} e {n2}")
    print(f"Média calculada: {media:.2f}")
    print(f"Número de faltas: {faltas}")
    print("\n" + "=" * 50)
    
except ValueError:
    print("Erro: Por favor, digite valores numéricos válidos.")
````

````python
# Estrutura condicional para determinar a situação do aluno
# Usa operadores lógicos (and) para combinar múltiplas condições

if (media >= 7) and (faltas < 10):
    # Primeira condição: média suficiente E poucas faltas
    print("✓ APROVADO")
    print("Parabéns! Você atingiu a média necessária e teve frequência adequada.")
    
elif (media >= 5) and (faltas < 10):
    # Segunda condição: média intermediária E poucas faltas
    print("⚠ RECUPERAÇÃO")
    print("Você precisará fazer a prova de recuperação.")
    
else:
    # Qualquer outra situação: média insuficiente OU muitas faltas
    print("✗ REPROVADO")
    if faltas >= 10:
        print("Motivo: Número de faltas excedeu o limite permitido.")
    if media < 5:
        print("Motivo: Média abaixo do mínimo necessário.")

# Exibindo resumo final
print("\n" + "=" * 50)
print(f"Resumo: Média = {media:.2f} | Faltas = {faltas}")
print("=" * 50)
````

### Fluxo de Execução

A estrutura condicional é avaliada **sequencialmente de cima para baixo**:

1. **if**: Avalia a primeira condição
   * Se `True`: executa o bloco e **pula** o restante
   * Se `False`: continua para o próximo

2. **elif**: Avalia a segunda condição (só se o `if` foi `False`)
   * Se `True`: executa o bloco e **pula** o restante
   * Se `False`: continua para o próximo

3. **else**: Executado apenas se **todas** as condições anteriores foram `False`

**Importante:** Apenas **um** bloco será executado, mesmo que múltiplas condições sejam verdadeiras!

## 13. Laços de Repetição While

### Laço de Repetição While

O laço `while` executa um bloco de código **repetidamente enquanto** uma condição for verdadeira.

**Sintaxe básica:**
```python
while condição:
    # código executado enquanto a condição for True
    # atualização da condição (importante para evitar loop infinito)
```

**Componentes essenciais:**
1. **Condição**: Expressão booleana avaliada antes de cada iteração
2. **Corpo do loop**: Código indentado que será repetido
3. **Atualização**: Modificação da variável de controle para eventualmente tornar a condição False

**⚠️ Cuidado:** Se a condição nunca se tornar False, o loop será **infinito**!

````python
# Laço while básico: contador de 1 a 10

# Inicialização: variável de controle do loop
num = 1

# Loop: enquanto num for menor ou igual a 10
while (num <= 10):
    print(num)  # Exibe o número atual
    num += 1    # Incrementa o contador (essencial para evitar loop infinito)

# Após sair do loop (quando num > 10)
print("Finalizado!")
````

### Como o While Funciona

**Fluxo de execução:**
1. **Avalia a condição** antes de cada iteração
2. Se **True**: executa o corpo do loop e retorna ao passo 1
3. Se **False**: **sai do loop** e continua com o código após o while

**Diferença entre while e if:**
* `if`: executa o bloco **uma vez** se a condição for True
* `while`: executa o bloco **repetidamente** enquanto a condição for True

### Operadores Compostos de Atribuição

É comum usar operadores compostos para atualizar variáveis em loops:

* `num += 1` → equivalente a `num = num + 1` (incremento)
* `num -= 1` → equivalente a `num = num - 1` (decremento)
* `num *= 2` → equivalente a `num = num * 2` (multiplicação)
* `num /= 2` → equivalente a `num = num / 2` (divisão)

**Uso comum em loops:**
* `num += 1` - incrementa de 1 em 1
* `num += 2` - incrementa de 2 em 2
* `num *= 2` - dobra o valor a cada iteração

````python
# Exemplo 1: Contador decrescente
print("Contagem regressiva:")
contador = 10

while contador > 0:
    print(contador)
    contador -= 1  # Decremento: subtrai 1 a cada iteração

print("Lançamento!\n")

# Exemplo 2: Incremento de 2 em 2
print("Números pares de 2 a 10:")
par = 2

while par <= 10:
    print(par)
    par += 2  # Incrementa de 2 em 2

print("\nFinalizado!")
````

### Loop Infinito com `while True`

Um loop `while True` cria um **loop infinito** que só termina quando encontra:
* Um comando `break` (sai do loop imediatamente)
* Um comando `return` (em funções)
* Uma exceção/erro

**Sintaxe:**
```python
while True:
    # código
    if condição_de_saída:
        break  # sai do loop
```

**Uso comum:**
* Menus interativos
* Validação de entrada até ser válida
* Jogos (loop principal)
* Servidores (aguardando conexões)

````python
# Loop infinito com comando break
# Útil para menus e validação de entrada

nome = None

print("Sistema de Cadastro")
print("=" * 40)

while True:
    # Solicita entrada do usuário
    nome = input("Digite o seu nome ou 'x' para sair: ")
    
    # Verifica condição de saída
    if nome == 'x' or nome == 'X':
        print("Programa encerrado.")
        break  # Sai do loop imediatamente
    
    # Processa a entrada válida
    print(f"Bem-vindo(a), {nome}!\n")
````

### O Comando `continue`

O comando `continue` **pula** o restante da iteração atual e **volta ao início do loop** para avaliar a condição novamente.

**Diferença entre break e continue:**
* `break`: **sai completamente** do loop
* `continue`: **pula** para a próxima iteração

**Uso comum:**
* Pular valores indesejados
* Ignorar casos especiais
* Validação condicional dentro do loop

````python
# Exemplo: Imprimindo apenas números ímpares de 1 a 10
print("Números ímpares de 1 a 10:")

num = 0

while num < 10:
    num += 1  # Incrementa primeiro
    
    # Se o número for par, pula para a próxima iteração
    if num % 2 == 0:
        continue  # Volta ao início do while (não executa o print)
    
    # Este código só é executado para números ímpares
    print(num)

print("\nFinalizado!")
````

### Break vs Continue - Exemplo Visual

**Com `break`:**
```python
while condição:
    if algo_especial:
        break  # SAI DO LOOP
    código_normal
# Continua aqui após o break
```

**Com `continue`:**
```python
while condição:
    if pular_este:
        continue  # VOLTA AO INÍCIO
    código_normal  # Não executado se continue foi chamado
```

**Resumo:**
* `break` = "**Pare tudo** e saia do loop"
* `continue` = "**Pule esta iteração** e vá para a próxima"

````python
# Demonstração: Break vs Continue

print("Exemplo com CONTINUE (pula apenas o 5):")
num = 0
while num < 8:
    num += 1
    if num == 5:
        print(f"  [Pulando o {num}]")
        continue  # Pula o 5, mas continua o loop
    print(num)

print("\n" + "=" * 40 + "\n")

print("Exemplo com BREAK (para no 5):")
num = 0
while num < 8:
    num += 1
    if num == 5:
        print(f"  [Parando no {num}]")
        break  # Para completamente no 5
    print(num)

print("\nLoop encerrado!")
````

### ⚠️ Armadilhas Comuns e Boas Práticas

**1. Loop Infinito Acidental**
```python
# ❌ ERRADO - loop infinito!
num = 1
while num <= 10:
    print(num)
    # Esqueceu de incrementar num!
```

**2. Sempre Atualize a Variável de Controle**
```python
# ✅ CORRETO
num = 1
while num <= 10:
    print(num)
    num += 1  # Atualização essencial
```

**3. Cuidado com Loops Infinitos Intencionais**
* Use `while True` apenas quando tiver uma condição clara de saída (`break`)
* Sempre garanta que o `break` será alcançado eventualmente

**4. Validação de Entrada**
* Use `while True` com `break` para validar entradas do usuário
* Continue solicitando até obter entrada válida

````python
# Exemplo prático: Validação de entrada com while
# Solicita um número entre 1 e 10 até que seja válido

print("Digite um número entre 1 e 10:")
print("=" * 40)

# Loop infinito até obter entrada válida
while True:
    try:
        numero = int(input("Número: "))
        
        # Verifica se está no intervalo válido
        if 1 <= numero <= 10:
            print(f"\n✓ Entrada válida! Você digitou: {numero}")
            break  # Sai do loop quando a entrada for válida
        else:
            print("✗ Erro: O número deve estar entre 1 e 10. Tente novamente.\n")
            
    except ValueError:
        print("✗ Erro: Digite apenas números inteiros. Tente novamente.\n")

print("Programa finalizado!")
````

### Exercício: Tabuada Interativa

Crie um programa que:
1. Solicita um número ao usuário
2. Exibe a tabuada desse número (1 a 10)
3. Pergunta se deseja continuar
4. Repete até que o usuário digite 'n' para sair

**Requisitos:**
* Use `while True` para o menu principal
* Use outro `while` para calcular a tabuada

````python
# Programa: Tabuada Interativa

print("=" * 50)
print("CALCULADORA DE TABUADA")
print("=" * 50)

# Loop principal do programa
while True:
    try:
        # Solicita o número para calcular a tabuada
        numero = int(input("\nDigite um número para ver sua tabuada: "))
        
        print(f"\nTabuada do {numero}:")
        print("-" * 30)
        
        # Loop para calcular e exibir a tabuada
        multiplicador = 1
        while multiplicador <= 10:
            resultado = numero * multiplicador
            print(f"{numero} x {multiplicador:2d} = {resultado:3d}")
            multiplicador += 1
        
        print("-" * 30)
        
        # Pergunta se deseja continuar
        continuar = input("\nDeseja calcular outra tabuada? (s/n): ").lower()
        
        if continuar == 'n':
            print("\nPrograma encerrado. Até logo!")
            break  # Sai do loop principal
        elif continuar != 's':
            print("Opção inválida. Encerrando...")
            break
            
    except ValueError:
        print("\n✗ Erro: Digite apenas números inteiros válidos.")
        continuar = input("Deseja tentar novamente? (s/n): ").lower()
        if continuar == 'n':
            print("\nPrograma encerrado.")
            break
````

```python
for variavel in sequencia:
```

## 15. Encadeamento de Laços de Repetição

### Laços de Repetição Aninhados (Nested Loops)

Laços aninhados são laços **dentro de outros laços**. O laço interno é executado completamente a **cada iteração** do laço externo.

**Estrutura básica:**
```python
for externo in sequencia_externa:
    # Código do laço externo
    for interno in sequencia_interna:
        # Código do laço interno (executa N vezes para cada iteração externa)
    # Continua o laço externo
```

**Como funciona:**
* Para **cada** iteração do laço externo
* O laço interno executa **todas** as suas iterações
* Depois volta para a próxima iteração do laço externo

**Aplicações comuns:**
* Processar matrizes (tabelas 2D)
* Criar padrões e desenhos
* Gerar combinações
* Percorrer estruturas bidimensionais

````python
# Exemplo básico: laço dentro de outro laço
# Laço externo: rodadas de 1 a 5
# Laço interno: contagem regressiva de 5 a 1

print("DEMONSTRAÇÃO: LAÇOS ANINHADOS")
print("=" * 50)

# Laço externo: controla as rodadas
for cont_ext in range(1, 6):
    print(f"\n🔹 Rodada {cont_ext}")
    print("-" * 30)
    
    # Laço interno: executa completamente a cada rodada
    # Conta de 5 até 1 (range com step negativo)
    for cont_int in range(5, 0, -1):
        print(f"  Valor: {cont_int}")

# Após todos os laços terminarem
print("\n" + "=" * 50)
print("✓ Fim dos laços")
print(f"\nTotal de iterações do laço externo: 5")
print(f"Total de iterações do laço interno: 5 × 5 = 25")
````

### Fluxo de Execução

**Exemplo com números pequenos:**
```python
for i in range(1, 4):      # Externo: 1, 2, 3
    for j in range(1, 3):  # Interno: 1, 2
        print(i, j)
```

**Saída:**
```
1 1  ← i=1, j executa: 1, 2
1 2
2 1  ← i=2, j executa: 1, 2
2 2
3 1  ← i=3, j executa: 1, 2
3 2
```

**Observação importante:**
* Se o laço externo tem **N** iterações
* E o laço interno tem **M** iterações
* O código dentro do laço interno executa **N × M** vezes

````python
# Visualização clara: mostrando cada iteração dos laços aninhados

print("VISUALIZAÇÃO: LAÇOS ANINHADOS")
print("=" * 50)

contador_total = 0  # Conta o total de execuções do laço interno

# Laço externo: 3 iterações
for externo in range(1, 4):
    print(f"\n>>> LAÇO EXTERNO: iteração {externo}")
    
    # Laço interno: 4 iterações
    for interno in range(1, 5):
        contador_total += 1
        print(f"    LAÇO INTERNO: iteração {interno} (execução #{contador_total} no total)")

print("\n" + "=" * 50)
print(f"Resumo: Laço externo executou 3 vezes")
print(f"        Laço interno executou 4 vezes por cada externa = 12 vezes no total")
````

### Gerando Conjuntos de Dados Aleatórios

Laços aninhados são úteis para gerar **múltiplos conjuntos** de dados, onde:

**Caso de uso:**

### Laços Aninhados para Tabelas

Uma aplicação clássica de laços aninhados é criar **tabelas** e **matrizes**:

**Padrão comum:**
```python
for linha in range(linhas):
    for coluna in range(colunas):
        # processa célula (linha, coluna)
```

**Exemplos:**

````python
# Exemplo: Criando uma tabela de multiplicação

print("TABELA DE MULTIPLICAÇÃO (1 a 5)")
print("=" * 50)

# Cabeçalho da tabela
print("\n    ", end="")
for coluna in range(1, 6):
    print(f"{coluna:4d}", end="")
print("\n" + "-" * 30)

# Laço externo: linhas (primeiro número)
for linha in range(1, 6):
    print(f"{linha} | ", end="")  # Rótulo da linha
    
    # Laço interno: colunas (segundo número)
    for coluna in range(1, 6):
        resultado = linha * coluna
        print(f"{resultado:4d}", end="")  # :4d = 4 espaços de largura
    
    print()  # Nova linha após cada linha da tabela

print("=" * 50)
````

### Break e Continue com Laços Aninhados


### Desafio: Encontrar Números Primos com Laços Aninhados

Use laços aninhados para encontrar todos os números primos entre 2 e 50.

**Lógica:**
* Laço externo: percorre cada número de 2 a 50
* Laço interno: testa se o número é divisível por algum valor
* Se não for divisível por nenhum (exceto 1 e ele mesmo), é primo

````python
# Programa: Encontrar Números Primos com Laços Aninhados

print("=" * 50)
print("NÚMEROS PRIMOS ENTRE 2 E 50")
print("=" * 50)

primos_encontrados = []  # Lista para armazenar os primos

# Laço externo: testa cada número de 2 a 50
for numero in range(2, 51):
    eh_primo = True  # Assume que é primo inicialmente
    
    # Laço interno: testa divisibilidade
    # Testa divisores de 2 até a raiz quadrada do número (otimização)
    for divisor in range(2, int(numero ** 0.5) + 1):
        if numero % divisor == 0:
            # Encontrou um divisor, não é primo
            eh_primo = False
            break  # Não precisa testar mais divisores
    
    # Se passou por todos os testes, é primo
    if eh_primo:
        primos_encontrados.append(numero)

# Exibindo os resultados
print("\nNúmeros primos encontrados:")
print("-" * 50)

# Exibe em linhas de 10 números
for i, primo in enumerate(primos_encontrados):
    print(f"{primo:3d}", end="  ")
    if (i + 1) % 10 == 0:  # Nova linha a cada 10 números
        print()

print("\n\n" + "=" * 50)
print(f"Total de números primos encontrados: {len(primos_encontrados)}")
print("=" * 50)
````

## 19. Mini-Projeto: Jogo de Advinhação

- O programa gera um número aleatório entre 1 e 15.
- O usuário deve tentar adivinhar o número gerado.
- Após cada tentativa, o programa informa se o valor está abaixo ou acima do número sorteado.
- Se o usuário acertar, exibe: **'Parabéns, você acertou!'**
- O usuário tem até **3 tentativas** por jogo.
- Pressione **'s'** para iniciar um novo jogo.

<!--
Comentários:
- O objetivo é praticar estruturas de repetição, condicionais e interação com o usuário.
- O uso de comentários no código é essencial para documentação e compreensão.
-->

````python
import random
import os

# Variável para controlar se o usuário deseja jogar novamente
novo_jogo = "s"

# Loop principal do jogo: repete enquanto o usuário quiser jogar novamente
while novo_jogo == "s":
    print(f"Bem vindo ao jogo de adivinhação!")
    print(f"Escolha um número entre 1 e 10.")
    print(f"Você tem 3 chances para acertar o número.")
    print(f"Boa sorte!")

    # Gera o número secreto aleatório entre 1 e 10
    numero_secreto = random.randint(1, 10)
    chances = 3  # Define o número de chances

    print(f"O número secreto foi gerado. Agora é sua vez de adivinhar!")
    # Loop para as tentativas do usuário
    while chances > 0:
        chute = int(input(f"\nDigite seu chute: "))  # Solicita o chute do usuário
        if chute == numero_secreto:
            print(f"\nParabéns! Você acertou o número secreto!")  # Usuário acertou
            break
        elif chute < numero_secreto:
            print(f"\nSeu chute foi menor que o número secreto.")  # Chute menor
        else:
            print(f"\nSeu chute foi maior que o número secreto.")  # Chute maior
        chances -= 1  # Decrementa as chances
        if chances == 0:
            print(f"\nSuas chances acabaram. O número secreto era {numero_secreto}.")  # Usuário perdeu
            break
    # Pergunta se o usuário deseja jogar novamente
    novo_jogo = input(f"\nDeseja jogar novamente? (s/n):")
    novo_jogo = novo_jogo.lower()

    os.system("clear")  # Limpa a tela (pode não funcionar em todos os sistemas)
# Esta célula implementa um jogo de adivinhação simples, onde o usuário tenta adivinhar um número aleatório gerado pelo computador. O jogo oferece três chances para acertar o número. Após cada tentativa, o usuário é informado se o chute foi maior ou menor que o número secreto. Se o usuário acertar o número, ele ganha uma mensagem de parabéns. Se o usuário não acertar o número em três tentativas, ele perde o jogo e é informado qual era o número secreto. O jogo pode ser reiniciado pelo usuário, permitindo jogos múltiplos.
````
