# Laço While em Python

Este notebook cobre o uso do laço `while` para repetição controlada por condição:

* **While básico** - Estrutura fundamental do laço
* **Condições de parada** - Como controlar a execução
* **break** - Interromper o laço prematuramente
* **continue** - Pular para a próxima iteração
* **While-else** - Bloco executado ao término normal
* **Loops infinitos** - Uso controlado de while True
* **Contadores e acumuladores** - Padrões comuns

O laço `while` é fundamental para:
* Repetir código enquanto uma condição for verdadeira
* Processar dados até encontrar uma condição de parada
* Implementar menus e interfaces iterativas
* Validar entradas do usuário

## 1. Laço While Básico

O `while` executa um bloco de código **repetidamente** enquanto uma condição for verdadeira.

**Sintaxe:**
```python
while condicao:
    # código executado enquanto condição for True
    comando1
    comando2
```

**Como funciona:**
1. Avalia a condição
2. Se `True`, executa o bloco e retorna ao passo 1
3. Se `False`, encerra o laço e continua o programa

**Importante:**
* A condição deve **eventualmente se tornar False**
* Sem isso, cria-se um **loop infinito**
* Use uma variável de controle que muda a cada iteração

````python
# Exemplo 1: Contar de 1 a 10
contador = 1

while contador <= 10:
    print(contador)
    contador += 1  # Incrementa para evitar loop infinito

print("Laço terminado!")
print()

# Exemplo 2: Contagem regressiva
numero = 5

print("Contagem regressiva:")
while numero > 0:
    print(numero)
    numero -= 1
print("🚀 Lançamento!")
print()

# Exemplo 3: Somatório
soma = 0
i = 1

while i <= 5:
    soma += i
    print(f"Adicionando {i}, soma parcial: {soma}")
    i += 1

print(f"\nSoma total de 1 a 5: {soma}")
print()

# Exemplo 4: Tabuada
numero = 7
multiplicador = 1

print(f"Tabuada do {numero}:")
while multiplicador <= 10:
    resultado = numero * multiplicador
    print(f"{numero} x {multiplicador} = {resultado}")
    multiplicador += 1
````

## 2. Comandos break e continue

### break
* **Interrompe** o laço imediatamente
* Sai do while e continua o programa após o laço
* Útil para condições de parada especiais

### continue
* **Pula** para a próxima iteração
* Ignora o resto do código no bloco atual
* Volta a avaliar a condição do while

**Quando usar:**
* `break`: Quando encontrar condição de saída antecipada
* `continue`: Quando quiser pular certas iterações

````python
# Exemplo 1: break - Procurar um número
numeros = [10, 25, 30, 45, 50, 60]
i = 0
alvo = 30

print(f"Procurando o número {alvo}...")
while i < len(numeros):
    print(f"Verificando posição {i}: {numeros[i]}")
    if numeros[i] == alvo:
        print(f"✅ Encontrado na posição {i}!")
        break
    i += 1
else:
    print(f"❌ Número {alvo} não encontrado")

print()

# Exemplo 2: continue - Pular números pares
contador = 0

print("Números ímpares de 1 a 10:")
while contador < 10:
    contador += 1
    if contador % 2 == 0:
        continue  # Pula números pares
    print(contador)

print()

# Exemplo 3: break com condição de erro
soma = 0
numero = 1

print("Somando números até ultrapassar 50:")
while numero <= 20:
    soma += numero
    print(f"Soma: {soma} (adicionado {numero})")
    
    if soma > 50:
        print("⚠️ Limite de 50 ultrapassado! Interrompendo...")
        break
    
    numero += 1

print(f"Soma final: {soma}")
print()

# Exemplo 4: continue - Filtrar valores
i = 0
valores = [5, -3, 8, -1, 12, -7, 15]
soma_positivos = 0

print("Somando apenas valores positivos:")
while i < len(valores):
    valor = valores[i]
    i += 1
    
    if valor < 0:
        print(f"Pulando {valor} (negativo)")
        continue
    
    soma_positivos += valor
    print(f"Adicionando {valor}, soma: {soma_positivos}")

print(f"\nSoma total dos positivos: {soma_positivos}")
````

## 3. Cláusula while-else

O `else` em um while é executado **somente se** o laço terminar normalmente (sem `break`).

**Sintaxe:**
```python
while condicao:
    comando1
    if condicao_especial:
        break
else:
    # Executado apenas se o while terminou sem break
    comando2
```

**Quando usar:**
* Para distinguir saída normal de saída por `break`
* Comum em buscas: else indica "não encontrado"
* Validações: else indica "todas as condições passaram"

````python
# Exemplo 1: Busca com while-else
lista = [10, 20, 30, 40, 50]
procurado = 35
i = 0

print(f"Procurando {procurado} na lista...")
while i < len(lista):
    if lista[i] == procurado:
        print(f"✅ Encontrado na posição {i}!")
        break
    i += 1
else:
    # Executado apenas se não houve break
    print(f"❌ {procurado} não foi encontrado na lista")

print()

# Exemplo 2: Validação de senha (simulada)
tentativas = 0
max_tentativas = 3
senha_correta = "python123"
senhas_testadas = ["abc", "123", "senha"]  # Simula entrada do usuário

print(f"Sistema de login (máx. {max_tentativas} tentativas)")
while tentativas < max_tentativas:
    senha = senhas_testadas[tentativas] if tentativas < len(senhas_testadas) else "errada"
    tentativas += 1
    
    print(f"Tentativa {tentativas}: Testando senha '{senha}'...")
    
    if senha == senha_correta:
        print("✅ Login bem-sucedido!")
        break
    else:
        print("❌ Senha incorreta")
else:
    # Executado se esgotou as tentativas sem break
    print(f"\n🚫 Conta bloqueada após {max_tentativas} tentativas falhadas")

print()

# Exemplo 3: Verificar se todos são positivos
numeros = [5, 10, 15, 20, 25]
i = 0

print("Verificando se todos os números são positivos...")
while i < len(numeros):
    if numeros[i] <= 0:
        print(f"❌ Encontrado número não-positivo: {numeros[i]}")
        break
    print(f"✓ {numeros[i]} é positivo")
    i += 1
else:
    print("\n✅ Todos os números são positivos!")

print()

# Exemplo 4: Número primo
numero = 17
divisor = 2
eh_primo = True

if numero < 2:
    eh_primo = False
else:
    while divisor <= numero // 2:
        if numero % divisor == 0:
            print(f"{numero} é divisível por {divisor}")
            eh_primo = False
            break
        divisor += 1
    else:
        # Executado se não encontrou divisores
        print(f"{numero} não tem divisores além de 1 e ele mesmo")

if eh_primo:
    print(f"✅ {numero} é primo!")
else:
    print(f"❌ {numero} não é primo")
````

## 4. While True (Loop Infinito Controlado)

`while True` cria um **loop infinito** que deve ser terminado com `break`.

**Sintaxe:**
```python
while True:
    comando1
    if condicao_saida:
        break
    comando2
```

**Quando usar:**
* Menus interativos
* Servidores e processos contínuos
* Quando a condição de parada está no meio da lógica
* Validações de entrada repetidas

**Importante:**
* **SEMPRE** tenha uma forma de sair (`break`)
* Sem `break`, o programa trava em loop infinito

````python
# Exemplo 1: Menu simples (simulado sem input)
opcoes_simuladas = ['1', '2', '3', '4']  # Simula escolhas do usuário
indice_opcao = 0

print("Sistema de Menu (simulação)")
print()

while True:
    print("\n=== MENU ===")
    print("1. Ver saldo")
    print("2. Depositar")
    print("3. Sacar")
    print("4. Sair")
    
    # Simula entrada do usuário
    if indice_opcao < len(opcoes_simuladas):
        opcao = opcoes_simuladas[indice_opcao]
        indice_opcao += 1
    else:
        opcao = '4'  # Força saída
    
    print(f"Escolha: {opcao}")
    
    if opcao == '1':
        print("💰 Saldo: R$ 1.000,00")
    elif opcao == '2':
        print("✅ Depósito realizado")
    elif opcao == '3':
        print("✅ Saque realizado")
    elif opcao == '4':
        print("\n👋 Até logo!")
        break
    else:
        print("❌ Opção inválida!")

print()

# Exemplo 2: Processar até condição
contador = 0
soma = 0

print("Somando números até ultrapassar 100:")
while True:
    contador += 1
    soma += contador
    print(f"Contador: {contador}, Soma: {soma}")
    
    if soma > 100:
        print(f"\n✅ Soma ultrapassou 100! Parando...")
        break

print(f"Total de números somados: {contador}")
print(f"Soma final: {soma}")
print()

# Exemplo 3: Validar entrada (simulado)
entradas_simuladas = [-5, 0, 150, 42]  # Simula entradas do usuário
indice_entrada = 0

print("Digite um número entre 1 e 100 (simulação):")
while True:
    numero = entradas_simuladas[indice_entrada]
    indice_entrada += 1
    print(f"Entrada: {numero}")
    
    if 1 <= numero <= 100:
        print(f"✅ Número válido: {numero}")
        break
    else:
        print("❌ Número inválido! Deve estar entre 1 e 100.")
        if indice_entrada >= len(entradas_simuladas):
            print("(Fim da simulação)")
            break

print()

# Exemplo 4: Coletar nomes (simulado, baseado no código original)
nomes_simulados = ["Maria", "João", "Ana", "x"]  # Simula entradas
nomes_coletados = []
indice_nome = 0

print("Coletando nomes (digite 'x' para parar):")
while True:
    nome = nomes_simulados[indice_nome] if indice_nome < len(nomes_simulados) else "x"
    indice_nome += 1
    
    print(f"Nome digitado: {nome}")
    
    if nome.lower() == "x":
        break
    
    nomes_coletados.append(nome)
    print(f"Bem-vindo(a), {nome}!")

print(f"\n👋 Até logo!")
print(f"Total de nomes coletados: {len(nomes_coletados)}")
print(f"Nomes: {', '.join(nomes_coletados)}")
````

## 5. Exemplo Consolidado: Jogo de Adivinhação

Vamos criar um jogo completo que usa várias técnicas de while.

````python
# Jogo de Adivinhação de Números
# Demonstra while, break, continue e else

import random

# Configurações do jogo
numero_secreto = random.randint(1, 100)
max_tentativas = 7
tentativa = 0
acertou = False

# Simula palpites do jogador (para demonstração)
palpites_simulados = [50, 75, 60, 70, 65, 68, 67]  # Chega em 67

print("=" * 60)
print("JOGO DE ADIVINHAÇÃO")
print("=" * 60)
print(f"Adivinhe o número entre 1 e 100!")
print(f"Você tem {max_tentativas} tentativas.")
print(f"(Número secreto: {numero_secreto})\n")  # Mostra para demonstração

# Loop principal do jogo
while tentativa < max_tentativas:
    tentativa += 1
    
    # Simula entrada do jogador
    if tentativa - 1 < len(palpites_simulados):
        palpite = palpites_simulados[tentativa - 1]
    else:
        palpite = random.randint(1, 100)
    
    print(f"\nTentativa {tentativa}/{max_tentativas}")
    print(f"Seu palpite: {palpite}")
    
    # Validação de entrada
    if palpite < 1 or palpite > 100:
        print("❌ Número inválido! Deve estar entre 1 e 100.")
        tentativa -= 1  # Não conta como tentativa válida
        continue
    
    # Verifica o palpite
    if palpite == numero_secreto:
        print("\n" + "=" * 60)
        print("🎉 PARABÉNS! VOCÊ ACERTOU! 🎉")
        print("=" * 60)
        print(f"O número secreto era: {numero_secreto}")
        print(f"Você acertou em {tentativa} tentativa(s)!")
        acertou = True
        break
    elif palpite < numero_secreto:
        diferenca = numero_secreto - palpite
        if diferenca <= 5:
            print("🔥 Muito quente! O número é MAIOR")
        elif diferenca <= 15:
            print("🌡️ Quente! O número é MAIOR")
        else:
            print("❄️ Frio! O número é MAIOR")
    else:  # palpite > numero_secreto
        diferenca = palpite - numero_secreto
        if diferenca <= 5:
            print("🔥 Muito quente! O número é MENOR")
        elif diferenca <= 15:
            print("🌡️ Quente! O número é MENOR")
        else:
            print("❄️ Frio! O número é MENOR")
    
    # Mostra tentativas restantes
    restantes = max_tentativas - tentativa
    if restantes > 0:
        print(f"Tentativas restantes: {restantes}")
else:
    # Executado apenas se não houve break (esgotou tentativas)
    if not acertou:
        print("\n" + "=" * 60)
        print("😞 GAME OVER - Tentativas esgotadas!")
        print("=" * 60)
        print(f"O número secreto era: {numero_secreto}")
        print("Mais sorte na próxima vez!")

# Estatísticas finais
print("\n" + "=" * 60)
print("ESTATÍSTICAS")
print("=" * 60)
print(f"Total de tentativas usadas: {tentativa}")
if acertou:
    percentual = (tentativa / max_tentativas) * 100
    print(f"Eficiência: {percentual:.1f}% das tentativas")
    if tentativa <= 3:
        print("🏆 Desempenho: EXCELENTE!")
    elif tentativa <= 5:
        print("🥇 Desempenho: BOM!")
    else:
        print("👍 Desempenho: Regular")
print("=" * 60)
````

## Boas Práticas com While

### 1. Evite Loops Infinitos Acidentais
```python
# ❌ ERRADO - Loop infinito (contador não incrementa)
i = 0
while i < 10:
    print(i)
    # Esqueceu de incrementar i!

# ✅ CORRETO
i = 0
while i < 10:
    print(i)
    i += 1
```

### 2. Inicialize Variáveis de Controle
```python
# ✅ Sempre inicialize antes do while
contador = 0
soma = 0
encontrou = False

while contador < 10:
    # lógica aqui
    contador += 1
```

### 3. Use break para Condições Complexas
```python
# ❌ EVITE - Condição complexa
while x < 100 and not erro and usuario_ativo and tentativas < 5:
    # difícil de ler
    pass

# ✅ PREFIRA - break para clareza
while True:
    if x >= 100 or erro or not usuario_ativo or tentativas >= 5:
        break
    # lógica aqui
```

### 4. Prefira for Quando Possível
```python
# ❌ DESNECESSÁRIO - while para iterar lista
i = 0
while i < len(lista):
    print(lista[i])
    i += 1

# ✅ MELHOR - use for
for item in lista:
    print(item)

# ✅ While apropriado - condição desconhecida
while not arquivo.eof():
    linha = arquivo.readline()
```

### 5. While-else para Buscas
```python
# ✅ Padrão elegante para buscas
i = 0
while i < len(lista):
    if lista[i] == alvo:
        print("Encontrado!")
        break
    i += 1
else:
    print("Não encontrado")
```

### 6. Cuidado com While True
```python
# ❌ PERIGOSO - sem saída clara
while True:
    processar()
    # Como sair?

# ✅ SEGURO - saída explícita
while True:
    resultado = processar()
    if resultado == "sair":
        break
```

### 7. Limite Iterações em Produção
```python
# ✅ Proteção contra loops infinitos
max_iteracoes = 1000
iteracoes = 0

while condicao:
    iteracoes += 1
    if iteracoes > max_iteracoes:
        print("⚠️ Limite de iterações excedido")
        break
    # lógica aqui
```

### 8. Documente Condições Complexas
```python
# ✅ Comente lógica não óbvia
while saldo > 0 and not conta_bloqueada:
    # Continua processando enquanto há saldo
    # e a conta não foi bloqueada por segurança
    processar_transacao()
```

### 9. Use Continue com Parcimônia
```python
# ❌ CONFUSO - múltiplos continue
while i < 10:
    if condicao1:
        continue
    if condicao2:
        continue
    if condicao3:
        continue
    processar()

# ✅ MAIS CLARO - lógica positiva
while i < 10:
    if not condicao1 and not condicao2 and not condicao3:
        processar()
```

### 10. Valide Entradas em Loops
```python
# ✅ Sempre valide entradas do usuário
while True:
    entrada = input("Digite um número: ")
    
    try:
        numero = int(entrada)
        if 1 <= numero <= 100:
            break
        else:
            print("Número fora do intervalo (1-100)")
    except ValueError:
        print("Entrada inválida. Digite um número.")
```

### Comparação: While vs For

**Use while quando:**
* Não sabe quantas iterações serão necessárias
* A condição de parada é complexa
* Precisa de controle fino sobre o incremento
* Está esperando um evento/condição externa

**Use for quando:**
* Sabe o número de iterações
* Está iterando sobre uma sequência (lista, range, etc.)
* Não precisa modificar o contador manualmente
* Quer código mais legível e pythonica
