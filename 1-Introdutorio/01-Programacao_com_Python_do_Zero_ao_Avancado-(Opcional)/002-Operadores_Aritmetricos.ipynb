# Operadores Aritméticos em Python

Este notebook cobre os operadores aritméticos básicos do Python:
* **Adição** (`+`)
* **Subtração** (`-`)
* **Multiplicação** (`*`)
* **Divisão** (`/`)
* **Divisão Inteira** (`//`)
* **Módulo** (`%`)
* **Exponenciação** (`**`)

Cada operador será demonstrado com exemplos práticos e casos de uso.

## 1. Adição (+)

O operador `+` soma dois ou mais valores numéricos.

**Sintaxe:** `resultado = valor1 + valor2`

**Casos de uso:**
* Somar números inteiros e decimais
* Calcular totais e soma de valores
* Concatenar strings (quando usado com texto)

```python
# Adição de inteiros
x = 7
y = 9
z = x + y
print(f"A soma de {x} + {y} = {z}")

# Adição de decimais
preco1 = 15.50
preco2 = 23.75
total = preco1 + preco2
print(f"\nTotal da compra: R$ {total:.2f}")

# Adição múltipla
a, b, c = 10, 20, 30
soma_total = a + b + c
print(f"\nSoma de {a} + {b} + {c} = {soma_total}")
```

## 2. Subtração (-)

O operador `-` subtrai um valor de outro.

**Sintaxe:** `resultado = valor1 - valor2`

**Casos de uso:**
* Calcular diferenças entre valores
* Decrementar contadores
* Calcular descontos e reduções

```python
# Subtração simples
x = 50
y = 18
diferenca = x - y
print(f"A diferença entre {x} e {y} = {diferenca}")

# Calculando desconto
preco_original = 100.00
desconto = 15.00
preco_final = preco_original - desconto
print(f"\nPreço original: R$ {preco_original:.2f}")
print(f"Desconto: R$ {desconto:.2f}")
print(f"Preço final: R$ {preco_final:.2f}")

# Subtração de negativos
a = 10
b = -5
resultado = a - b  # 10 - (-5) = 10 + 5 = 15
print(f"\n{a} - ({b}) = {resultado}")
```

## 3. Multiplicação (*)

O operador `*` multiplica dois valores.

**Sintaxe:** `resultado = valor1 * valor2`

**Casos de uso:**
* Calcular produtos e totais
* Repetir strings
* Escalar valores

```python
# Multiplicação simples
quantidade = 5
preco_unitario = 12.50
total = quantidade * preco_unitario
print(f"{quantidade} itens x R$ {preco_unitario:.2f} = R$ {total:.2f}")

# Multiplicação de negativos
a = -3
b = 4
resultado = a * b
print(f"\n{a} * {b} = {resultado}")

# Repetição de strings
texto = "Python! "
repetido = texto * 3
print(f"\n{repetido}")
```

## 4. Divisão (/)

O operador `/` divide um valor por outro, retornando sempre um número decimal (float).

**Sintaxe:** `resultado = valor1 / valor2`

**Observações:**
* Sempre retorna float, mesmo que a divisão seja exata
* Divisão por zero gera erro (ZeroDivisionError)

```python
# Divisão simples
total = 100
pessoas = 4
valor_por_pessoa = total / pessoas
print(f"R$ {total} dividido por {pessoas} pessoas = R$ {valor_por_pessoa:.2f} cada")

# Divisão com resultado decimal
a = 10
b = 3
resultado = a / b
print(f"\n{a} / {b} = {resultado:.4f}")

# Divisão exata (ainda retorna float)
x = 20
y = 5
z = x / y
print(f"\n{x} / {y} = {z} (tipo: {type(z).__name__})")
```

## 5. Divisão Inteira (//)

O operador `//` divide um valor por outro e retorna apenas a parte inteira do resultado (arredonda para baixo).

**Sintaxe:** `resultado = valor1 // valor2`

**Casos de uso:**
* Quando precisa apenas da parte inteira da divisão
* Calcular quantidades exatas sem frações

```python
# Divisão inteira vs divisão normal
a = 17
b = 5

divisao_normal = a / b
divisao_inteira = a // b

print(f"Divisão normal: {a} / {b} = {divisao_normal}")
print(f"Divisão inteira: {a} // {b} = {divisao_inteira}")

# Exemplo prático: distribuir itens
total_itens = 25
itens_por_caixa = 7
caixas_completas = total_itens // itens_por_caixa
print(f"\n{total_itens} itens em caixas de {itens_por_caixa}: {caixas_completas} caixas completas")
```

## 6. Módulo (%)

O operador `%` retorna o resto da divisão entre dois números.

**Sintaxe:** `resto = valor1 % valor2`

**Casos de uso:**
* Verificar se um número é par ou ímpar
* Calcular itens restantes após distribuição
* Operações cíclicas

```python
# Calculando o resto da divisão
a = 17
b = 5
resto = a % b
print(f"O resto de {a} dividido por {b} = {resto}")

# Verificando se número é par ou ímpar
numeros = [10, 15, 22, 7]
for num in numeros:
    if num % 2 == 0:
        print(f"{num} é par")
    else:
        print(f"{num} é ímpar")

# Exemplo prático: itens restantes
total_itens = 25
itens_por_caixa = 7
itens_restantes = total_itens % itens_por_caixa
print(f"\nItens restantes (não formam caixa completa): {itens_restantes}")
```

## 7. Exponenciação (**)

O operador `**` eleva um número à potência de outro.

**Sintaxe:** `resultado = base ** expoente`

**Casos de uso:**
* Cálculos de potência
* Raízes (usando expoentes fracionários)
* Crescimento exponencial

```python
# Potência simples
base = 2
expoente = 8
resultado = base ** expoente
print(f"{base} elevado a {expoente} = {resultado}")

# Potências comuns
print(f"\n3² = {3 ** 2}")
print(f"5³ = {5 ** 3}")
print(f"10⁴ = {10 ** 4}")

# Raíz quadrada (expoente 0.5)
numero = 16
raiz_quadrada = numero ** 0.5
print(f"\nRaíz quadrada de {numero} = {raiz_quadrada}")

# Raíz cúbica (expoente 1/3)
numero = 27
raiz_cubica = numero ** (1/3)
print(f"Raíz cúbica de {numero} = {raiz_cubica:.2f}")
```

## 8. Exemplo Consolidado: Todos os Operadores

Vamos aplicar todos os operadores aritméticos em um único exemplo para comparar seus resultados.

```python
# Definindo dois valores para demonstrar todos os operadores
x = 40
y = 6

print(f"Valores: x = {x}, y = {y}\n")
print("=" * 50)

# Aplicando todos os operadores
adicao = x + y
subtracao = x - y
multiplicacao = x * y
divisao = x / y
divisao_inteira = x // y
modulo = x % y
exponenciacao = x ** 2  # x ao quadrado

# Exibindo resultados organizados
print(f"Adição:           {x} + {y}  = {adicao}")
print(f"Subtração:        {x} - {y}  = {subtracao}")
print(f"Multiplicação:    {x} * {y}  = {multiplicacao}")
print(f"Divisão:          {x} / {y}  = {divisao:.2f}")
print(f"Divisão Inteira:  {x} // {y} = {divisao_inteira}")
print(f"Módulo (resto):   {x} % {y}  = {modulo}")
print(f"Exponenciação:    {x} ** 2 = {exponenciacao}")
print("=" * 50)
```

## 9. Precedência de Operadores

Quando uma expressão contém múltiplos operadores, Python segue uma ordem de precedência:

**Ordem (da maior para a menor precedência):**
1. **Parênteses** `()`
2. **Exponenciação** `**`
3. **Multiplicação, Divisão, Divisão Inteira, Módulo** `*`, `/`, `//`, `%`
4. **Adição, Subtração** `+`, `-`

Use parênteses para deixar a ordem de execução explícita e melhorar a legibilidade.

```python
# Sem parênteses - segue a precedência
resultado1 = 2 + 3 * 4
print(f"2 + 3 * 4 = {resultado1}")  # Multiplicação primeiro: 2 + 12 = 14

# Com parênteses - força a ordem
resultado2 = (2 + 3) * 4
print(f"(2 + 3) * 4 = {resultado2}")  # Adição primeiro: 5 * 4 = 20

# Exemplo mais complexo
resultado3 = 10 + 5 * 2 ** 3 / 4 - 1
print(f"\n10 + 5 * 2 ** 3 / 4 - 1 = {resultado3}")
print("Ordem: 2**3=8, 5*8=40, 40/4=10, 10+10=20, 20-1=19")

# Com parênteses para clareza
resultado4 = 10 + ((5 * (2 ** 3)) / 4) - 1
print(f"10 + ((5 * (2 ** 3)) / 4) - 1 = {resultado4}")
print("Mesma operação, mas mais clara!")
```

## Boas Práticas com Operadores Aritméticos

### 1. Clareza e Legibilidade
* Use parênteses mesmo quando não são obrigatórios para tornar expressões complexas mais claras
* Quebre expressões longas em várias linhas com variáveis intermediárias
* Adicione espaços ao redor dos operadores: `a + b` em vez de `a+b`

### 2. Divisão por Zero
* Sempre valide divisões para evitar `ZeroDivisionError`
* Use blocos try-except quando necessário

### 3. Tipos de Dados
* Lembre-se: `/` sempre retorna float, mesmo com inteiros
* Use `//` quando precisar apenas da parte inteira
* Tenha cuidado ao misturar int e float em operações

### 4. Performance
* Exponenciação pode ser custosa: `x ** 2` é ok, mas para potências grandes considere bibliotecas como NumPy
* Evite cálculos repetitivos: armazene resultados em variáveis

### 5. Precisão
* Operações com float podem ter imprecisões: `0.1 + 0.2` ≠ `0.3` exatamente
* Para valores monetários, considere usar `Decimal` do módulo `decimal`
