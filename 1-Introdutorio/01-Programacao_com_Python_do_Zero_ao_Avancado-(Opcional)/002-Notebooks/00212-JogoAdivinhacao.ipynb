## 1️⃣ Estrutura Básica do Jogo

### 🎯 Conceito

Um jogo de adivinhação combina vários elementos de programação:

* **Geração de número aleatório** - Usando `random.randint()`
* **Loop de tentativas** - Limitando o número de chances
* **Validação de entrada** - Garantindo dados válidos
* **Feedback ao jogador** - Dicas de "alto" ou "baixo"
* **Condição de vitória/derrota** - Encerrando o jogo apropriadamente

### 📝 Sintaxe Principal

```python
import random

# Gerar número secreto
numero_secreto = random.randint(min, max)

# Loop de tentativas
for tentativa in range(max_tentativas):
    palpite = int(input("Seu palpite: "))
    if palpite == numero_secreto:
        # Vitória
        break
    elif palpite > numero_secreto:
        # Feedback: muito alto
    else:
        # Feedback: muito baixo
```

### ✨ Casos de Uso

* Jogos educacionais para ensinar programação
* Exercícios de lógica e estruturas de controle
* Prototipagem de mecânicas de jogo
* Validação de entrada de usuário

````python
import random

# Exemplo 1: Jogo simples com 3 tentativas
print("=" * 50)
print("Exemplo 1: Jogo Simples")
print("=" * 50)

numero_secreto = random.randint(1, 10)
max_tentativas = 3

print(f"\n🎲 Adivinhe o número entre 1 e 10!")
print(f"Você tem {max_tentativas} tentativas.\n")

# Simulação de tentativas (ao invés de input())
tentativas_simuladas = [5, 7, 3]  # Simular palpites do jogador

for i in range(max_tentativas):
    palpite = tentativas_simuladas[i]
    print(f"Tentativa {i+1}/{max_tentativas}: {palpite}")
    
    if palpite == numero_secreto:
        print(f"\n✅ Parabéns! Você acertou em {i+1} tentativa(s)!")
        break
    elif palpite > numero_secreto:
        print("⬆️ Seu palpite foi alto.")
    else:
        print("⬇️ Seu palpite foi baixo.")
else:
    print(f"\n❌ Game Over! O número era {numero_secreto}.")

# Exemplo 2: Jogo com dificuldades diferentes
print("\n" + "=" * 50)
print("Exemplo 2: Níveis de Dificuldade")
print("=" * 50)

dificuldades = {
    "facil": {"min": 1, "max": 10, "tentativas": 5},
    "medio": {"min": 1, "max": 50, "tentativas": 7},
    "dificil": {"min": 1, "max": 100, "tentativas": 10}
}

nivel = "medio"
config = dificuldades[nivel]
numero = random.randint(config["min"], config["max"])

print(f"\n🎯 Nível: {nivel.upper()}")
print(f"Intervalo: {config['min']} a {config['max']}")
print(f"Tentativas: {config['tentativas']}")
print(f"Número gerado: {numero} (para demonstração)\n")

# Exemplo 3: Jogo com pontuação
print("\n" + "=" * 50)
print("Exemplo 3: Sistema de Pontuação")
print("=" * 50)

numero_secreto = random.randint(1, 20)
pontos_iniciais = 100
penalidade_por_erro = 15

print(f"\n🏆 Pontuação inicial: {pontos_iniciais} pontos")
print(f"Penalidade por erro: -{penalidade_por_erro} pontos\n")

tentativas_simuladas = [12, 8, 5]  # Simular tentativas
pontos = pontos_iniciais

for i, palpite in enumerate(tentativas_simuladas, 1):
    print(f"Tentativa {i}: {palpite} | Pontos: {pontos}")
    
    if palpite == numero_secreto:
        print(f"\n⭐ VENCEU! Pontuação final: {pontos} pontos")
        break
    else:
        pontos -= penalidade_por_erro
        if palpite > numero_secreto:
            print("⬆️ Muito alto")
        else:
            print("⬇️ Muito baixo")
else:
    print(f"\n🚫 Fim de jogo. Número: {numero_secreto} | Pontos finais: {pontos}")

print(f"\n📊 Resultado: {'Aprovou' if pontos > 0 else 'Precisa melhorar'}!")
````

## 2️⃣ Validação de Entrada

### 🔒 Conceito

Validação de entrada é essencial para evitar erros e garantir uma experiência de usuário robusta:

* **Verificação de tipo** - Garantir que o input é um número
* **Verificação de intervalo** - Garantir que está dentro dos limites
* **Tratamento de exceções** - Capturar erros de conversão
* **Mensagens claras** - Informar o usuário sobre entradas inválidas
* **Loop de revalidação** - Permitir nova tentativa após erro

### 📝 Sintaxe de Validação

```python
while True:
    try:
        palpite = int(input("Digite um número: "))
        if min_valor <= palpite <= max_valor:
            break  # Entrada válida
        else:
            print(f"Digite um número entre {min_valor} e {max_valor}")
    except ValueError:
        print("Entrada inválida! Digite apenas números.")
```

### ⚠️ Boas Práticas

* Sempre validar entrada do usuário
* Fornecer mensagens de erro claras e específicas
* Permitir múltiplas tentativas de correção
* Não contar tentativas inválidas contra o jogador

````python
# Exemplo 1: Validação básica de tipo
print("=" * 50)
print("Exemplo 1: Validação de Tipo")
print("=" * 50)

# Simular entradas válidas e inválidas
entradas_simuladas = ["abc", "15.5", "-5", "8"]

for entrada in entradas_simuladas:
    print(f"\nTestando entrada: '{entrada}'")
    try:
        numero = int(entrada)
        print(f"✅ Válido! Convertido para: {numero}")
    except ValueError:
        print(f"❌ Inválido! Não é um número inteiro.")

# Exemplo 2: Validação de intervalo
print("\n" + "=" * 50)
print("Exemplo 2: Validação de Intervalo")
print("=" * 50)

min_valor = 1
max_valor = 10
entradas_teste = [0, 5, 11, 10, -3]

print(f"\nIntervalo permitido: {min_valor} a {max_valor}\n")

for entrada in entradas_teste:
    print(f"Testando: {entrada} ->", end=" ")
    if min_valor <= entrada <= max_valor:
        print(f"✅ Dentro do intervalo")
    else:
        print(f"❌ Fora do intervalo")

# Exemplo 3: Função de validação completa
print("\n" + "=" * 50)
print("Exemplo 3: Função de Validação Completa")
print("=" * 50)

def validar_palpite(entrada_str, min_val, max_val):
    """
    Valida um palpite do usuário.
    
    Retorna: (valido: bool, valor: int ou None, mensagem: str)
    """
    try:
        valor = int(entrada_str)
        if valor < min_val:
            return False, None, f"⚠️ Número muito baixo! Mínimo: {min_val}"
        elif valor > max_val:
            return False, None, f"⚠️ Número muito alto! Máximo: {max_val}"
        else:
            return True, valor, "✅ Entrada válida!"
    except ValueError:
        return False, None, "❌ Erro! Digite apenas números inteiros."

# Testar a função
testes = ["5", "abc", "0", "15", "10"]
print(f"\nTestando validação para intervalo 1-10:\n")

for teste in testes:
    valido, valor, mensagem = validar_palpite(teste, 1, 10)
    print(f"Entrada: '{teste:>4}' -> {mensagem}")
    if valido:
        print(f"             Valor processado: {valor}")

# Exemplo 4: Contador de tentativas válidas vs inválidas
print("\n" + "=" * 50)
print("Exemplo 4: Estatísticas de Validação")
print("=" * 50)

entradas_jogo = ["3", "abc", "15", "7", "xyz", "5"]
validas = 0
invalidas = 0

print("\nProcessando entradas do jogo:\n")

for i, entrada in enumerate(entradas_jogo, 1):
    valido, valor, msg = validar_palpite(entrada, 1, 10)
    status = "✅" if valido else "❌"
    print(f"{i}. {status} '{entrada}' - {msg}")
    
    if valido:
        validas += 1
    else:
        invalidas += 1

print(f"\n📊 Estatísticas:")
print(f"   Entradas válidas: {validas}")
print(f"   Entradas inválidas: {invalidas}")
print(f"   Taxa de sucesso: {validas/len(entradas_jogo)*100:.1f}%")
````

## 3️⃣ Loop de Jogo e Reinício

### 🔄 Conceito

Permitir que o jogador jogue múltiplas rodadas sem reiniciar o programa:

* **Loop externo** - `while` para controlar múltiplas partidas
* **Loop interno** - `for` para controlar tentativas dentro de uma partida
* **Confirmação de reinício** - Perguntar se deseja jogar novamente
* **Limpeza de tela** - Melhorar a experiência visual entre rodadas
* **Contador de estatísticas** - Rastrear vitórias e derrotas

### 📝 Padrão de Implementação

```python
jogar_novamente = True
vitorias = 0
derrotas = 0

while jogar_novamente:
    # Lógica do jogo aqui
    
    resposta = input("Jogar novamente? (s/n): ").lower()
    jogar_novamente = resposta == 's'
    
print(f"Placar final: {vitorias} vitórias, {derrotas} derrotas")
```

### ✨ Melhorias

* Salvar histórico de partidas
* Exibir estatísticas ao final
* Oferecer mudança de dificuldade entre rodadas
* Implementar ranking ou high scores

````python
import random

# Exemplo 1: Loop básico com reinício
print("=" * 50)
print("Exemplo 1: Loop Básico de Jogo")
print("=" * 50)

rodadas_simuladas = ["s", "s", "n"]  # Simular respostas: sim, sim, não
rodada_atual = 0

for resposta_reinicio in rodadas_simuladas:
    rodada_atual += 1
    numero_secreto = random.randint(1, 10)
    
    print(f"\n🎮 Rodada {rodada_atual}")
    print(f"Número secreto: {numero_secreto} (para demonstração)")
    
    # Simular uma tentativa por rodada
    palpite = random.randint(1, 10)
    print(f"Palpite: {palpite}")
    
    if palpite == numero_secreto:
        print("✅ Acertou!")
    else:
        print(f"❌ Errou! Era {numero_secreto}")
    
    print(f"\nJogar novamente? {resposta_reinicio}")
    if resposta_reinicio != "s":
        break

print(f"\n🏁 Fim do jogo após {rodada_atual} rodada(s)")

# Exemplo 2: Sistema de estatísticas
print("\n" + "=" * 50)
print("Exemplo 2: Sistema de Estatísticas")
print("=" * 50)

vitorias = 0
derrotas = 0
total_tentativas = 0

# Simular 5 rodadas
resultados_simulados = [True, False, True, True, False]  # True = vitória

for i, venceu in enumerate(resultados_simulados, 1):
    numero = random.randint(1, 20)
    tentativas_na_rodada = random.randint(1, 5)
    total_tentativas += tentativas_na_rodada
    
    print(f"\n🎯 Rodada {i}: ", end="")
    
    if venceu:
        vitorias += 1
        print(f"✅ VITÓRIA em {tentativas_na_rodada} tentativa(s)!")
    else:
        derrotas += 1
        print(f"❌ DERROTA após {tentativas_na_rodada} tentativa(s). Era {numero}")

print("\n" + "=" * 50)
print("📊 ESTATÍSTICAS FINAIS")
print("=" * 50)
print(f"Vitórias: {vitorias}")
print(f"Derrotas: {derrotas}")
print(f"Total de rodadas: {vitorias + derrotas}")
print(f"Taxa de vitória: {vitorias/(vitorias + derrotas)*100:.1f}%")
print(f"Média de tentativas por rodada: {total_tentativas/(vitorias + derrotas):.1f}")

# Exemplo 3: Histórico de partidas
print("\n" + "=" * 50)
print("Exemplo 3: Histórico de Partidas")
print("=" * 50)

historico = []

for rodada in range(1, 4):
    numero = random.randint(1, 15)
    palpite = random.randint(1, 15)
    acertou = palpite == numero
    tentativas_usadas = random.randint(1, 3)
    
    partida = {
        "rodada": rodada,
        "numero_secreto": numero,
        "ultimo_palpite": palpite,
        "vitoria": acertou,
        "tentativas": tentativas_usadas
    }
    historico.append(partida)

print("\n📜 Histórico de Partidas:\n")
print(f"{'Rod':<5} {'Resultado':<12} {'Número':<8} {'Palpite':<9} {'Tentativas'}")
print("-" * 50)

for p in historico:
    resultado = "✅ Vitória" if p["vitoria"] else "❌ Derrota"
    print(f"{p['rodada']:<5} {resultado:<12} {p['numero_secreto']:<8} {p['ultimo_palpite']:<9} {p['tentativas']}")

vitorias_total = sum(1 for p in historico if p["vitoria"])
print(f"\n🏆 Desempenho geral: {vitorias_total}/{len(historico)} vitórias")
````

## 4️⃣ Exemplo Consolidado - Jogo Completo

### 🎮 Jogo de Adivinhação Profissional

Este exemplo integra todos os conceitos apresentados:

* ✅ Validação robusta de entrada
* ✅ Múltiplas rodadas com estatísticas
* ✅ Sistema de pontuação
* ✅ Níveis de dificuldade
* ✅ Feedback contextual detalhado
* ✅ Histórico de partidas
* ✅ Tratamento de erros completo

### 🎯 Características

O jogo implementa boas práticas de programação, incluindo:

* Funções modulares e reusavéis
* Validação de entrada em todos os pontos
* Mensagens claras e informativas
* Estrutura de código organizada e legível

````python
import random

def validar_numero(entrada, minimo, maximo):
    """
    Valida se a entrada é um número inteiro dentro do intervalo.
    
    Retorna: (valido, valor, mensagem)
    """
    try:
        valor = int(entrada)
        if valor < minimo:
            return False, None, f"⚠️ Número abaixo do mínimo ({minimo})"
        elif valor > maximo:
            return False, None, f"⚠️ Número acima do máximo ({maximo})"
        else:
            return True, valor, "✅ Válido"
    except ValueError:
        return False, None, "❌ Entrada inválida - digite apenas números"

def jogar_rodada(minimo, maximo, max_tentativas):
    """
    Executa uma rodada completa do jogo.
    
    Retorna: (venceu: bool, tentativas_usadas: int, numero_secreto: int)
    """
    numero_secreto = random.randint(minimo, maximo)
    
    print(f"\n🎲 Adivinhe o número entre {minimo} e {maximo}!")
    print(f"Você tem {max_tentativas} tentativas.\n")
    
    # Simular tentativas (em ambiente real seria input())
    tentativas_simuladas = [random.randint(minimo, maximo) for _ in range(max_tentativas)]
    
    for i in range(max_tentativas):
        palpite = tentativas_simuladas[i]
        tentativa_num = i + 1
        
        print(f"Tentativa {tentativa_num}/{max_tentativas}: {palpite}")
        
        if palpite == numero_secreto:
            print(f"\n✨ 🎉 PARABÉNS! Você acertou em {tentativa_num} tentativa(s)! 🎉 ✨")
            return True, tentativa_num, numero_secreto
        elif palpite > numero_secreto:
            diferenca = palpite - numero_secreto
            if diferenca <= 3:
                print("🔥 Quente! Mas ainda alto...")
            else:
                print("❄️ Frio! Muito alto.")
        else:
            diferenca = numero_secreto - palpite
            if diferenca <= 3:
                print("🔥 Quente! Mas ainda baixo...")
            else:
                print("❄️ Frio! Muito baixo.")
    
    print(f"\n🚫 Game Over! O número era {numero_secreto}")
    return False, max_tentativas, numero_secreto

def exibir_estatisticas(historico):
    """
    Exibe estatísticas consolidadas do jogo.
    """
    if not historico:
        print("\n📊 Nenhuma partida jogada ainda.")
        return
    
    vitorias = sum(1 for p in historico if p["venceu"])
    derrotas = len(historico) - vitorias
    total_tentativas = sum(p["tentativas"] for p in historico)
    
    print("\n" + "=" * 60)
    print("📊 ESTATÍSTICAS GERAIS")
    print("=" * 60)
    print(f"Total de partidas: {len(historico)}")
    print(f"Vitórias: {vitorias} 🏆")
    print(f"Derrotas: {derrotas} 🚫")
    print(f"Taxa de vitória: {vitorias/len(historico)*100:.1f}%")
    print(f"Média de tentativas: {total_tentativas/len(historico):.1f}")
    
    if vitorias > 0:
        tentativas_vitorias = [p["tentativas"] for p in historico if p["venceu"]]
        print(f"Melhor vitória: {min(tentativas_vitorias)} tentativa(s) ⭐")

# ============================================================================
# JOGO PRINCIPAL
# ============================================================================

print("\n" + "=" * 60)
print("🎮 BEM-VINDO AO JOGO DE ADIVINHAÇÃO! 🎮")
print("=" * 60)

# Configurações de dificuldade
NIVEIS = {
    1: {"nome": "Fácil", "min": 1, "max": 10, "tentativas": 5},
    2: {"nome": "Médio", "min": 1, "max": 50, "tentativas": 7},
    3: {"nome": "Difícil", "min": 1, "max": 100, "tentativas": 10}
}

# Selecionar nível (simulado)
nivel_escolhido = 2  # Médio
config = NIVEIS[nivel_escolhido]

print(f"\n🎯 Nível selecionado: {config['nome']}")
print(f"Intervalo: {config['min']} a {config['max']}")
print(f"Tentativas por rodada: {config['tentativas']}\n")

# Histórico de partidas
historico_partidas = []

# Simular 3 rodadas
for rodada in range(1, 4):
    print("\n" + "=" * 60)
    print(f"🏆 RODADA {rodada}")
    print("=" * 60)
    
    venceu, tentativas, numero = jogar_rodada(
        config["min"], 
        config["max"], 
        config["tentativas"]
    )
    
    historico_partidas.append({
        "rodada": rodada,
        "venceu": venceu,
        "tentativas": tentativas,
        "numero": numero
    })
    
    # Simular decisão de continuar
    if rodada < 3:
        print(f"\n\u27a1️ Continuando para próxima rodada...")
    else:
        print(f"\n🏁 Última rodada completada!")

# Exibir estatísticas finais
exibir_estatisticas(historico_partidas)

print("\n" + "=" * 60)
print("👋 Obrigado por jogar!")
print("=" * 60)
````

### 10. 📚 Código Legível e Mantenedor

**Escreva código que outros (e você no futuro) possam entender:**

* **Comente lógicas complexas**, não o óbvio
* **Use nomes descritivos** para variáveis e funções
* **Mantenha funções curtas** (idealmente < 30 linhas)
* **Siga convenções** (PEP 8 para Python)
* **Evite duplicação** de código (DRY - Don't Repeat Yourself)
* **Teste regularmente** durante o desenvolvimento
