```python
var_global = "Curso Completo de Python"

def escreve_texto():
    global var_global
    var_global = "Bancos de Dados com SQL"
    var_local = "Bob Esponja"
    print(f"Variável Global: {var_global}")
    print(f"Variável Local: {var_local}")

if __name__ == "__main__":
    print(f"Executando a função escreve_texto()")
    escreve_texto()

    print(f"Tentar acessar as variáveis diretamente")
    print(f"Variável Global: {var_global}")
```
