## 1️⃣ Funções Matemáticas Built-in

### 📖 Conceito
Python possui funções matemáticas integradas que não requerem importação de módulos. Essas funções estão sempre disponíveis e são otimizadas para desempenho.

### 🔧 Funções Principais
```python
abs(x)      # Valor absoluto
pow(x, y)   # Potenciação (x elevado a y)
round(x, n) # Arredondamento (n casas decimais)
max(...)    # Valor máximo
min(...)    # Valor mínimo
sum(...)    # Soma de iteráveis
divmod(x, y) # Retorna (quociente, resto)
```

### ✅ Características
* **Sempre disponíveis**: Não precisam de import
* **Rápidas**: Implementadas em C
* **Versáteis**: Trabalham com int, float e outros tipos numéricos
* **Úteis para validações**: Especialmente abs() para diferenças

````python
# Exemplo 1: abs() - Valor absoluto ✅
temperatura_ontem = 15
temperatura_hoje = -5
diferenca = abs(temperatura_hoje - temperatura_ontem)
print(f"🌡️ Diferença de temperatura: {diferenca}°C")

# Exemplo 2: pow() - Potenciação 📈
base = 2
expoente = 10
resultado = pow(base, expoente)
print(f"🔢 {base}^{expoente} = {resultado}")

# Exemplo 3: round() - Arredondamento 🎯
preco = 19.9876
preco_arredondado = round(preco, 2)
print(f"💰 Preço: R$ {preco_arredondado}")

# Exemplo 4: max() e min() - Valores extremos 📊
notas = [7.5, 8.2, 6.8, 9.5, 7.0, 8.8]
maior_nota = max(notas)
menor_nota = min(notas)
print(f"📚 Maior nota: {maior_nota} | Menor nota: {menor_nota}")

# Exemplo 5: sum() - Soma de iteráveis ➕
vendas_semanais = [1200, 1500, 980, 1350, 1800, 2100, 1650]
total_vendas = sum(vendas_semanais)
media_vendas = sum(vendas_semanais) / len(vendas_semanais)
print(f"💵 Total vendas: R$ {total_vendas:.2f}")
print(f"📊 Média diária: R$ {media_vendas:.2f}")

# Exemplo 6: divmod() - Divisão com quociente e resto 🔢
total_minutos = 137
horas, minutos = divmod(total_minutos, 60)
print(f"⏰ {total_minutos} minutos = {horas}h {minutos}min")
````

## 2️⃣ Módulo Math - Funções Básicas

### 📖 Conceito
O módulo `math` fornece acesso às funções matemáticas definidas pelo padrão C. Oferece funções para arredondamento, raízes e operações mais avançadas.

### 🔧 Funções de Arredondamento e Raízes
```python
import math

math.ceil(x)    # Arredonda para cima
math.floor(x)   # Arredonda para baixo
math.trunc(x)   # Remove parte decimal
math.sqrt(x)    # Raiz quadrada
math.isqrt(x)   # Raiz quadrada inteira (Python 3.8+)
math.fabs(x)    # Valor absoluto (retorna float)
```

### ⚠️ Diferenças Importantes
* **ceil vs floor**: ceil sempre para cima, floor sempre para baixo
* **round vs trunc**: round arredonda, trunc apenas remove decimais
* **sqrt vs isqrt**: sqrt retorna float, isqrt retorna int
* **abs vs fabs**: abs mantém tipo, fabs sempre retorna float

````python
import math

# Exemplo 1: ceil() - Arredondamento para cima ⬆️
pessoas_por_van = 35 / 8
vans_necessarias = math.ceil(pessoas_por_van)
print(f"🚐 Pessoas: 35 | Vans necessárias: {vans_necessarias}")

# Exemplo 2: floor() - Arredondamento para baixo ⬇️
dinheiro_disponivel = 127.80
preco_ingresso = 15.50
ingressos_compraveis = math.floor(dinheiro_disponivel / preco_ingresso)
print(f"🎫 Pode comprar {ingressos_compraveis} ingressos")

# Exemplo 3: trunc() - Truncamento (remove decimais) ✂️
pi_aproximado = math.trunc(math.pi)
print(f"🥧 Pi truncado: {pi_aproximado} (original: {math.pi})")

# Exemplo 4: sqrt() - Raiz quadrada √ 
area_quadrado = 64
lado = math.sqrt(area_quadrado)
perimetro = 4 * lado
print(f"📐 Área: {area_quadrado}m² → Lado: {lado}m → Perímetro: {perimetro}m")

# Exemplo 5: isqrt() - Raiz quadrada inteira (Python 3.8+) 🔢
numero = 50
raiz_inteira = math.isqrt(numero)
print(f"✅ Raiz inteira de {numero}: {raiz_inteira}")
print(f"📊 Verificação: {raiz_inteira}² = {raiz_inteira**2}, {raiz_inteira+1}² = {(raiz_inteira+1)**2}")

# Exemplo 6: Comparação de arredondamentos 🎯
valor = 3.7
print(f"\n🔍 Comparando arredondamentos de {valor}:")
print(f"   ceil():  {math.ceil(valor)}")
print(f"   floor(): {math.floor(valor)}")
print(f"   round(): {round(valor)}")
print(f"   trunc(): {math.trunc(valor)}")
````

## 3️⃣ Exponenciais e Logaritmos

### 📖 Conceito
Exponenciais e logaritmos são operações inversas fundamentais em matemática, úteis para crescimento/decaimento, escalas e transformações de dados.

### 🔧 Funções Disponíveis
```python
import math

# Exponenciais
math.exp(x)        # e elevado a x (e^x)
math.expm1(x)      # e^x - 1 (preciso para x pequeno)
math.pow(x, y)     # x elevado a y

# Logaritmos
math.log(x)        # Logaritmo natural (base e)
math.log10(x)      # Logaritmo base 10
math.log2(x)       # Logaritmo base 2
math.log(x, base)  # Logaritmo em base customizada
math.log1p(x)      # log(1 + x) (preciso para x pequeno)
```

### 🎯 Casos de Uso
* **Crescimento exponencial**: Populações, juros compostos, infeçções
* **Logaritmos**: Escalas (pH, Richter, decibéis), complexidade, normalização

````python
import math

# Exemplo 1: exp() - Crescimento exponencial 📈
populacao_inicial = 1000
taxa_crescimento = 0.05  # 5% ao ano
anos = 10
populacao_final = populacao_inicial * math.exp(taxa_crescimento * anos)
print(f"🌍 População após {anos} anos: {populacao_final:.0f} habitantes")

# Exemplo 2: log10() - Escala de Richter (terremotos) 🌋
energia_terremoto = 10**15  # Joules
magnitude = math.log10(energia_terremoto) - 4.8
print(f"⚡ Energia: {energia_terremoto:.2e} J → Magnitude Richter: {magnitude:.1f}")

# Exemplo 3: log2() - Complexidade de algoritmos 💻
n_elementos = 1000000
operacoes_linear = n_elementos
operacoes_log = n_elementos * math.log2(n_elementos)
print(f"🔍 Para {n_elementos} elementos:")
print(f"   Busca linear: {operacoes_linear:,.0f} operações")
print(f"   Busca binária: {math.log2(n_elementos):.0f} operações")

# Exemplo 4: Juros compostos com diferentes freqüências 💰
capital_inicial = 10000
taxa_anual = 0.08  # 8%
periodo_anos = 5

# Capitalização contínua
montante_continuo = capital_inicial * math.exp(taxa_anual * periodo_anos)
print(f"\n💵 Capital inicial: R$ {capital_inicial:,.2f}")
print(f"📈 Taxa: {taxa_anual*100}% ao ano por {periodo_anos} anos")
print(f"✅ Capitalização contínua: R$ {montante_continuo:,.2f}")

# Exemplo 5: pH (escala logarítmica) 🧪
concentracao_h = 1e-7  # Mols por litro
pH = -math.log10(concentracao_h)
if pH < 7:
    tipo = "ácida"
elif pH > 7:
    tipo = "básica/alcalina"
else:
    tipo = "neutra"
print(f"\n🧪blim H⁺: {concentracao_h} → pH: {pH:.1f} (solução {tipo})")

# Exemplo 6: Decaimento radioativo ☢️
massa_inicial = 100  # gramas
meia_vida = 5730  # anos (Carbono-14)
tempo_decorrido = 11460  # anos
massa_final = massa_inicial * math.exp(-math.log(2) * tempo_decorrido / meia_vida)
print(f"\n☢️  Decaimento C-14: {massa_inicial}g → {massa_final:.2f}g em {tempo_decorrido} anos")
````

## 4️⃣ Funções Trigonométricas

### 📖 Conceito
Funções trigonométricas relacionam ângulos e distâncias. São essenciais em geometria, física, engenharia, computação gráfica e processamento de sinais.

### 🔧 Funções Principais
```python
import math

# Funções trigonométricas (entrada em radianos)
math.sin(x)       # Seno
math.cos(x)       # Cosseno
math.tan(x)       # Tangente

# Funções inversas (retornam radianos)
math.asin(x)      # Arco-seno
math.acos(x)      # Arco-cosseno
math.atan(x)      # Arco-tangente
math.atan2(y, x)  # Arco-tangente de y/x (preserva quadrante)

# Conversão de unidades
math.degrees(x)   # Radianos para graus
math.radians(x)   # Graus para radianos

# Hiperbólicas
math.sinh(x), math.cosh(x), math.tanh(x)
```

### ⚠️ Importante
* **Todas as funções trigonométricas usam RADIANOS**
* Use `math.radians()` para converter graus em radianos
* Use `math.degrees()` para converter radianos em graus

````python
import math

# Exemplo 1: Cálculo de altura usando trigonometria 🏗️
distancia = 50  # metros da base do prédio
angulo_graus = 60  # ângulo de elevação
angulo_rad = math.radians(angulo_graus)
altura = distancia * math.tan(angulo_rad)
print(f"🏛️ Distância: {distancia}m | Ângulo: {angulo_graus}° → Altura: {altura:.2f}m")

# Exemplo 2: Comprimento de arco e área de setor circular ⚪
raio = 10  # metros
angulo_central = 45  # graus
angulo_rad = math.radians(angulo_central)
comprimento_arco = raio * angulo_rad
area_setor = 0.5 * raio**2 * angulo_rad
print(f"\n🔵 Círculo raio {raio}m, ângulo {angulo_central}°:")
print(f"   Comprimento do arco: {comprimento_arco:.2f}m")
print(f"   Área do setor: {area_setor:.2f}m²")

# Exemplo 3: Coordenadas polares para cartesianas 📍
raio_polar = 5
angulo_polar = 30  # graus
angulo_rad = math.radians(angulo_polar)
x = raio_polar * math.cos(angulo_rad)
y = raio_polar * math.sin(angulo_rad)
print(f"\n🧭 Polar ({raio_polar}, {angulo_polar}°) → Cartesiano ({x:.2f}, {y:.2f})")

# Exemplo 4: Cartesianas para polares usando atan2 📍
ponto_x = 3
ponto_y = 4
raio_calculado = math.sqrt(ponto_x**2 + ponto_y**2)
angulo_calculado_rad = math.atan2(ponto_y, ponto_x)
angulo_calculado_graus = math.degrees(angulo_calculado_rad)
print(f"\n🔴 Cartesiano ({ponto_x}, {ponto_y}) → Polar (r={raio_calculado:.2f}, θ={angulo_calculado_graus:.2f}°)")

# Exemplo 5: Lei dos cossenos (cálculo de distância) 📏
lado_a = 8
lado_b = 10
angulo_entre = 60  # graus
angulo_rad = math.radians(angulo_entre)
lado_c = math.sqrt(lado_a**2 + lado_b**2 - 2*lado_a*lado_b*math.cos(angulo_rad))
print(f"\n🔺 Triângulo: lados {lado_a} e {lado_b}, ângulo {angulo_entre}° → lado oposto: {lado_c:.2f}")

# Exemplo 6: Verificação da identidade trigonométrica fundamental ✅
angulo_teste = 42
angulo_rad = math.radians(angulo_teste)
sen_quadrado = math.sin(angulo_rad)**2
cos_quadrado = math.cos(angulo_rad)**2
soma = sen_quadrado + cos_quadrado
print(f"\n📊 Identidade sin²({angulo_teste}°) + cos²({angulo_teste}°) = {soma:.10f} ≈ 1 ✅")
````

## 5️⃣ Constantes Matemáticas

### 📖 Conceito
O módulo `math` fornece constantes matemáticas fundamentais com precisão máxima disponível no sistema. Essas constantes são essenciais para cálculos científicos e de engenharia.

### 🔧 Constantes Disponíveis
```python
import math

math.pi         # Pi (π ≈ 3.14159...)
math.e          # Número de Euler (e ≈ 2.71828...)
math.tau        # Tau (2π ≈ 6.28318...) - Python 3.6+
math.inf        # Infinito positivo (float('inf'))
math.nan        # Not a Number (float('nan'))
```

### 🎯 Uso das Constantes
* **math.pi**: Cálculos com círculos, ondas, trigonometria
* **math.e**: Crescimento exponencial, logaritmos naturais
* **math.tau**: Alternativa a 2π (mais intuitivo em alguns contextos)
* **math.inf**: Representa infinito, comparações, limites
* **math.nan**: Representa valores indefinidos/inválidos

* Use `math.isnan(x)` para testar se x é NaN (não use `x == math.nan`)
* Use `math.isinf(x)` para testar se x é infinito
* Use `math.isfinite(x)` para verificar se x é um número finito

````python
import math

# Exemplo 1: math.pi - Cálculos com círculos ⚪
raio_pizza = 15  # cm
area_pizza = math.pi * raio_pizza**2
circunferencia = 2 * math.pi * raio_pizza
print(f"🍕 Pizza raio {raio_pizza}cm:")
print(f"   Área: {area_pizza:.2f} cm²")
print(f"   Circunferência: {circunferencia:.2f} cm")

# Exemplo 2: math.e - Juros compostos contínuos 💰
capital = 5000
taxa = 0.06  # 6% ao ano
tempo = 3  # anos
montante = capital * math.e**(taxa * tempo)
print(f"\n💵 Capital: R$ {capital:,.2f} a {taxa*100}% por {tempo} anos")
print(f"📈 Montante (juros contínuos): R$ {montante:,.2f}")

# Exemplo 3: math.tau - Radianos em círculo completo 🔄
voltas = 2.5
angulo_radianos = voltas * math.tau
angulo_graus = math.degrees(angulo_radianos)
print(f"\n🎯 {voltas} voltas = {angulo_radianos:.2f} radianos = {angulo_graus:.0f}°")

# Exemplo 4: math.inf - Representa infinito ♾️
valores = [10, 50, 200, math.inf, 1000, -500]
maior = max(valores)
menor = min(valores)
print(f"\n🔢 Lista: {valores[:3]}... (com inf)")
print(f"   Maior valor: {maior}")
print(f"   É infinito? {math.isinf(maior)}")

# Exemplo 5: math.nan - Representa valores indefinidos ⚠️
resultado_invalido = math.sqrt(-1) if False else math.nan  # Simula operação inválida
print(f"\n❌ Resultado inválido: {resultado_invalido}")
print(f"   É NaN? {math.isnan(resultado_invalido)}")
print(f"   NaN == NaN? {resultado_invalido == math.nan} (sempre False!)")

# Exemplo 6: Testes com valores especiais 🔍
valores_teste = [42, math.inf, -math.inf, math.nan, 0.0]
print(f"\n🧪 Testando valores especiais:")
for val in valores_teste:
    print(f"   {str(val):>10} | finito: {math.isfinite(val):5} | infinito: {math.isinf(val):5} | NaN: {math.isnan(val):5}")

# Exemplo 7: Precisão das constantes 🎯
print(f"\n🔬 Precisão das constantes:")
print(f"   Pi:  {math.pi:.15f}")
print(f"   e:   {math.e:.15f}")
print(f"   Tau: {math.tau:.15f} (2π)")
````

## 6️⃣ Boas Práticas com Operações Matemáticas

### 1️⃣ Prefira Funções Built-in
* Use funções nativas quando possível (mais rápidas)
* `abs()`, `pow()`, `round()` não precisam de import

### 2️⃣ Import Seletivo do Módulo Math
* Para poucos usos: `from math import sqrt, pi`
* Para muitos usos: `import math` (evita conflitos)
* Nunca: `from math import *` (polui namespace)

### 3️⃣ Conversão Graus ↔ Radianos
* **Sempre** converta graus para radianos antes de usar funções trig
* Use `math.radians()` e `math.degrees()` explicitamente
* Documente a unidade nas variáveis: `angulo_graus`, `angulo_rad`

### 4️⃣ Validação de Entrada
* Valide domínio: `sqrt()` requer x ≥ 0
* `log()` e `log10()` requerem x > 0
* Use tratamento de exceções para entradas inválidas

### 5️⃣ Precisão e Arredondamento
* Arredonde apenas para apresentação, não em cálculos intermediários
* Use `round()` para casas decimais, `ceil()`/`floor()` para inteiros
* Cuidado com erros de precisão em ponto flutuante

### 6️⃣ Testes com Valores Especiais
* Teste NaN com `math.isnan()`, não com `== math.nan`
* Teste infinito com `math.isinf()`
* Use `math.isfinite()` para validar números válidos

### 7️⃣ Performance com Potências
* Para quadrado: `x**2` é mais rápido que `pow(x, 2)`
* Para raiz quadrada: `math.sqrt(x)` > `x**0.5` > `pow(x, 0.5)`
* Use `math.isqrt()` quando precisar de raízes inteiras

### 8️⃣ Escolha Logaritmos Apropriados
* `log()` - base e (natural), padrão em ciências
* `log10()` - base 10, comum em escalas (pH, decibéis)
* `log2()` - base 2, ideal para computação/bits

### 9️⃣ Evite Overflow/Underflow
* Use `math.expm1(x)` em vez de `exp(x) - 1` para x pequeno
* Use `math.log1p(x)` em vez de `log(1 + x)` para x pequeno
* Essas funções têm melhor precisão numérica

### 🔟 Considere Bibliotecas Especializadas
* **NumPy**: Arrays, matrizes, operações vetorizadas
* **SciPy**: Funções científicas avançadas
* **Decimal**: Precisão arbitrária para finanças
* **SymPy**: Matemática simbólica

````python
import math

# ✅ BOM: Import explícito e conversão clara
angulo_graus = 45
angulo_rad = math.radians(angulo_graus)
resultado = math.sin(angulo_rad)
print(f"✅ BOM: sin({angulo_graus}°) = {resultado:.4f}")

# ❌ RUIM: Esquece conversão (interpreta graus como radianos)
# resultado_errado = math.sin(45)  # Isso calcularia sin(45 rad), não sin(45°)!
print(f"❌ RUIM: sin(45) sem conversão = {math.sin(45):.4f} (ERRADO!)")

# ✅ BOM: Validar entrada antes de calcular
def calcular_raiz_segura(x):
    if x < 0:
        return None
    return math.sqrt(x)

valor_teste = -16
resultado_seguro = calcular_raiz_segura(valor_teste)
if resultado_seguro is None:
    print(f"\n⚠️  Valor {valor_teste} inválido para raiz quadrada")
else:
    print(f"\n✅ √{valor_teste} = {resultado_seguro}")

# ✅ BOM: Arredondar apenas no final
preco1, preco2, preco3 = 19.99, 34.50, 12.75
total = preco1 + preco2 + preco3  # NÃO arredondar aqui
total_formatado = round(total, 2)  # Arredondar só para exibição
print(f"\n💰 Total: R$ {total_formatado}")

# ✅ BOM: Testar NaN corretamente
valor_indefinido = math.nan
print(f"\n🔍 Testando NaN:")
print(f"   math.isnan(valor): {math.isnan(valor_indefinido)} ✅")
print(f"   valor == math.nan: {valor_indefinido == math.nan} (sempre False, não use!)")

# ✅ BOM: Performance - comparando potências
import time
n = 10_000_000

# Operador ** para quadrado
start = time.perf_counter()
for _ in range(100):
    resultado = 5**2
tempo1 = time.perf_counter() - start

# Função pow para quadrado  
start = time.perf_counter()
for _ in range(100):
    resultado = pow(5, 2)
tempo2 = time.perf_counter() - start

print(f"\n⏱️  Performance (100 repetições):")
print(f"   x**2:     {tempo1*1000:.4f}ms")
print(f"   pow(x,2): {tempo2*1000:.4f}ms")
print(f"   🏆 Operador ** é ~{tempo2/tempo1:.1f}x mais rápido!")

# ✅ BOM: Escolher logaritmo apropriado
print(f"\n📏 Logaritmos diferentes para contextos diferentes:")
valor = 1000
print(f"   ln({valor}) = {math.log(valor):.4f} (ciências naturais)")
print(f"   log10({valor}) = {math.log10(valor):.4f} (escalas, pH, decibéis)")
print(f"   log2({valor}) = {math.log2(valor):.4f} (computação, bits)")

# ✅ BOM: Usar funções especiais para precisão
x_pequeno = 1e-10
print(f"\n🎯 Precisão com valores pequenos (x={x_pequeno}):")
print(f"   exp(x)-1 direto:   {math.exp(x_pequeno)-1}")
print(f"   expm1(x):          {math.expm1(x_pequeno)} (mais preciso!)")
print(f"   log(1+x) direto:   {math.log(1+x_pequeno)}")
print(f"   log1p(x):          {math.log1p(x_pequeno)} (mais preciso!)")

print("\n✅ Siga essas práticas para código matemático robusto e eficiente!")
````
