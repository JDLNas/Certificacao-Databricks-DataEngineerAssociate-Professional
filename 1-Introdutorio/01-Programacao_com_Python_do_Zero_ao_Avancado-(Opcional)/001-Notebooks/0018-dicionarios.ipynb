# 📚 Dicionários em Python

## 📋 Visão Geral
Este notebook explora dicionários (dict), uma das estruturas de dados mais poderosas do Python. Dicionários armazenam dados em pares chave-valor, permitindo acesso rápido e organização eficiente de informações relacionadas.

## 🎯 Objetivos de Aprendizagem
* Criar e acessar dicionários com chaves e valores
* Adicionar, modificar e remover itens
* Usar métodos principais: keys(), values(), items(), get(), pop(), update()
* Iterar sobre dicionários de diferentes formas
* Trabalhar com dicionários aninhados
* Criar dicionários com compreensão (dict comprehension)
* Aplicar boas práticas em manipulação de dicionários

## 1️⃣ Criação e Conceitos Básicos

### 📖 Conceito
Dicionários são estruturas de dados **mutáveis** que armazenam pares **chave-valor**. Cada chave é única e mapeia para um valor específico. São otimizados para busca rápida por chave.

### 🔧 Criação de Dicionários
```python
# Dicionário vazio
vazio = {}
vazio2 = dict()

# Com valores iniciais
pessoa = {"nome": "Maria", "idade": 30}

# Com dict()
pessoa2 = dict(nome="Maria", idade=30)

# Com lista de tuplas
pessoa3 = dict([("nome", "Maria"), ("idade", 30)])
```

### 📊 Características
* **Chaves únicas**: Não pode haver chaves duplicadas
* **Chaves imutáveis**: Strings, números, tuplas (não listas)
* **Valores flexíveis**: Qualquer tipo de dado
* **Ordem preservada**: Python 3.7+ mantém ordem de inserção
* **Acesso rápido**: O(1) para busca por chave

### ✅ Quando Usar
* Mapear relacionamentos (ID → Nome, CPF → Dados)
* Configurar parâmetros e opções
* Contar ocorrências
* Cachear resultados
* Estruturar dados relacionados

````python
# Exemplo 1: Criando dicionários de diferentes formas 📝
aluno1 = {"nome": "Ana", "idade": 20, "curso": "Engenharia"}
aluno2 = dict(nome="Carlos", idade=22, curso="Medicina")
aluno3 = dict([("nome", "Maria"), ("idade", 21), ("curso", "Direito")])

print("🎓 Alunos criados de 3 formas:")
print(f"   Aluno 1: {aluno1}")
print(f"   Aluno 2: {aluno2}")
print(f"   Aluno 3: {aluno3}")

# Exemplo 2: Elemento químico (dados relacionados) 🧪
elemento = {
    "Z": 3,
    "Nome": "Lítio",
    "Simbolo": "Li",
    "Grupo": "Metais Alcalinos",
    "Densidade": 0.534,
    "Ponto_Fusao": 180.5
}

print(f"\n🧪 Elemento Químico:")
print(f"   Nome: {elemento['Nome']} ({elemento['Simbolo']})")
print(f"   Número Atômico: {elemento['Z']}")
print(f"   Grupo: {elemento['Grupo']}")
print(f"   Densidade: {elemento['Densidade']} g/cm³")

# Exemplo 3: Produto (e-commerce) 📦
produto = {
    "id": 12345,
    "nome": "Notebook Dell",
    "preco": 3500.00,
    "estoque": 15,
    "categoria": "Eletrônicos",
    "ativo": True
}

print(f"\n📦 Produto:")
print(f"   ID: {produto['id']}")
print(f"   Nome: {produto['nome']}")
print(f"   Preço: R$ {produto['preco']:.2f}")
print(f"   Estoque: {produto['estoque']} unidades")
print(f"   Status: {'Ativo' if produto['ativo'] else 'Inativo'}")

# Exemplo 4: Tamanho e tipo 📊
config = {"debug": True, "timeout": 30, "server": "localhost"}
print(f"\n🔢 Configuração possui {len(config)} itens")
print(f"🔍 Tipo: {type(config)}")

# Exemplo 5: Diferentes tipos de valores 🌈
misto = {
    "texto": "Python",
    "numero": 42,
    "decimal": 3.14,
    "lista": [1, 2, 3],
    "tupla": (4, 5, 6),
    "booleano": True,
    "nulo": None
}

print(f"\n🌈 Dicionário com tipos variados:")
for chave, valor in misto.items():
    print(f"   {chave}: {valor} (tipo: {type(valor).__name__})")

# Exemplo 6: Verificando existência de chave ✅❌
carro = {"marca": "Toyota", "modelo": "Corolla", "ano": 2023}
print(f"\n🚗 Verificando chaves:")
print(f"   'marca' existe? {'marca' in carro}")
print(f"   'cor' existe? {'cor' in carro}")
print(f"   'ano' não existe? {'ano' not in carro}")
````

## 2️⃣ Acesso e Modificação

### 📖 Conceito
Dicionários permitem acesso rápido aos valores através de suas chaves. Por serem mutáveis, podemos adicionar, modificar e remover itens dinamicamente.

### 🔧 Operações Básicas

**Acessar Valores:**
```python
dict[chave]          # Lança KeyError se não existir
dict.get(chave)      # Retorna None se não existir
dict.get(chave, default)  # Retorna default se não existir
```

**Adicionar/Modificar:**
```python
dict[chave] = valor       # Adiciona ou modifica
dict.update({k: v})       # Atualiza múltiplos
dict.setdefault(k, v)     # Adiciona se não existir
```

**Remover:**
```python
del dict[chave]           # Remove, lança KeyError se não existir
dict.pop(chave)           # Remove e retorna valor
dict.pop(chave, default)  # Remove ou retorna default
dict.popitem()            # Remove e retorna último par
dict.clear()              # Remove todos os itens
```

### ⚠️ Diferenças Importantes
* **dict[key] vs get()**: [] lança erro, get() retorna None
* **del vs pop()**: del não retorna valor, pop() retorna
* **clear() vs del dict**: clear() esvazia, del remove variável

````python
# Exemplo 1: Acessando valores 🔑
pessoa = {"nome": "João", "idade": 35, "cidade": "São Paulo"}

# Acesso direto com []
nome = pessoa["nome"]
print(f"👤 Nome: {nome}")

# Acesso seguro com get()
telefone = pessoa.get("telefone")  # None se não existir
email = pessoa.get("email", "nao_informado@email.com")  # Valor padrão
print(f"📞 Telefone: {telefone}")
print(f"📧 Email: {email}")

# Exemplo 2: Modificando valores existentes ✏️
produto = {"nome": "Teclado", "preco": 150.00, "estoque": 20}
print(f"\n📦 Produto original: {produto}")

produto["preco"] = 135.00  # Modificar preço
produto["estoque"] = 18     # Atualizar estoque
print(f"✏️ Produto atualizado: {produto}")

# Exemplo 3: Adicionando novos itens ➕
carro = {"marca": "Honda", "modelo": "Civic"}
print(f"\n🚗 Carro inicial: {carro}")

carro["ano"] = 2023
carro["cor"] = "Prata"
carro["km"] = 15000
print(f"➕ Com novos campos: {carro}")

# Exemplo 4: update() - Múltiplas atualizações 🔄
aluno = {"nome": "Maria", "idade": 20}
print(f"\n🎓 Aluno inicial: {aluno}")

aluno.update({"curso": "Engenharia", "período": 5, "cr": 8.5})
print(f"🔄 Após update(): {aluno}")

# update() também pode sobrescrever
aluno.update({"idade": 21, "período": 6})
print(f"🔄 Atualizado novamente: {aluno}")

# Exemplo 5: setdefault() - Adicionar se não existir 🎯
config = {"host": "localhost", "port": 8080}
print(f"\n⚙️ Config inicial: {config}")

# Adiciona se não existir
config.setdefault("timeout", 30)
print(f"➕ Com timeout: {config}")

# Não sobrescreve se já existir
config.setdefault("port", 9090)
print(f"🚫 Port não mudou: {config}")

# Exemplo 6: Removendo itens ❌
elemento = {"Z": 3, "Nome": "Lítio", "Grupo": "Alcalinos", "Período": 1}
print(f"\n🧪 Elemento completo: {elemento}")

# del - remove item
del elemento["Período"]
print(f"❌ Removeu Período: {elemento}")

# pop() - remove e retorna
grupo = elemento.pop("Grupo")
print(f"↩️ pop() retornou: {grupo}")
print(f"📄 Elemento agora: {elemento}")

# pop() com valor padrão
densidade = elemento.pop("Densidade", "N/A")
print(f"↩️ Densidade (inexistente): {densidade}")

# Exemplo 7: clear() - Limpar tudo 🧹
temp = {"a": 1, "b": 2, "c": 3}
print(f"\n📊 Antes: {temp}")
temp.clear()
print(f"🧹 Depois de clear(): {temp}")
````

## 3️⃣ Métodos Principais

### 📖 Conceito
Dicionários possuem métodos especializados para acessar chaves, valores e pares. Esses métodos retornam **views** dinâmicas que refletem mudanças no dicionário original.

### 🔧 Métodos de Acesso

**keys()** - Retorna todas as chaves:
```python
dict.keys()    # dict_keys(['chave1', 'chave2'])
list(dict.keys())  # ['chave1', 'chave2']
```

**values()** - Retorna todos os valores:
```python
dict.values()  # dict_values([valor1, valor2])
list(dict.values())  # [valor1, valor2]
```

**items()** - Retorna pares chave-valor:
```python
dict.items()   # dict_items([('k1', 'v1'), ('k2', 'v2')])
for k, v in dict.items():
    print(k, v)
```

**copy()** - Cria cópia rasa:
```python
copia = dict.copy()  # Cria novo dicionário
```

### 📊 Views Dinâmicas
* keys(), values() e items() retornam **views**
* Views refletem mudanças no dicionário original
* Podem ser convertidas para listas com `list()`
* São eficientes em memória

### ✅ Casos de Uso
* **keys()**: Verificar chaves, iterar sobre chaves
* **values()**: Processar todos os valores, calcular estatísticas
* **items()**: Iterar sobre pares, transformar dicionário
* **copy()**: Criar versão modificada sem alterar original

````python
# Exemplo 1: keys() - Acessando todas as chaves 🔑
aluno = {"nome": "Pedro", "idade": 22, "curso": "Ciência da Computação", "período": 7}

chaves = aluno.keys()
print(f"🔑 Chaves: {chaves}")
print(f"🔑 Como lista: {list(chaves)}")
print(f"🔍 Tipo: {type(chaves)}")

# Verificando existência
print(f"\n✅ 'nome' nas chaves? {'nome' in chaves}")
print(f"❌ 'email' nas chaves? {'email' in chaves}")

# Exemplo 2: values() - Acessando todos os valores 📊
notas = {"Matemática": 8.5, "Física": 7.0, "Química": 9.0, "História": 8.0}

valores = notas.values()
print(f"\n📊 Notas: {list(valores)}")
print(f"📈 Maior nota: {max(valores)}")
print(f"📉 Menor nota: {min(valores)}")
print(f"🧭 Média: {sum(valores) / len(valores):.2f}")

# Exemplo 3: items() - Acessando pares chave-valor 🔗
produto = {"id": 101, "nome": "Mouse", "preco": 45.90, "estoque": 50}

itens = produto.items()
print(f"\n🔗 Items: {list(itens)}")
print(f"\n📦 Detalhes do produto:")
for chave, valor in produto.items():
    print(f"   {chave}: {valor}")

# Exemplo 4: Views dinâmicas 🔄
config = {"debug": False, "port": 8080}
chaves_view = config.keys()

print(f"\n🔍 Chaves antes: {list(chaves_view)}")
config["timeout"] = 30  # Adicionar novo item
print(f"🔄 Chaves depois: {list(chaves_view)}")
print("✅ View reflete mudanças automaticamente!")

# Exemplo 5: copy() - Criando cópias 📋
original = {"a": 1, "b": 2, "c": 3}
copia = original.copy()

print(f"\n📊 Original: {original}")
print(f"📋 Cópia: {copia}")

# Modificando cópia não afeta original
copia["d"] = 4
copia["a"] = 10

print(f"\n✏️ Cópia modificada: {copia}")
print(f"🚫 Original inalterado: {original}")

# Exemplo 6: Usando keys() para filtrar 🔍
estudantes = {
    "001": {"nome": "Ana", "nota": 8.5},
    "002": {"nome": "Bruno", "nota": 7.0},
    "003": {"nome": "Carla", "nota": 9.0}
}

print(f"\n🎓 IDs dos estudantes: {list(estudantes.keys())}")
print(f"🔢 Total de estudantes: {len(estudantes.keys())}")

# Exemplo 7: Convertendo views para listas 📝
elemento = {"Z": 6, "Nome": "Carbono", "Simbolo": "C", "Massa": 12.011}

lista_chaves = list(elemento.keys())
lista_valores = list(elemento.values())
lista_pares = list(elemento.items())

print(f"\n🧪 Elemento Carbono:")
print(f"   Chaves como lista: {lista_chaves}")
print(f"   Valores como lista: {lista_valores}")
print(f"   Pares como lista: {lista_pares}")
````

## 4️⃣ Iteração sobre Dicionários

### 📖 Conceito
Existem várias formas de iterar sobre dicionários, cada uma adequada para diferentes necessidades. A escolha depende se você precisa de chaves, valores ou ambos.

### 🔧 Formas de Iteração

**Iterar sobre chaves (padrão):**
```python
for chave in dict:
    print(chave)
# Equivalente a:
for chave in dict.keys():
    print(chave)
```

**Iterar sobre valores:**
```python
for valor in dict.values():
    print(valor)
```

**Iterar sobre pares (mais comum):**
```python
for chave, valor in dict.items():
    print(chave, valor)
```

**Com enumerate():**
```python
for indice, (chave, valor) in enumerate(dict.items()):
    print(indice, chave, valor)
```

### ✅ Melhores Práticas
* Use `items()` quando precisar de chave E valor
* Use `values()` quando só precisar dos valores
* Evite acessar `dict[key]` dentro do loop de `keys()`
* Use desempacotamento: `for k, v in dict.items()`

````python
# Exemplo 1: Iteração sobre chaves 🔑
pessoa = {"nome": "Lucas", "idade": 28, "profissão": "Engenheiro", "cidade": "Brasília"}

print("🔑 Iterando sobre chaves:")
for chave in pessoa:
    print(f"   {chave}")

# Ou explicitamente com keys()
print("\n🔑 Com keys() (equivalente):")
for chave in pessoa.keys():
    print(f"   {chave}")

# Exemplo 2: Iteração sobre valores 📊
notas = {"Português": 7.5, "Matemática": 8.0, "Inglês": 9.0, "Biologia": 7.0}

print("\n📊 Notas:")
for nota in notas.values():
    print(f"   {nota}")

total = sum(notas.values())
media = total / len(notas)
print(f"\n🧭 Média das notas: {media:.2f}")

# Exemplo 3: Iteração sobre pares (mais comum) 🔗
produtos = {
    "Notebook": 3500.00,
    "Mouse": 45.90,
    "Teclado": 250.00,
    "Monitor": 1200.00
}

print("\n📦 Lista de produtos:")
for produto, preco in produtos.items():
    print(f"   {produto}: R$ {preco:.2f}")

# Exemplo 4: Com enumerate() para índices 🔢
estudantes = {
    "001": "Ana Silva",
    "002": "Bruno Costa",
    "003": "Carla Souza",
    "004": "Daniel Lima"
}

print("\n🎓 Estudantes (com índice):")
for indice, (matricula, nome) in enumerate(estudantes.items(), 1):
    print(f"   {indice}. Matrícula {matricula}: {nome}")

# Exemplo 5: Filtrando durante iteração 🔍
inventario = {
    "maçã": 50,
    "banana": 30,
    "laranja": 5,
    "uva": 2,
    "melancia": 15
}

print("\n⚠️ Produtos com estoque baixo (< 10):")
for fruta, quantidade in inventario.items():
    if quantidade < 10:
        print(f"   {fruta}: {quantidade} unidades")

# Exemplo 6: Construindo novo dicionário durante iteração 🔄
precos_originais = {"Camiseta": 50, "Calça": 100, "Tênis": 200}
desconto = 0.20  # 20%

precos_promocionais = {}
for item, preco in precos_originais.items():
    preco_promocional = preco * (1 - desconto)
    precos_promocionais[item] = preco_promocional

print(f"\n🏷️ Preços originais: {precos_originais}")
print(f"💰 Preços promocionais (20% off): {precos_promocionais}")

# Exemplo 7: Acessando valores de forma segura ✅
config = {"host": "localhost", "port": 8080, "debug": True}

print("\n⚙️ Configurações:")
for chave in ["host", "port", "timeout", "debug"]:
    # Usando get() para evitar KeyError
    valor = config.get(chave, "N/A")
    print(f"   {chave}: {valor}")
````

## 5️⃣ Dicionários Aninhados

### 📖 Conceito
Dicionários podem conter outros dicionários como valores, criando estruturas hierárquicas. Isso permite modelar dados complexos e relacionados.

### 🔧 Estrutura Aninhada
```python
empresa = {
    "departamentos": {
        "TI": {
            "funcionários": 15,
            "gerente": "João"
        },
        "RH": {
            "funcionários": 8,
            "gerente": "Maria"
        }
    }
}

# Acessar valores aninhados
empresa["departamentos"]["TI"]["gerente"]  # "João"
```

### 📊 Níveis de Aninhamento
* **1 nível**: Dicionário simples
* **2 níveis**: Dicionário de dicionários (mais comum)
* **3+ níveis**: Possível, mas considere classes/JSON

### ✅ Casos de Uso
* Estruturar dados de APIs (JSON)
* Modelar entidades relacionadas
* Configurar sistemas complexos
* Armazenar dados hierárquicos

### ⚠️ Cuidados
* Acesso pode lançar KeyError em múltiplos níveis
* Use `get()` em cada nível para segurança
* Copy() cria cópia rasa (shallow), não profunda

````python
# Exemplo 1: Estrutura de alunos com informações completas 🎓
alunos = {
    "001": {
        "nome": "Ana Silva",
        "idade": 20,
        "curso": "Engenharia",
        "notas": {"Calc1": 8.5, "Física1": 7.0, "Programação": 9.0}
    },
    "002": {
        "nome": "Carlos Lima",
        "idade": 22,
        "curso": "Medicina",
        "notas": {"Anatomia": 9.0, "Fisiologia": 8.5, "Bioquímica": 7.5}
    }
}

print("🎓 Aluno 001:")
print(f"   Nome: {alunos['001']['nome']}")
print(f"   Curso: {alunos['001']['curso']}")
print(f"   Notas: {alunos['001']['notas']}")

# Calculando média
notas_001 = alunos['001']['notas'].values()
media_001 = sum(notas_001) / len(notas_001)
print(f"   Média: {media_001:.2f}")

# Exemplo 2: Empresa com departamentos 🏢
empresa = {
    "nome": "Tech Solutions",
    "fundacao": 2010,
    "departamentos": {
        "TI": {
            "funcionários": 25,
            "gerente": "Roberto Santos",
            "orçamento": 500000
        },
        "RH": {
            "funcionários": 10,
            "gerente": "Mariana Costa",
            "orçamento": 200000
        },
        "Vendas": {
            "funcionários": 30,
            "gerente": "Paulo Silva",
            "orçamento": 300000
        }
    }
}

print(f"\n🏢 Empresa: {empresa['nome']}")
print(f"\n📄 Departamentos:")
for dept, info in empresa['departamentos'].items():
    print(f"   {dept}:")
    print(f"      Gerente: {info['gerente']}")
    print(f"      Funcionários: {info['funcionários']}")
    print(f"      Orçamento: R$ {info['orçamento']:,.2f}")

# Exemplo 3: Elementos químicos com propriedades detalhadas 🧪
elementos = {
    "H": {
        "nome": "Hidrogênio",
        "numero_atomico": 1,
        "massa": 1.008,
        "propriedades": {
            "estado": "gasoso",
            "densidade": 0.00008988,
            "ponto_fusao": -259.14
        }
    },
    "O": {
        "nome": "Oxigênio",
        "numero_atomico": 8,
        "massa": 15.999,
        "propriedades": {
            "estado": "gasoso",
            "densidade": 0.001429,
            "ponto_fusao": -218.79
        }
    }
}

print(f"\n🧪 Elemento H:")
print(f"   Nome: {elementos['H']['nome']}")
print(f"   Número Atômico: {elementos['H']['numero_atomico']}")
print(f"   Estado: {elementos['H']['propriedades']['estado']}")
print(f"   Densidade: {elementos['H']['propriedades']['densidade']} g/cm³")

# Exemplo 4: Acesso seguro com get() aninhado ✅
produto = {
    "id": 123,
    "nome": "Notebook",
    "especificacoes": {
        "processador": "Intel i7",
        "ram": "16GB"
    }
}

# Acesso seguro em múltiplos níveis
processador = produto.get("especificacoes", {}).get("processador", "N/A")
armazenamento = produto.get("especificacoes", {}).get("armazenamento", "N/A")

print(f"\n💻 Produto:")
print(f"   Processador: {processador}")
print(f"   Armazenamento: {armazenamento}")

# Exemplo 5: Modificando valores aninhados ✏️
config = {
    "servidor": {
        "host": "localhost",
        "port": 8080,
        "ssl": {
            "ativo": False,
            "certificado": None
        }
    }
}

print(f"\n⚙️ Config original: {config}")

# Modificar valor aninhado
config["servidor"]["port"] = 9090
config["servidor"]["ssl"]["ativo"] = True
config["servidor"]["ssl"]["certificado"] = "/path/to/cert"

print(f"✏️ Config modificada: {config}")

# Exemplo 6: Iterando sobre estrutura aninhada 🔁
inventario_lojas = {
    "Loja Centro": {
        "Notebook": 10,
        "Mouse": 50,
        "Teclado": 30
    },
    "Loja Norte": {
        "Notebook": 5,
        "Mouse": 25,
        "Teclado": 15
    }
}

print(f"\n🏦 Inventário por loja:")
for loja, produtos in inventario_lojas.items():
    print(f"\n   {loja}:")
    for produto, quantidade in produtos.items():
        print(f"      {produto}: {quantidade} unidades")
    print(f"      Total de itens: {sum(produtos.values())}")
````

## 6️⃣ Dict Comprehension (Compreensão de Dicionários)

### 📖 Conceito
Dict comprehension é uma forma concisa e elegante de criar dicionários usando uma sintaxe similar a list comprehension. Permite transformar, filtrar e criar dicionários em uma única linha.

### 🔧 Sintaxe Básica
```python
# Forma geral
{chave_expr: valor_expr for item in iterable}

# Com condicional
{chave_expr: valor_expr for item in iterable if condicao}

# De outro dicionário
{k: v*2 for k, v in dict.items()}
```

### 📊 Exemplos de Padrões

**Criar de range:**
```python
quadrados = {x: x**2 for x in range(5)}
# {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
```

**Transformar dicionário:**
```python
precos = {"a": 10, "b": 20}
com_desconto = {k: v*0.9 for k, v in precos.items()}
```

**Filtrar dicionário:**
```python
notas = {"a": 7, "b": 5, "c": 9}
aprovados = {k: v for k, v in notas.items() if v >= 7}
```

**Inverter chave-valor:**
```python
original = {"a": 1, "b": 2}
invertido = {v: k for k, v in original.items()}
# {1: 'a', 2: 'b'}
```

### ✅ Quando Usar
* Transformar todos os valores
* Filtrar itens por condição
* Criar mapeamentos a partir de listas
* Inverter dicionários
* Converter tipos

````python
# Exemplo 1: Criar dicionário de quadrados 🔢
quadrados = {x: x**2 for x in range(1, 11)}
print(f"🔢 Quadrados de 1 a 10: {quadrados}")

# Exemplo 2: Criar de listas (zip) 🔗
nomes = ["Ana", "Bruno", "Carla"]
idades = [20, 22, 21]

pessoas = {nome: idade for nome, idade in zip(nomes, idades)}
print(f"\n👤 Pessoas: {pessoas}")

# Exemplo 3: Transformar valores (desconto) 💰
precos_originais = {"Notebook": 3500, "Mouse": 50, "Teclado": 250}
desconto = 0.15  # 15%

precos_promocionais = {produto: preco * (1 - desconto) 
                        for produto, preco in precos_originais.items()}

print(f"\n💰 Preços originais: {precos_originais}")
print(f"🏷️ Com 15% desconto: {precos_promocionais}")

# Exemplo 4: Filtrar dicionário (aprovados) ✅
notas = {
    "Ana": 8.5,
    "Bruno": 6.0,
    "Carla": 9.0,
    "Daniel": 5.5,
    "Elena": 7.5
}

aprovados = {aluno: nota for aluno, nota in notas.items() if nota >= 7.0}
reprovados = {aluno: nota for aluno, nota in notas.items() if nota < 7.0}

print(f"\n🎓 Todos: {notas}")
print(f"✅ Aprovados (>= 7.0): {aprovados}")
print(f"❌ Reprovados (< 7.0): {reprovados}")

# Exemplo 5: Inverter chave-valor 🔄
codigo_para_cidade = {"SP": "São Paulo", "RJ": "Rio de Janeiro", "MG": "Minas Gerais"}
cidade_para_codigo = {cidade: codigo for codigo, cidade in codigo_para_cidade.items()}

print(f"\n🗺️ Original: {codigo_para_cidade}")
print(f"🔄 Invertido: {cidade_para_codigo}")

# Exemplo 6: Contar caracteres 🔢
palavras = ["python", "java", "javascript", "c", "ruby"]
comprimentos = {palavra: len(palavra) for palavra in palavras}

print(f"\n💬 Palavras: {palavras}")
print(f"🔢 Comprimentos: {comprimentos}")

# Exemplo 7: Criar categoria por preço 🏷️
produtos_precos = {
    "Caneta": 2.50,
    "Caderno": 15.00,
    "Mochila": 80.00,
    "Lápis": 1.50,
    "Estojo": 25.00
}

def categorizar_preco(preco):
    if preco < 5:
        return "Barato"
    elif preco < 50:
        return "Médio"
    else:
        return "Caro"

categorias = {produto: categorizar_preco(preco) 
              for produto, preco in produtos_precos.items()}

print(f"\n🏷️ Categorias por preço:")
for produto, categoria in categorias.items():
    preco = produtos_precos[produto]
    print(f"   {produto}: R$ {preco:.2f} - {categoria}")

# Exemplo 8: Múltiplas condições (alunos com bônus) 🌟
notas_finais = {
    "Ana": 8.5,
    "Bruno": 9.5,
    "Carla": 7.0,
    "Daniel": 9.8,
    "Elena": 8.0
}

# Adicionar bônus de 0.5 para notas >= 9.0
com_bonus = {aluno: (nota + 0.5 if nota >= 9.0 else nota) 
             for aluno, nota in notas_finais.items()}

print(f"\n🌟 Notas com bônus (para >= 9.0):")
for aluno in notas_finais:
    original = notas_finais[aluno]
    final = com_bonus[aluno]
    bonus = " +BÔNUS" if final > original else ""
    print(f"   {aluno}: {original} → {final}{bonus}")
````

````python
# Exemplo 1: Contador de frequências 📊
palavras = ["python", "java", "python", "javascript", "python", "java", "c++"]

# Método manual
contador = {}
for palavra in palavras:
    contador[palavra] = contador.get(palavra, 0) + 1

print(f"📊 Contador de palavras:")
for palavra, freq in sorted(contador.items()):
    print(f"   {palavra}: {freq}x")

# Exemplo 2: Agrupamento de dados por categoria 📦
produtos = [
    {"nome": "Notebook", "categoria": "Eletrônicos", "preco": 3500},
    {"nome": "Mouse", "categoria": "Eletrônicos", "preco": 50},
    {"nome": "Cadeira", "categoria": "Móveis", "preco": 800},
    {"nome": "Mesa", "categoria": "Móveis", "preco": 1200},
    {"nome": "Teclado", "categoria": "Eletrônicos", "preco": 250}
]

grupos = {}
for produto in produtos:
    categoria = produto["categoria"]
    grupos.setdefault(categoria, []).append(produto["nome"])

print(f"\n📦 Produtos por categoria:")
for categoria, itens in grupos.items():
    print(f"   {categoria}: {itens}")

# Exemplo 3: Mapeamento ID → Nome (índice) 🔍
funcionarios = [
    {"id": 101, "nome": "Ana Silva", "cargo": "Engenheira"},
    {"id": 102, "nome": "Bruno Costa", "cargo": "Designer"},
    {"id": 103, "nome": "Carla Souza", "cargo": "Gerente"}
]

# Criar índice para busca rápida por ID
indice_id = {f["id"]: f for f in funcionarios}

print(f"\n🔍 Buscar funcionário por ID:")
id_busca = 102
if id_busca in indice_id:
    func = indice_id[id_busca]
    print(f"   ID {id_busca}: {func['nome']} - {func['cargo']}")

# Exemplo 4: Agregação de vendas 💰
vendas = [
    {"vendedor": "Ana", "valor": 1500},
    {"vendedor": "Bruno", "valor": 2000},
    {"vendedor": "Ana", "valor": 1200},
    {"vendedor": "Carla", "valor": 1800},
    {"vendedor": "Bruno", "valor": 2500}
]

total_por_vendedor = {}
for venda in vendas:
    vendedor = venda["vendedor"]
    valor = venda["valor"]
    total_por_vendedor[vendedor] = total_por_vendedor.get(vendedor, 0) + valor

print(f"\n💰 Total de vendas por vendedor:")
for vendedor, total in sorted(total_por_vendedor.items(), key=lambda x: x[1], reverse=True):
    print(f"   {vendedor}: R$ {total:,.2f}")

# Exemplo 5: Cache/Memoização ⚡
# Simular função cara sem cache
import time

calculos_sem_cache = 0
def fibonacci_sem_cache(n):
    global calculos_sem_cache
    calculos_sem_cache += 1
    if n <= 1:
        return n
    return fibonacci_sem_cache(n-1) + fibonacci_sem_cache(n-2)

# Com cache
cache = {}
calculos_com_cache = 0
def fibonacci_com_cache(n):
    global calculos_com_cache
    calculos_com_cache += 1
    if n in cache:
        return cache[n]
    if n <= 1:
        cache[n] = n
    else:
        cache[n] = fibonacci_com_cache(n-1) + fibonacci_com_cache(n-2)
    return cache[n]

print(f"\n⚡ Fibonacci(10):")
calculos_sem_cache = 0
resultado1 = fibonacci_sem_cache(10)
print(f"   Sem cache: {resultado1} (cálculos: {calculos_sem_cache})")

calculos_com_cache = 0
resultado2 = fibonacci_com_cache(10)
print(f"   Com cache: {resultado2} (cálculos: {calculos_com_cache})")
print(f"   📈 Redução: {calculos_sem_cache - calculos_com_cache} cálculos")

# Exemplo 6: Inversão de mapeamento 🔄
codigo_estado = {
    "SP": "São Paulo",
    "RJ": "Rio de Janeiro",
    "MG": "Minas Gerais",
    "ES": "Espírito Santo"
}

estado_codigo = {v: k for k, v in codigo_estado.items()}

print(f"\n🔄 Mapeamentos:")
print(f"   Código → Estado: {codigo_estado}")
print(f"   Estado → Código: {estado_codigo}")
print(f"   Buscar 'Rio de Janeiro': {estado_codigo['Rio de Janeiro']}")

# Exemplo 7: Estatísticas de notas 📈
notas_turma = {
    "Ana": [8.5, 9.0, 7.5, 9.5],
    "Bruno": [6.0, 7.0, 6.5, 7.5],
    "Carla": [9.5, 9.0, 10.0, 9.5],
    "Daniel": [7.0, 8.0, 7.5, 8.5]
}

estatisticas = {}
for aluno, notas in notas_turma.items():
    estatisticas[aluno] = {
        "media": sum(notas) / len(notas),
        "maior": max(notas),
        "menor": min(notas)
    }

print(f"\n📈 Estatísticas da turma:")
for aluno, stats in estatisticas.items():
    print(f"   {aluno}:")
    print(f"      Média: {stats['media']:.2f}")
    print(f"      Maior: {stats['maior']}")
    print(f"      Menor: {stats['menor']}")
````

## 8️⃣ Boas Práticas com Dicionários

### 🎯 10 Recomendações Essenciais

#### 1️⃣ Use get() para Acesso Seguro
```python
# ✅ Bom - Não lança erro
valor = dict.get(chave, valor_padrao)

# ❌ Evite - Pode lançar KeyError
valor = dict[chave] if chave in dict else valor_padrao
```

#### 2️⃣ Use setdefault() para Inicialização
```python
# ✅ Bom - Inicializa se não existir
dict.setdefault(chave, []).append(item)

# ❌ Evite - Mais verboso
if chave not in dict:
    dict[chave] = []
dict[chave].append(item)
```

#### 3️⃣ Use items() para Iterar
```python
# ✅ Bom - Acessa chave e valor diretamente
for chave, valor in dict.items():
    print(chave, valor)

# ❌ Evite - Acesso extra
for chave in dict:
    valor = dict[chave]
    print(chave, valor)
```

#### 4️⃣ Use Dict Comprehension
```python
# ✅ Bom - Conciso
quadrados = {x: x**2 for x in range(10)}

# ❌ Evite - Mais verboso
quadrados = {}
for x in range(10):
    quadrados[x] = x**2
```

#### 5️⃣ Use update() para Múltiplas Atualizações
```python
# ✅ Bom
config.update({"host": "localhost", "port": 8080})

# ❌ Evite
config["host"] = "localhost"
config["port"] = 8080
```

#### 6️⃣ Use in para Verificar Chaves
```python
# ✅ Bom - Direto e claro
if "chave" in dict:
    ...

# ❌ Evite
if "chave" in dict.keys():
    ...
```

#### 7️⃣ Escolha Chaves Imutáveis
```python
# ✅ Bom - String, número, tupla
dict["chave"] = valor
dict[42] = valor
dict[(1, 2)] = valor

# ❌ Erro - Lista é mutável
dict[[1, 2]] = valor  # TypeError
```

#### 8️⃣ Use pop() com Valor Padrão
```python
# ✅ Bom - Não lança erro
valor = dict.pop(chave, None)

# ❌ Evite - Requer try/except
try:
    valor = dict.pop(chave)
except KeyError:
    valor = None
```

#### 9️⃣ Documente Estruturas Complexas
```python
# ✅ Bom - Documenta estrutura
def processar_usuario(usuario):
    """
    Args:
        usuario: dict com 'id' (int), 'nome' (str), 'email' (str)
    """
    ...
```

#### 🔟 Use copy() para Evitar Efeitos Colaterais
```python
# ✅ Bom - Cria cópia independente
config_teste = config.copy()
config_teste["debug"] = True  # Não afeta config

# ❌ Evite - Altera original
config_teste = config
config_teste["debug"] = True  # Afeta config!
```

### 📊 Complexidade de Operações
* **Acesso dict[key]**: O(1) - Muito rápido
* **in (verificação)**: O(1)
* **get()**: O(1)
* **pop()**: O(1)
* **update()**: O(len(outro_dict))
* **Iteração**: O(n)

### 🎓 Resumo Final
* Dicionários são a estrutura para **mapeamento chave-valor**
* Use **get()** para acesso seguro
* Use **items()** para iterar sobre pares
* **Dict comprehension** para criação concisa
* Perfeito para: índices, contadores, cache, configurações
