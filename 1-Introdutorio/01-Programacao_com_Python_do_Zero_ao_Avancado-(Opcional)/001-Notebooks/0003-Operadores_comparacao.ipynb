# Operadores de Comparação em Python

Este notebook cobre os operadores de comparação que retornam valores booleanos (`True` ou `False`):

* **Igual a** (`==`)
* **Diferente de** (`!=`)
* **Maior que** (`>`)
* **Menor que** (`<`)
* **Maior ou igual a** (`>=`)
* **Menor ou igual a** (`<=`)

Operadores de comparação são fundamentais para:
* Estruturas condicionais (if/else)
* Loops com condições
* Filtragem de dados
* Validações lógicas

## 1. Igualdade (==)

O operador `==` verifica se dois valores são **iguais**.

**Sintaxe:** `valor1 == valor2`

**Retorno:**
* `True` se os valores forem iguais
* `False` se os valores forem diferentes

**Importante:** Não confunda `==` (comparação) com `=` (atribuição)!

````python
# Comparando números inteiros
a = 10
b = 10
c = 5

print(f"{a} == {b} ? {a == b}")  # True
print(f"{a} == {c} ? {a == c}")  # False

# Comparando strings
nome1 = "Python"
nome2 = "Python"
nome3 = "python"  # Case sensitive!

print(f"\n'{nome1}' == '{nome2}' ? {nome1 == nome2}")  # True
print(f"'{nome1}' == '{nome3}' ? {nome1 == nome3}")  # False (maiúsculas/minúsculas importam)

# Comparando floats
preco1 = 19.99
preco2 = 19.99
preco3 = 20.00

print(f"\nR$ {preco1} == R$ {preco2} ? {preco1 == preco2}")  # True
print(f"R$ {preco1} == R$ {preco3} ? {preco1 == preco3}")  # False
````

## 2. Diferente de (!=)

O operador `!=` verifica se dois valores são **diferentes**.

**Sintaxe:** `valor1 != valor2`

**Retorno:**
* `True` se os valores forem diferentes
* `False` se os valores forem iguais

**Nota:** `!=` é o oposto lógico de `==`

````python
# Comparando números
x = 15
y = 20
z = 15

print(f"{x} != {y} ? {x != y}")  # True (são diferentes)
print(f"{x} != {z} ? {x != z}")  # False (são iguais)

# Comparando strings
status1 = "ativo"
status2 = "inativo"
status3 = "ativo"

print(f"\n'{status1}' != '{status2}' ? {status1 != status2}")  # True
print(f"'{status1}' != '{status3}' ? {status1 != status3}")  # False

# Comparando tipos diferentes
numero = 10
texto = "10"

print(f"\n{numero} (int) != '{texto}' (str) ? {numero != texto}")  # True (tipos diferentes)
````

## 3. Maior que (>)

O operador `>` verifica se o valor da esquerda é **maior** que o valor da direita.

**Sintaxe:** `valor1 > valor2`

**Retorno:**
* `True` se valor1 for maior que valor2
* `False` caso contrário

**Casos de uso:**
* Validações de idade, preço, quantidade
* Filtragem de dados numéricos
* Ordenação e comparações

````python
# Comparando números
a = 50
b = 30
c = 50

print(f"{a} > {b} ? {a > b}")  # True
print(f"{b} > {a} ? {b > a}")  # False
print(f"{a} > {c} ? {a > c}")  # False (são iguais, não maior)

# Exemplo prático: verificação de idade
idade = 25
idade_minima = 18

maior_idade = idade > idade_minima
print(f"\nIdade {idade} > {idade_minima}? {maior_idade}")

# Comparando preços
preco_produto = 150.00
orcamento = 100.00

acima_orcamento = preco_produto > orcamento
print(f"\nPreço R$ {preco_produto:.2f} > Orçamento R$ {orcamento:.2f}? {acima_orcamento}")
````

## 4. Menor que (<)

O operador `<` verifica se o valor da esquerda é **menor** que o valor da direita.

**Sintaxe:** `valor1 < valor2`

**Retorno:**
* `True` se valor1 for menor que valor2
* `False` caso contrário

````python
# Comparando números
a = 10
b = 25
c = 10

print(f"{a} < {b} ? {a < b}")  # True
print(f"{b} < {a} ? {b < a}")  # False
print(f"{a} < {c} ? {a < c}")  # False (são iguais)

# Exemplo prático: verificação de estoque
estoque_atual = 5
estoque_minimo = 10

abaixo_minimo = estoque_atual < estoque_minimo
print(f"\nEstoque {estoque_atual} < Mínimo {estoque_minimo}? {abaixo_minimo}")
if abaixo_minimo:
    print("⚠️ Alerta: Estoque abaixo do mínimo!")

# Comparando temperaturas
temperatura = 15.5
limite_congelamento = 0

print(f"\nTemperatura {temperatura}°C < {limite_congelamento}°C? {temperatura < limite_congelamento}")
````

## 5. Maior ou Igual a (>=)

O operador `>=` verifica se o valor da esquerda é **maior ou igual** ao valor da direita.

**Sintaxe:** `valor1 >= valor2`

**Retorno:**
* `True` se valor1 for maior ou igual a valor2
* `False` se valor1 for menor que valor2

````python
# Comparando números
a = 30
b = 20
c = 30

print(f"{a} >= {b} ? {a >= b}")  # True (maior)
print(f"{a} >= {c} ? {a >= c}")  # True (igual)
print(f"{b} >= {a} ? {b >= a}")  # False (menor)

# Exemplo prático: verificação de nota mínima
nota_aluno = 7.0
nota_minima = 7.0

aprovado = nota_aluno >= nota_minima
print(f"\nNota {nota_aluno} >= Nota mínima {nota_minima}? {aprovado}")
if aprovado:
    print("✅ Aprovado!")
else:
    print("❌ Reprovado")

# Exemplo: elegibilidade para desconto
valor_compra = 100.00
minimo_desconto = 100.00

tem_desconto = valor_compra >= minimo_desconto
print(f"\nCompra R$ {valor_compra:.2f} >= Mínimo R$ {minimo_desconto:.2f}? {tem_desconto}")
````

## 6. Menor ou Igual a (<=)

O operador `<=` verifica se o valor da esquerda é **menor ou igual** ao valor da direita.

**Sintaxe:** `valor1 <= valor2`

**Retorno:**
* `True` se valor1 for menor ou igual a valor2
* `False` se valor1 for maior que valor2

````python
# Comparando números
a = 15
b = 20
c = 15

print(f"{a} <= {b} ? {a <= b}")  # True (menor)
print(f"{a} <= {c} ? {a <= c}")  # True (igual)
print(f"{b} <= {a} ? {b <= a}")  # False (maior)

# Exemplo prático: verificação de limite
velocidade_atual = 80
limite_velocidade = 80

dentro_limite = velocidade_atual <= limite_velocidade
print(f"\nVelocidade {velocidade_atual} km/h <= Limite {limite_velocidade} km/h? {dentro_limite}")

# Exemplo: capacidade de sala
participantes = 45
capacidade_maxima = 50

pode_entrar = participantes <= capacidade_maxima
print(f"\nParticipantes {participantes} <= Capacidade {capacidade_maxima}? {pode_entrar}")
if pode_entrar:
    print("✅ Ainda há vagas!")
else:
    print("⚠️ Sala cheia!")
````

## 7. Exemplo Consolidado: Todos os Operadores

Vamos aplicar todos os operadores de comparação com os mesmos valores para entender melhor como funcionam.

````python
# Definindo dois valores para demonstrar todos os operadores
n1 = 40
n2 = 25

print(f"Valores: n1 = {n1}, n2 = {n2}\n")
print("=" * 60)

# Aplicando todos os operadores de comparação
print(f"n1 == n2  (igual a)          : {n1} == {n2}  = {n1 == n2}")
print(f"n1 != n2  (diferente de)     : {n1} != {n2}  = {n1 != n2}")
print(f"n1 > n2   (maior que)        : {n1} > {n2}   = {n1 > n2}")
print(f"n1 < n2   (menor que)        : {n1} < {n2}   = {n1 < n2}")
print(f"n1 >= n2  (maior ou igual)   : {n1} >= {n2}  = {n1 >= n2}")
print(f"n1 <= n2  (menor ou igual)   : {n1} <= {n2}  = {n1 <= n2}")

print("=" * 60)

# Testando com valores iguais
print(f"\nQuando os valores são iguais (n1 = n2 = 30):\n")
n1 = n2 = 30
print(f"n1 == n2 : {n1 == n2}  (True)")
print(f"n1 != n2 : {n1 != n2}  (False)")
print(f"n1 > n2  : {n1 > n2}   (False)")
print(f"n1 < n2  : {n1 < n2}   (False)")
print(f"n1 >= n2 : {n1 >= n2}  (True)")
print(f"n1 <= n2 : {n1 <= n2}  (True)")
````

## 8. Encadeamento de Comparações

Python permite encadear múltiplas comparações de forma elegante:

**Sintaxe:** `valor1 < valor2 < valor3`

Isso é equivalente a: `(valor1 < valor2) and (valor2 < valor3)`

**Vantagens:**
* Código mais legível
* Menos repetição de variáveis
* Performance otimizada (avaliação em curto-circuito)

````python
# Verificando se um valor está em um intervalo
nota = 7.5

# Forma tradicional (mais verbosa)
if nota >= 0 and nota <= 10:
    print(f"Nota {nota}: Válida (forma tradicional)")

# Forma encadeada (mais pythonica)
if 0 <= nota <= 10:
    print(f"Nota {nota}: Válida (forma encadeada)")

# Múltiplos encadeamentos
x = 5
y = 10
z = 15

print(f"\n{x} < {y} < {z} ? {x < y < z}")  # True
print(f"{x} < {z} < {y} ? {x < z < y}")  # False

# Exemplo prático: classificação de temperatura
temperatura = 22

if 18 <= temperatura <= 25:
    print(f"\nTemperatura {temperatura}°C: Confortável! 🌡️")
elif temperatura < 18:
    print(f"\nTemperatura {temperatura}°C: Frio! ❄️")
else:
    print(f"\nTemperatura {temperatura}°C: Quente! 🔥")

# Verificando ordem crescente
a, b, c = 5, 10, 15
print(f"\n{a} <= {b} <= {c} (ordem crescente)? {a <= b <= c}")
````

## Boas Práticas com Operadores de Comparação

### 1. Não Confunda `=` com `==`
```python
# ❌ ERRADO - Atribuição em vez de comparação
if x = 10:  # SyntaxError!
    print("Erro")

# ✅ CORRETO - Comparação
if x == 10:
    print("Correto")
```

### 2. Comparação de Tipos
* Python permite comparar tipos diferentes, mas tenha cuidado:
  * `10 == "10"` retorna `False` (int vs string)
  * Use `int()`, `str()`, `float()` para converter quando necessário

### 3. Comparação de Floats
* Floats podem ter imprecisões:
```python
# Cuidado com comparações exatas
0.1 + 0.2 == 0.3  # Pode retornar False!

# Use tolerância para floats
import math
math.isclose(0.1 + 0.2, 0.3)  # Melhor abordagem
```

### 4. Use Encadeamento
* Prefira `0 <= x <= 10` em vez de `x >= 0 and x <= 10`
* Mais legível e pythonica

### 5. Strings são Case-Sensitive
* `"Python" == "python"` retorna `False`
* Use `.lower()` ou `.upper()` para comparações case-insensitive:
```python
nome1.lower() == nome2.lower()  # Ignora maiúsculas/minúsculas
```

### 6. Valores Booleanos
* Não precisa comparar explicitamente com `True` ou `False`:
```python
# ❌ DESNECESSÁRIO
if ativo == True:
    pass

# ✅ MELHOR
if ativo:
    pass
```

### 7. Use `is` para Identidade, Não Igualdade
* `==` compara valores
* `is` compara identidade (mesmo objeto na memória)
```python
if valor is None:  # Correto para None
if valor == None:  # Funciona, mas não é recomendado
```
