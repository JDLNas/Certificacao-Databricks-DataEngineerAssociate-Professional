# 📝 Manipulação de Strings em Python

## 📋 Visão Geral
Este notebook explora a manipulação de strings (texto) em Python, uma das habilidades mais importantes para processamento de dados. Você aprenderá a criar, modificar, buscar, formatar e validar strings usando os métodos nativos do Python.

* Limpar espaços em branco com strip(), lstrip(), rstrip()
* Formatar e alinhar texto com center(), ljust(), rjust(), zfill()

## 1️⃣ Strings Básicas

### 📖 Conceito
Strings são sequências de caracteres usadas para armazenar e manipular texto. Em Python, strings são **imutáveis** - não podem ser alteradas após criação, mas podem gerar novas strings.

### 🔧 Criação de Strings
```python
# Aspas simples ou duplas
texto1 = 'Python'
texto2 = "Python"

# Aspas triplas (multilinhas)
texto3 = '''Texto
com várias
linhas'''

# String vazia
vazia = ""
```

### 📊 Indexação e Fatiamento
```python
texto = "Python"
texto[0]      # 'P' (primeiro caractere)
texto[-1]     # 'n' (último caractere)
texto[2:5]    # 'tho' (do índice 2 até 4)
texto[:3]     # 'Pyt' (do início até índice 2)
texto[3:]     # 'hon' (do índice 3 até o fim)
len(texto)    # 6 (comprimento)
```

### ✅ Características
* **Imutáveis**: Não podem ser alteradas no local
* **Iteráveis**: Podem ser percorridas com loops
* **Indexadas**: Acesso por posição (0-based)
* **Unicode**: Suportam caracteres internacionais

````python
# Exemplo 1: Criação de strings 📝
nome = "Ana Silva"
cidade = 'São Paulo'
codigo = '''ABC-123
DEF-456'''
print(f"👤 Nome: {nome}")
print(f"🏙️ Cidade: {cidade}")
print(f"🔢 Código:\n{codigo}")

# Exemplo 2: Indexação 🔍
produto = "Notebook"
print(f"\n💻 Produto: {produto}")
print(f"   Primeira letra: {produto[0]}")
print(f"   Última letra: {produto[-1]}")
print(f"   Terceira letra: {produto[2]}")

# Exemplo 3: Fatiamento ✂️
email = "usuario@empresa.com.br"
print(f"\n📧 Email: {email}")
print(f"   Primeiros 7 caracteres: {email[:7]}")
print(f"   Últimos 6 caracteres: {email[-6:]}")
print(f"   Do 8 ao 15: {email[8:15]}")

# Exemplo 4: len() - Comprimento 📊
senha = "Py@123!Secure"
frase = "Python é incrível!"
print(f"\n🔐 Senha: {senha} (tamanho: {len(senha)})")
print(f"💬 Frase: {frase} (tamanho: {len(frase)})")

# Exemplo 5: Iteração sobre string 🔁
palavra = "Python"
print(f"\n🔤 Letras de '{palavra}':")
for i, letra in enumerate(palavra):
    print(f"   Índice {i}: {letra}")

# Exemplo 6: Imutabilidade 🔒
texto_original = "Hello"
print(f"\n🔒 Texto original: {texto_original}")
# texto_original[0] = 'h'  # Isso causaria TypeError!
texto_novo = 'h' + texto_original[1:]  # Criar nova string
print(f"✅ Texto novo: {texto_novo}")
print(f"📊 Original ainda é: {texto_original} (imutável)")
````

## 2️⃣ Concatenação e Repetição

### 📖 Conceito
Concatenação junta strings, enquanto repetição multiplica uma string. São operações fundamentais para construir texto dinâmico.

### 🔧 Operações

**Concatenação (+):**
```python
nome = "João"
sobrenome = "Silva"
nome_completo = nome + " " + sobrenome  # "João Silva"
```

**Repetição (*):**
```python
linha = "-" * 50  # "--------------------------------------------------"
repetido = "Ha" * 3  # "HaHaHa"
```

**F-strings (recomendado):**
```python
nome = "Maria"
idade = 30
mensagem = f"{nome} tem {idade} anos"  # "Maria tem 30 anos"
```

### ⚠️ Importante
* Concatenação com + só funciona entre strings
* Use `str()` para converter outros tipos
* F-strings são mais rápidas e legíveis que +
* Evite concatenação em loops (use join())

````python
# Exemplo 1: Concatenação com + ➕
primeiro_nome = "Carlos"
sobrenome = "Pereira"
nome_completo = primeiro_nome + " " + sobrenome
print(f"👤 Nome completo: {nome_completo}")

# Exemplo 2: Concatenação de múltiplas strings 🔗
titulo = "Dr."
nome = "Roberto"
apelido = "Santos"
especialidade = "Cardiologia"
apresentacao = titulo + " " + nome + " " + apelido + " - " + especialidade
print(f"\n👨‍⚕️ Apresentação: {apresentacao}")

# Exemplo 3: F-strings (moderna e recomendada) 🎉
produto = "Laptop"
preco = 3500.00
quantidade = 3
total = preco * quantidade
mensagem = f"💻 Produto: {produto} | Preço: R$ {preco:.2f} | Qtd: {quantidade} | Total: R$ {total:.2f}"
print(f"\n{mensagem}")

# Exemplo 4: Repetição com * ✖️
separador = "=" * 60
titulo_secao = "RELATÓRIO DE VENDAS"
print(f"\n{separador}")
print(f"{titulo_secao.center(60)}")
print(separador)

# Exemplo 5: Repetição para padrões 🔲
echos = "Eco! " * 5
risadas = "Ha" * 10
print(f"\n🔊 {echos}")
print(f"😂 {risadas}")

# Exemplo 6: join() - Melhor para listas 🧵
palavras = ["Python", "é", "uma", "linguagem", "poderosa"]
frase = " ".join(palavras)
print(f"\n💬 Frase: {frase}")

itens = ["Maçã", "Banana", "Laranja", "Uva"]
lista_formatada = ", ".join(itens)
print(f"🍎 Lista: {lista_formatada}")

# Exemplo 7: Conversão com str() 🔄
idade = 25
altura = 1.75
mensagem_convertida = "Idade: " + str(idade) + " anos, Altura: " + str(altura) + "m"
print(f"\n📊 {mensagem_convertida}")
# Melhor com f-string:
mensagem_fstring = f"Idade: {idade} anos, Altura: {altura}m"
print(f"✅ {mensagem_fstring} (f-string é melhor!)")
````

## 3️⃣ Métodos de Busca

### 📖 Conceito
Métodos de busca localizam substrings dentro de strings maiores. São essenciais para validação, extração e processamento de texto.

### 🔧 Métodos Principais

**find(substring):**
```python
texto = "Python é legal"
texto.find("legal")      # 10 (índice inicial)
texto.find("Java")       # -1 (não encontrado)
```

**index(substring):**
```python
texto.index("legal")     # 10 (índice inicial)
texto.index("Java")      # ValueError! (lança erro)
```

**in / not in:**
```python
"Python" in texto        # True
"Java" not in texto      # True
```

**count(substring):**
```python
texto = "banana"
texto.count("a")         # 3
texto.count("na")        # 2
```

### ⚠️ Diferenças Importantes
* **find()**: Retorna -1 se não encontrar (seguro)
* **index()**: Lança ValueError se não encontrar (erro)
* **in**: Retorna True/False (mais simples para validação)
* **count()**: Conta ocorrências (podem sobrepor)

````python
# Exemplo 1: find() - Busca segura 🔍
email = "contato@empresa.com.br"
posicao_arroba = email.find("@")
posicao_ponto = email.find(".")
print(f"📧 Email: {email}")
print(f"   Posição do @: {posicao_arroba}")
print(f"   Posição do primeiro ponto: {posicao_ponto}")

# Busca que não encontra
posicao_invalida = email.find("#")
print(f"   Posição do #: {posicao_invalida} (não encontrado)")

# Exemplo 2: Extraindo usuário e domínio 📝
email_completo = "usuario.nome@dominio.com.br"
arroba_pos = email_completo.find("@")
if arroba_pos != -1:
    usuario = email_completo[:arroba_pos]
    dominio = email_completo[arroba_pos+1:]
    print(f"\n👤 Usuário: {usuario}")
    print(f"🌐 Domínio: {dominio}")

# Exemplo 3: in / not in - Verificação simples ✅❌
produtos = "carbonato de sódio e óxido de zinco"
print(f"\n🧪 Produtos: {produtos}")
print(f"   Contém 'sódio'? {'sódio' in produtos}")
print(f"   Contém 'magnésio'? {'magnésio' in produtos}")
print(f"   Não contém 'magnésio'? {'magnésio' not in produtos}")

# Exemplo 4: count() - Contando ocorrências 📊
frase = "banana"
palavra_longa = "paralelepipedo"
print(f"\n🍌 '{frase}' tem {frase.count('a')} letras 'a'")
print(f"🏛️ '{palavra_longa}' tem {palavra_longa.count('e')} letras 'e'")
print(f"🔢 '{palavra_longa}' tem {palavra_longa.count('p')} letras 'p'")

# Exemplo 5: count() em texto real 💬
texto_analise = "Python é legal. Python é poderoso. Python é versátil."
qtd_python = texto_analise.count("Python")
qtd_espacos = texto_analise.count(" ")
qtd_pontos = texto_analise.count(".")
print(f"\n📝 Texto: {texto_analise}")
print(f"   'Python' aparece {qtd_python} vezes")
print(f"   {qtd_espacos} espaços")
print(f"   {qtd_pontos} pontos")

# Exemplo 6: Validando formatos 🔐
codigos = ["ABC-123", "XYZ456", "DEF-789", "GHI012"]
print(f"\n📄 Validando códigos:")
for codigo in codigos:
    if "-" in codigo:
        print(f"   ✅ {codigo} - formato válido")
    else:
        print(f"   ❌ {codigo} - formato inválido (falta hífen)")
````

## 4️⃣ Métodos de Transformação

### 📖 Conceito
Métodos de transformação modificam o conteúdo da string, alterando maiúsculas/minúsculas, substituindo partes ou aplicando outras transformações. **Sempre retornam novas strings** (original permanece inalterada).

### 🔧 Métodos Principais

**Alteração de Caixa:**
```python
texto = "Python É Legal"
texto.upper()        # "PYTHON É LEGAL"
texto.lower()        # "python é legal"
texto.capitalize()   # "Python é legal" (só primeira letra)
texto.title()        # "Python É Legal" (cada palavra)
texto.swapcase()     # "pYTHON é lEGAL" (inverte)
```

**Substituição:**
```python
texto.replace("old", "new")     # Substitui todas ocorrências
texto.replace("old", "new", 2)  # Substitui apenas 2 vezes
```

**Divisão e Junção:**
```python
"a,b,c".split(",")              # ['a', 'b', 'c']
"a b c".split()                  # ['a', 'b', 'c'] (espaços)
" | ".join(["a", "b", "c"])      # "a | b | c"
```

### ✅ Casos de Uso
* **upper/lower**: Comparações case-insensitive
* **title**: Títulos, nomes próprios
* **capitalize**: Início de frases
* **replace**: Normalização, limpeza de dados
* **split/join**: Processamento de CSV, listas

````python
# Exemplo 1: upper() e lower() - Maiúsculas/Minúsculas 🔤
nome_completo = "Maria Silva Santos"
email_entrada = "MARIA.SILVA@EMPRESA.COM"
print(f"👤 Nome: {nome_completo}")
print(f"   Maiúsculas: {nome_completo.upper()}")
print(f"   Minúsculas: {nome_completo.lower()}")
print(f"\n📧 Email entrada: {email_entrada}")
print(f"   Normalizado: {email_entrada.lower()}")

# Exemplo 2: capitalize() vs title() 🎯
frase = "python é uma linguagem poderosa"
print(f"\n💬 Original: {frase}")
print(f"   capitalize(): {frase.capitalize()}")
print(f"   title(): {frase.title()}")

# Exemplo 3: title() para nomes próprios 🎭
nomes_brutos = ["JOSE SILVA", "ana paula", "CaRlOs PeReIrA"]
print(f"\n📝 Normalizando nomes:")
for nome in nomes_brutos:
    nome_formatado = nome.title()
    print(f"   {nome:20s} → {nome_formatado}")

# Exemplo 4: replace() - Substituição 🔄
texto_original = "Olá, meu nome é João. João Silva."
texto_substituido = texto_original.replace("João", "Pedro")
print(f"\n📝 Original: {texto_original}")
print(f"✅ Substituído: {texto_substituido}")

# Replace limitado
texto_limitado = texto_original.replace("João", "Pedro", 1)
print(f"🔢 Substituir apenas 1: {texto_limitado}")

# Exemplo 5: replace() para limpeza de dados 🧹
telefone_sujo = "(11) 98765-4321"
telefone_limpo = telefone_sujo.replace("(", "").replace(")", "").replace(" ", "").replace("-", "")
print(f"\n📞 Telefone sujo: {telefone_sujo}")
print(f"✅ Telefone limpo: {telefone_limpo}")

# Exemplo 6: split() - Dividir string ✂️
frase_completa = "Python é incrível e poderoso"
palavras = frase_completa.split()
print(f"\n💬 Frase: {frase_completa}")
print(f"   Palavras: {palavras}")
print(f"   Total de palavras: {len(palavras)}")

# Split com delimitador customizado
csv_linha = "João,Silva,30,Engenheiro"
dados = csv_linha.split(",")
print(f"\n📄 CSV: {csv_linha}")
print(f"   Nome: {dados[0]}, Sobrenome: {dados[1]}, Idade: {dados[2]}, Profissão: {dados[3]}")

# Exemplo 7: join() - Juntar lista em string 🧵
palavras_lista = ["Python", "é", "fantástico"]
frase_juntada = " ".join(palavras_lista)
print(f"\n🔗 Lista: {palavras_lista}")
print(f"   Juntada: {frase_juntada}")

# Join com diferentes separadores
cidades = ["São Paulo", "Rio de Janeiro", "Belo Horizonte"]
print(f"\n🌆 Cidades:")
print(f"   Com vírgula: {', '.join(cidades)}")
print(f"   Com pipe: {' | '.join(cidades)}")
print(f"   Com quebra: {chr(10).join(cidades)}")
````

## 5️⃣ Métodos de Limpeza

### 📖 Conceito
Métodos de limpeza removem caracteres indesejados (geralmente espaços em branco) das extremidades das strings. Essenciais para normalização de dados de entrada.

### 🔧 Métodos Principais

**Remoção de Espaços:**
```python
texto = "  Python  "
texto.strip()     # "Python" (ambos os lados)
texto.lstrip()    # "Python  " (esquerda/left)
texto.rstrip()    # "  Python" (direita/right)
```

**Remoção de Caracteres Específicos:**
```python
texto = "***Python***"
texto.strip("*")  # "Python"
texto.lstrip("*") # "Python***"
texto.rstrip("*") # "***Python"
```

**Múltiplos Caracteres:**
```python
texto = "...---Python---..."
texto.strip(".-")  # "Python" (remove . e -)
```

### ✅ Casos de Uso
* **strip()**: Limpeza de dados de formulários
* **lstrip()**: Remover prefixos
* **rstrip()**: Remover sufixos
* Normalização antes de comparações
* Processamento de arquivos CSV/TXT

### ⚠️ Importante
* Remove apenas das **extremidades**, não do meio
* Remove **qualquer combinação** dos caracteres especificados
* Sem argumento, remove espaços, tabs (\t), quebras (\n)

````python
# Exemplo 1: strip() - Limpeza básica 🧹
nome_usuario = "  Maria Silva  "
email_usuario = "   maria@email.com   "
print("👤 Dados com espaços extras:")
print(f"   Nome: '{nome_usuario}'")
print(f"   Email: '{email_usuario}'")
print(f"\n✅ Após strip():")
print(f"   Nome: '{nome_usuario.strip()}'")
print(f"   Email: '{email_usuario.strip()}'")

# Exemplo 2: lstrip() - Remover à esquerda ⬅️
codigo_produto = "000012345"
print(f"\n📝 Código original: {codigo_produto}")
print(f"✅ Sem zeros à esquerda: {codigo_produto.lstrip('0')}")

# Exemplo 3: rstrip() - Remover à direita ➡️
frase_com_pontos = "Python é legal...."
print(f"\n💬 Frase: {frase_com_pontos}")
print(f"✅ Sem pontos extras: {frase_com_pontos.rstrip('.')}")

# Exemplo 4: Limpando caracteres especiais 🛠️
titulo_markdown = "### Título do Capítulo ###"
titulo_limpo = titulo_markdown.strip("# ")
print(f"\n📝 Markdown: {titulo_markdown}")
print(f"✅ Limpo: {titulo_limpo}")

# Exemplo 5: Limpeza de linhas de arquivo 📄
linhas_arquivo = [
    "  linha 1  \n",
    "linha 2\n",
    "  linha 3\n",
    "linha 4  "
]
print(f"\n📄 Processando linhas de arquivo:")
for i, linha in enumerate(linhas_arquivo, 1):
    linha_limpa = linha.strip()
    print(f"   Linha {i}: '{linha_limpa}'")

# Exemplo 6: Normalização antes de comparação 🔍
senha_digitada = "  Python123  "
senha_correta = "Python123"

print(f"\n🔐 Validação de senha:")
print(f"   Digitada: '{senha_digitada}'")
print(f"   Correta: '{senha_correta}'")

# Sem strip - falha
if senha_digitada == senha_correta:
    print("   ❌ Comparação direta: FALHOU")
else:
    print("   ❌ Comparação direta: FALHOU (espaços extras)")

# Com strip - sucesso
if senha_digitada.strip() == senha_correta:
    print("   ✅ Com strip(): SUCESSO!")

# Exemplo 7: Limpeza de URLs 🌐
url_suja = "///https://exemplo.com/pagina//"
url_limpa = url_suja.strip("/")
print(f"\n🌐 URL suja: {url_suja}")
print(f"✅ URL limpa: {url_limpa}")
````

## 6️⃣ Métodos de Formatação e Alinhamento

### 📖 Conceito
Métodos de formatação ajustam o tamanho e alinhamento de strings, adicionando caracteres de preenchimento. Úteis para formatação de relatórios, tabelas e saídas formatadas.

### 🔧 Métodos Principais

**center(largura, char):**
```python
"Python".center(20, "-")  # "-------Python-------"
```

**ljust(largura, char):**
```python
"Python".ljust(20, "-")   # "Python--------------"
```

**rjust(largura, char):**
```python
"Python".rjust(20, "-")   # "--------------Python"
```

**zfill(largura):**
```python
"42".zfill(5)             # "00042" (zeros à esquerda)
```

### 📊 Parâmetros
* **largura**: Tamanho total da string resultante
* **char**: Caractere de preenchimento (padrão: espaço)
* Se string já for maior, retorna sem alteração

### ✅ Casos de Uso
* **center**: Títulos, cabeçalhos
* **ljust**: Colunas alinhadas à esquerda
* **rjust**: Números, valores monetários
* **zfill**: Códigos, IDs, timestamps

````python
# Exemplo 1: center() - Centralizar texto 🎯
titulo = "RELATÓRIO DE VENDAS"
subtitulo = "Trimestre 1 - 2026"
largura = 60
print(titulo.center(largura, "="))
print(subtitulo.center(largura))
print("=" * largura)

# Exemplo 2: ljust() - Alinhar à esquerda ⬅️
print(f"\n📊 Tabela de produtos:")
print("-" * 50)
produtos = [
    ("Notebook", 3500.00),
    ("Mouse", 45.90),
    ("Teclado", 250.00),
    ("Monitor", 1200.00)
]
for produto, preco in produtos:
    print(f"{produto.ljust(20, '.')} R$ {preco:>8.2f}")
print("-" * 50)

# Exemplo 3: rjust() - Alinhar à direita ➡️
print(f"\n🔢 Números alinhados:")
numeros = [5, 42, 137, 1000, 25689]
for num in numeros:
    print(f"Número: {str(num).rjust(10)} | Quadrado: {str(num**2).rjust(15)}")

# Exemplo 4: zfill() - Preencher com zeros 🔢
print(f"\n🎫 IDs formatados com zfill():")
ids = [1, 42, 789, 3456]
for id_num in ids:
    id_formatado = str(id_num).zfill(6)
    print(f"   ID original: {id_num:4d} → Formatado: {id_formatado}")

# Exemplo 5: Formatação de horários ⏰
horas = [8, 14, 23]
minutos = [5, 30, 59]
segundos = [7, 0, 45]
print(f"\n⏰ Horários formatados:")
for h, m, s in zip(horas, minutos, segundos):
    horario = f"{str(h).zfill(2)}:{str(m).zfill(2)}:{str(s).zfill(2)}"
    print(f"   {horario}")

# Exemplo 6: Tabela formatada completa 📊
print(f"\n📋 RELATÓRIO DE FUNCIONÁRIOS")
print("=" * 70)
print(f"{'Nome'.ljust(20)} {'Cargo'.ljust(15)} {'Salário'.rjust(12)} {'ID'.rjust(8)}")
print("-" * 70)

funcionarios = [
    ("Ana Silva", "Gerente", 8500.00, 123),
    ("Carlos Souza", "Analista", 5200.00, 456),
    ("Maria Santos", "Desenvolvedor", 6800.00, 789)
]

for nome, cargo, salario, id_func in funcionarios:
    id_formatado = str(id_func).zfill(6)
    print(f"{nome.ljust(20)} {cargo.ljust(15)} R$ {salario:>9.2f} {id_formatado.rjust(8)}")
print("=" * 70)

# Exemplo 7: Códigos de barras 📦
print(f"\n📦 Códigos de produtos:")
codigos_curtos = ["ABC", "XYZ", "LMN", "QRS"]
for codigo in codigos_curtos:
    codigo_completo = codigo.ljust(3, "0").zfill(10)
    print(f"   {codigo} → {codigo_completo}")
````

## 7️⃣ Métodos de Validação

### 📖 Conceito
Métodos de validação verificam características da string, retornando True ou False. Essenciais para validação de entrada, parseamento e processamento de dados.

### 🔧 Métodos Principais

**Início e Fim:**
```python
texto = "Python.txt"
texto.startswith("Py")    # True (começa com)
texto.endswith(".txt")    # True (termina com)
```

**Tipo de Caracteres:**
```python
"123".isdigit()           # True (só dígitos)
"abc".isalpha()           # True (só letras)
"abc123".isalnum()        # True (letras ou dígitos)
"   ".isspace()           # True (só espaços)
"PYTHON".isupper()        # True (só maiúsculas)
"python".islower()        # True (só minúsculas)
"Python".istitle()        # True (formato título)
```

**Numeração:**
```python
"123".isnumeric()         # True (aceita mais tipos)
"123".isdecimal()         # True (só 0-9)
"½".isnumeric()          # True
"½".isdecimal()          # False
```

### ⚠️ Diferenças Importantes
* **isdigit()**: 0-9
* **isdecimal()**: 0-9 (mais restrito)
* **isnumeric()**: Aceita frações, romanos, etc.
* Todos retornam False para strings vazias

````python
# Exemplo 1: startswith() e endswith() 🔍
arquivos = ["relatorio.pdf", "dados.csv", "imagem.png", "documento.txt", "planilha.xlsx"]
print("📁 Validando arquivos:")
for arquivo in arquivos:
    if arquivo.endswith(".pdf"):
        print(f"   📝 {arquivo} - Documento PDF")
    elif arquivo.endswith(".csv"):
        print(f"   📊 {arquivo} - Planilha CSV")
    elif arquivo.endswith((".png", ".jpg")):
        print(f"   🖼️ {arquivo} - Imagem")
    else:
        print(f"   📄 {arquivo} - Outro tipo")

# Exemplo 2: isdigit() - Validar números 🔢
entradas = ["12345", "abc123", "999", "42.5", "-100"]
print(f"\n🔢 Validando se são dígitos:")
for entrada in entradas:
    if entrada.isdigit():
        print(f"   ✅ '{entrada}' - Válido (apenas dígitos)")
    else:
        print(f"   ❌ '{entrada}' - Inválido (contém outros caracteres)")

# Exemplo 3: isalpha() - Validar letras 🔤
nomes = ["Maria", "João123", "Ana Paula", "Carlos"]
print(f"\n👤 Validando nomes (apenas letras):")
for nome in nomes:
    if nome.isalpha():
        print(f"   ✅ '{nome}' - Válido")
    else:
        print(f"   ❌ '{nome}' - Inválido (contém números ou espaços)")

# Exemplo 4: isalnum() - Validar alfanuméricos 🔡
usuarios = ["usuario123", "user@123", "admin", "test_user", "user456"]
print(f"\n👤 Validando usuários (alfanuméricos):")
for usuario in usuarios:
    if usuario.isalnum():
        print(f"   ✅ '{usuario}' - Válido")
    else:
        print(f"   ❌ '{usuario}' - Inválido (caracteres especiais)")

# Exemplo 5: isupper(), islower(), istitle() 🔠
textos = ["PYTHON", "python", "Python", "Python É Legal"]
print(f"\n🔤 Analisando capitalização:")
for texto in textos:
    status = []
    if texto.isupper():
        status.append("MAIÚSCULAS")
    if texto.islower():
        status.append("minúsculas")
    if texto.istitle():
        status.append("Título")
    print(f"   '{texto}' - {', '.join(status) if status else 'Misto'}")

# Exemplo 6: Validação de CPF (apenas formato) 💳
print(f"\n💳 Validando formato de CPF:")
cpfs = ["12345678901", "123.456.789-01", "1234567890", "abcdefghijk"]
for cpf in cpfs:
    # Remove pontuação
    cpf_limpo = cpf.replace(".", "").replace("-", "")
    if cpf_limpo.isdigit() and len(cpf_limpo) == 11:
        print(f"   ✅ {cpf} - Formato válido")
    else:
        print(f"   ❌ {cpf} - Formato inválido")

# Exemplo 7: Validação de senha forte 🔐
print(f"\n🔐 Validando senhas:")
senhas = ["senha123", "SENHA123", "Senha123", "Senha@123", "abc"]
for senha in senhas:
    tem_maiuscula = any(c.isupper() for c in senha)
    tem_minuscula = any(c.islower() for c in senha)
    tem_numero = any(c.isdigit() for c in senha)
    tamanho_ok = len(senha) >= 8
    
    if tem_maiuscula and tem_minuscula and tem_numero and tamanho_ok:
        print(f"   ✅ '{senha}' - Senha forte")
    else:
        falhas = []
        if not tem_maiuscula: falhas.append("sem maiúsculas")
        if not tem_minuscula: falhas.append("sem minúsculas")
        if not tem_numero: falhas.append("sem números")
        if not tamanho_ok: falhas.append("muito curta")
        print(f"   ❌ '{senha}' - Fraca ({', '.join(falhas)})")
````

## 8️⃣ Boas Práticas com Strings

### 1️⃣ Use F-strings (Python 3.6+)
* Mais rápidas que concatenação e `.format()`
* Mais legíveis e expressivas
* Suportam expressões dentro de `{}`

### 2️⃣ Use join() para Concatenar Listas
* `" ".join(lista)` é muito mais rápido que loops com `+`
* Especialmente importante em loops grandes
* Python otimiza join() internamente

### 3️⃣ Strings são Imutáveis
* Cada operação cria nova string
* Evite concatenação em loops
* Use listas + join() para construção dinâmica

### 4️⃣ Normalize Antes de Comparar
* Use `.lower()` ou `.upper()` para comparações case-insensitive
* Use `.strip()` para remover espaços
* Combine: `texto.strip().lower()`

### 5️⃣ Prefer find() sobre index()
* `find()` retorna -1 se não encontrar (seguro)
* `index()` lança ValueError (precisa try/except)
* Use `in` para verificar presença (mais simples)

### 6️⃣ Use Métodos Apropriados para Validação
* Não use regex quando métodos nativos funcionam
* `startswith()`, `endswith()` são mais rápidos que regex
* `isdigit()`, `isalpha()` são mais claros que regex

### 7️⃣ Documente Formatos Esperados
* Comente formatos de strings: `# formato: YYYY-MM-DD`
* Use constantes para padrões: `FORMATO_EMAIL = r"..."`
* Valide entradas de usuário sempre

### 8️⃣ Cuidado com Caracteres Especiais
* Use `repr()` para debug de strings
* Atenção a `\n`, `\t`, espaços não visíveis
* Use raw strings (`r"..."`') para regex e paths

### 9️⃣ Encoding Explicitamente
* Sempre especifique encoding ao ler/escrever arquivos
* Use UTF-8 como padrão: `open(file, encoding='utf-8')`
* Evite problemas com caracteres especiais

### 🔟 Performance
* F-strings > `.format()` > concatenação com `+`
* `join()` > loop com `+=`
* Métodos nativos > regex > loops manuais

````python
import time

# ✅ BOM: Usar f-strings (moderno e rápido)
nome = "Maria"
idade = 30
cidade = "São Paulo"
mensagem_fstring = f"{nome}, {idade} anos, mora em {cidade}"
print(f"✅ F-string: {mensagem_fstring}")

# ❌ RUIM: Concatenação com + (lento e ilegível)
mensagem_concat = nome + ", " + str(idade) + " anos, mora em " + cidade
print(f"❌ Concatenação: {mensagem_concat}")

print("\n" + "="*60)

# ✅ BOM: join() para listas (muito mais rápido)
palavras = ["Python", "é", "incrível", "e", "poderoso"]
frase_join = " ".join(palavras)
print(f"✅ Com join(): {frase_join}")

# ❌ RUIM: Concatenação em loop (MUITO lento)
frase_loop = ""
for palavra in palavras:
    frase_loop += palavra + " "
frase_loop = frase_loop.strip()
print(f"❌ Com loop +=: {frase_loop}")

print("\n" + "="*60)

# ✅ BOM: Normalizar antes de comparar
email_usuario = "  MARIA.SILVA@EMAIL.COM  "
email_valido = "maria.silva@email.com"

if email_usuario.strip().lower() == email_valido:
    print("✅ BOM: Emails são iguais (após normalização)")

# ❌ RUIM: Comparar sem normalizar
if email_usuario == email_valido:
    print("❌ RUIM: Falha na comparação")
else:
    print("❌ RUIM: Emails são diferentes (sem normalização)")

print("\n" + "="*60)

# ✅ BOM: find() para busca segura
texto = "Python é legal"
posicao = texto.find("Java")
if posicao == -1:
    print("✅ BOM: 'Java' não encontrado (find retorna -1)")

# ❌ RUIM: index() sem tratamento de erro
try:
    posicao_erro = texto.index("Java")
except ValueError:
    print("❌ RUIM: index() lançou ValueError (precisa try/except)")

# ✅ MELHOR: usar 'in' para verificação
if "Java" not in texto:
    print("✅ MELHOR: Usar 'in' é mais simples e claro")

print("\n" + "="*60)

# ✅ BOM: Métodos nativos para validação
arquivo = "documento.pdf"
if arquivo.endswith(".pdf"):
    print(f"✅ BOM: {arquivo} é PDF (usando endswith)")

# ❌ RUIM: Regex desnecessário
import re
if re.search(r"\.pdf$", arquivo):
    print(f"❌ RUIM: Regex é overkill para isso")

print("\n" + "="*60)

# ✅ BOM: Performance com join vs concatenação
print("⏱️ Teste de performance (10.000 iterações):")

# Método 1: join (rápido)
inicio = time.perf_counter()
lista_temp = []
for i in range(10000):
    lista_temp.append(str(i))
resultado_join = "".join(lista_temp)
fim = time.perf_counter()
tempo_join = (fim - inicio) * 1000
print(f"   join(): {tempo_join:.2f}ms")

# Método 2: concatenação (lento)
inicio = time.perf_counter()
resultado_concat = ""
for i in range(10000):
    resultado_concat += str(i)
fim = time.perf_counter()
tempo_concat = (fim - inicio) * 1000
print(f"   Concatenação: {tempo_concat:.2f}ms")
print(f"   🏆 join() é {tempo_concat/tempo_join:.1f}x mais rápido!")

print("\n" + "="*60)

# ✅ BOM: Usar constantes para padrões
FORMATO_CPF = "XXX.XXX.XXX-XX"
FORMATO_TELEFONE = "(XX) XXXXX-XXXX"
FORMATO_CEP = "XXXXX-XXX"

print("✅ BOM: Usar constantes para padrões:")
print(f"   CPF: {FORMATO_CPF}")
print(f"   Telefone: {FORMATO_TELEFONE}")
print(f"   CEP: {FORMATO_CEP}")

print("\n✅ Siga essas práticas para código mais eficiente, legível e manutenível!")
````
