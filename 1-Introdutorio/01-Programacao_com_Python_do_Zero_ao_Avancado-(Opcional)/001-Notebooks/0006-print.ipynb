# Função print() em Python

Este notebook cobre o uso da função `print()` para exibir informações na saída padrão:

* **Print básico** - Exibir textos e variáveis
* **Concatenação** - Combinar textos e variáveis
* **Parâmetros** - `sep`, `end` e outros
* **Formatação .format()** - Método clássico de formatação
* **F-strings** - Método moderno e recomendado (Python 3.6+)
* **Formatação numérica** - Decimais, inteiros, porcentagens
* **Caracteres especiais** - `\n`, `\t`, aspas, etc.

A função `print()` é fundamental para:
* Debugar código exibindo valores
* Mostrar resultados ao usuário
* Gerar logs e mensagens
* Criar interfaces de linha de comando

## 1. Print Básico

A função `print()` exibe texto e valores na saída padrão.

**Sintaxe:**
```python
print(objeto1, objeto2, ...)
```

**Características:**
* Aceita um ou mais argumentos
* Converte automaticamente para string
* Adiciona quebra de linha (`\n`) ao final por padrão
* Pode imprimir variáveis, literais e expressões

````python
# Print de string literal
print("Olá, mundo!")
print()

# Print de variável
mensagem = "Função print() em Python"
print(mensagem)
print()

# Print de múltiplos valores
print("Aula de Python", "-", "Função print()")
print()

# Print de diferentes tipos
nome = "Maria"
idade = 25
altura = 1.68
print("Nome:", nome)
print("Idade:", idade)
print("Altura:", altura)
print()

# Print de expressões
a = 10
b = 5
print("Soma:", a + b)
print("Multiplicação:", a * b)
````

## 2. Parâmetros do print()

A função `print()` possui parâmetros opcionais para controlar a saída:

### Parâmetro `sep` (separador)
* Define o separador entre múltiplos argumentos
* Padrão: espaço (`" "`)
* **Sintaxe:** `print(valor1, valor2, sep="separador")`

### Parâmetro `end` (final)
* Define o que é impresso ao final
* Padrão: quebra de linha (`"\n"`)
* **Sintaxe:** `print(valor, end="final")`

### Outros parâmetros
* `file`: Especifica onde imprimir (padrão: sys.stdout)
* `flush`: Força a descarga do buffer (padrão: False)

````python
# Parâmetro sep - mudando o separador
print("Python", "Java", "JavaScript")
print("Python", "Java", "JavaScript", sep=" | ")
print("Python", "Java", "JavaScript", sep=" - ")
print("2026", "05", "17", sep="/")
print()

# Parâmetro end - controlando o final
print("Esta linha muda de linha normalmente")
print("Esta linha permanece", end="")
print(" e continua aqui!")
print()

print("Carregando", end="")
print(".", end="")
print(".", end="")
print(".")
print("Concluído!")
print()

# Combinando sep e end
print("A", "B", "C", sep="-", end=" | ")
print("D", "E", "F", sep="-")
print()

# Exemplo prático: formatação de tabela simples
print("Nome", "Idade", "Cidade", sep=" | ")
print("-" * 30)
print("João", "25", "São Paulo", sep=" | ")
print("Maria", "30", "Rio de Janeiro", sep=" | ")
````

## 3. Formatação com .format()

O método `.format()` permite inserir variáveis em strings de forma controlada.

**Sintaxe:**
```python
"Texto com {0} e {1}".format(valor1, valor2)
```

**Características:**
* `{0}`, `{1}`, etc. são marcadores posicionais
* Também aceita marcadores nomeados: `{nome}`, `{idade}`
* Permite formatação avançada de números
* Método clássico (antes do Python 3.6)

**Nota:** F-strings (seção seguinte) são mais modernas e recomendadas!

````python
# Formatação posicional
nome = "Maria"
idade = 30
mensagem = "O nome dela é {0} e ela tem {1} anos".format(nome, idade)
print(mensagem)
print()

# Ordem diferente
print("{1} tem {0} anos".format(idade, nome))
print()

# Reutilizando valores
print("{0} {0} {1}".format("Python", "!"))
print()

# Formatação nomeada
print("{nome} mora em {cidade}".format(nome="João", cidade="São Paulo"))
print()

# Formatação numérica
valor = 1234.5678
print("Valor: {:.2f}".format(valor))  # 2 casas decimais
print("Valor: {:,.2f}".format(valor))  # Com separador de milhar
print()

# Alinhamento
print("{:<10} | {}".format("Nome", "Idade"))  # Esquerda
print("{:>10} | {}".format("João", 25))      # Direita
print("{:^10} | {}".format("Maria", 30))     # Centralizado
````

## 4. F-strings (Formatted String Literals)

F-strings são o método **moderno e recomendado** para formatação de strings (Python 3.6+).

**Sintaxe:**
```python
f"Texto com {variavel} e {expressao}"
```

**Vantagens:**
* Mais legível e concisa
* Permite expressões dentro das chaves
* Melhor performance
* Suporta formatação avançada

**Como usar:**
1. Adicione `f` ou `F` antes das aspas
2. Coloque variáveis/expressões entre `{}`
3. Opcionalmente, adicione especificadores de formato após `:`

````python
# F-string básica
nome = "Fábio"
idade = 35
print(f"Olá, meu nome é {nome} e tenho {idade} anos")
print()

# Múltiplas variáveis
cidade = "São Paulo"
profissao = "Programador"
print(f"{nome} é {profissao} e mora em {cidade}")
print()

# Expressões dentro de f-strings
a = 10
b = 5
print(f"A soma de {a} com {b} é igual a {a + b}")
print(f"A multiplicação de {a} por {b} é {a * b}")
print()

# Chamadas de funções
texto = "python"
print(f"Texto original: {texto}")
print(f"Texto em maiúsculas: {texto.upper()}")
print(f"Tamanho do texto: {len(texto)} caracteres")
print()

# Expressões condicionais (operador ternário)
idade_usuario = 20
status = "Maior de idade" if idade_usuario >= 18 else "Menor de idade"
print(f"Status: {status}")

temperatura = 28
clima = "Quente" if temperatura > 25 else "Frio"
print(f"Clima: {clima}")
````

## 5. Formatação Numérica com F-strings

F-strings permitem formatar números de várias formas:

### Casas Decimais
* **Sintaxe:** `{valor:.Nf}` onde N é o número de casas decimais
* Exemplo: `{3.14159:.2f}` → `3.14`

### Separador de Milhares
* **Sintaxe:** `{valor:,}` ou `{valor:,.2f}`
* Exemplo: `{1234567:,}` → `1,234,567`

### Porcentagem
* **Sintaxe:** `{valor:.N%}`
* Exemplo: `{0.856:.1%}` → `85.6%`

### Notação Científica
* **Sintaxe:** `{valor:.Ne}` ou `{valor:.NE}`
* Exemplo: `{12345:.2e}` → `1.23e+04`

### Alinhamento e Preenchimento
* Esquerda: `{valor:<10}`
* Direita: `{valor:>10}`
* Centralizado: `{valor:^10}`
* Com preenchimento: `{valor:0>5}` → preenche com zeros

````python
# Casas decimais
peso = 73.5
altura = 1.75
imc = peso / (altura ** 2)

print(f"Peso: {peso:.2f} kg")
print(f"Altura: {altura:.2f} m")
print(f"IMC: {imc:.2f}")
print(f"IMC com 4 decimais: {imc:.4f}")
print()

# Separador de milhares
valor_grande = 1234567.89
print(f"Valor: {valor_grande:,.2f}")
print(f"Valor sem decimais: {valor_grande:,.0f}")
print()

# Porcentagens
taxa = 0.0875
conversao = 0.3456

print(f"Taxa de juros: {taxa:.2%}")
print(f"Taxa de conversão: {conversao:.1%}")
print()

# Notação científica
numero_grande = 1234567890
numero_pequeno = 0.000000123

print(f"Grande: {numero_grande:.2e}")
print(f"Pequeno: {numero_pequeno:.2e}")
print()

# Alinhamento (largura 15 caracteres)
print(f"{'Produto':<15} | {'Preço':>10}")
print("-" * 28)
print(f"{'Notebook':<15} | {2500.00:>10.2f}")
print(f"{'Mouse':<15} | {45.90:>10.2f}")
print(f"{'Teclado':<15} | {150.00:>10.2f}")
print()

# Preenchimento com zeros
numero = 42
print(f"Número com zeros: {numero:05d}")
print(f"Número com zeros: {numero:010d}")
````

## 6. Caracteres Especiais (Escape Sequences)

Caracteres especiais permitem formatar a saída de formas específicas:

| Sequência | Descrição | Exemplo |
|-----------|-------------|----------|
| `\n` | Nova linha (line feed) | `"Linha 1\nLinha 2"` |
| `\t` | Tabulação horizontal | `"Nome\tIdade"` |
| `\\` | Barra invertida literal | `"C:\\pasta"` |
| `\'` | Aspas simples | `"\'Texto\'"` |
| `\"` | Aspas duplas | `"\"Texto\""` |
| `\r` | Retorno de carro | `"Texto\rNovo"` |
| `\b` | Backspace | `"Texto\b!"` |

**Raw Strings:**
* Use `r"texto"` para ignorar escape sequences
* Útil para caminhos Windows e regex
* Exemplo: `r"C:\novo\teste"` (barra é literal)

````python
# Nova linha (\n)
print("Primeira linha\nSegunda linha\nTerceira linha")
print()

# Tabulação (\t)
nome = "Bob"
idade = 32
print(f"Nome: {nome}\tIdade: {idade}")
print(f"João\t25\tSão Paulo")
print(f"Maria\t30\tRio de Janeiro")
print()

# Barra invertida literal (\\)
caminho = "C:\\Users\\Documents\\arquivo.txt"
print(f"Caminho: {caminho}")
print()

# Raw string (ignora escape sequences)
caminho_raw = r"C:\novo\teste\arquivo.txt"
print(f"Caminho raw: {caminho_raw}")
print()

# Aspas dentro de strings
print("Ele disse: \"Python é incrível!\"")
print('Ela respondeu: \'Concordo totalmente!\'')
print()

# Usando f-strings com aspas
valor = 125.58
print(f"O valor é: '{valor:.2f}'")
print(f'O valor é: "{valor:.2f}"')
print()

# Combinando múltiplos caracteres especiais
print("Nome\tIdade\tCidade")
print("====\t=====\t======")
print("João\t25\tSão Paulo")
print("Maria\t30\tRio\nPedro\t28\tBrasília")
print()

# String multilinhas com triple quotes
mensagem = """Esta é uma mensagem
que ocupa múltiplas linhas
sem precisar usar \\n"""
print(mensagem)
````

## 7. Exemplo Consolidado: Relatório de Vendas

Vamos criar um relatório completo usando várias técnicas de print() e f-strings.

````python
# Dados do relatório
empresa = "Tech Solutions Ltda"
mes = "Maio"
ano = 2026

# Dados de vendas
produto1 = "Notebook"
quant1 = 15
preco1 = 3500.00
total1 = quant1 * preco1

produto2 = "Mouse"
quant2 = 50
preco2 = 45.90
total2 = quant2 * preco2

produto3 = "Teclado"
quant3 = 30
preco3 = 150.00
total3 = quant3 * preco3

total_geral = total1 + total2 + total3
taxa_imposto = 0.15
imposto = total_geral * taxa_imposto
total_liquido = total_geral - imposto

# Cabeçalho do relatório
print("=" * 70)
print(f"{empresa:^70}")
print(f"{'RELATÓRIO DE VENDAS':^70}")
print(f"{f'{mes}/{ano}':^70}")
print("=" * 70)
print()

# Tabela de produtos
print(f"{'Produto':<20} | {'Qtd':>6} | {'Preço Unit.':>12} | {'Total':>15}")
print("-" * 70)
print(f"{produto1:<20} | {quant1:>6} | R$ {preco1:>9,.2f} | R$ {total1:>12,.2f}")
print(f"{produto2:<20} | {quant2:>6} | R$ {preco2:>9,.2f} | R$ {total2:>12,.2f}")
print(f"{produto3:<20} | {quant3:>6} | R$ {preco3:>9,.2f} | R$ {total3:>12,.2f}")
print("-" * 70)
print()

# Resumo financeiro
print("RESUMO FINANCEIRO")
print("=" * 70)
print(f"Total Bruto:\t\t\t\t\tR$ {total_geral:>12,.2f}")
print(f"Impostos ({taxa_imposto:.1%}):\t\t\t\t\tR$ {imposto:>12,.2f}")
print(f"Total Líquido:\t\t\t\t\tR$ {total_liquido:>12,.2f}")
print("=" * 70)
print()

# Estatísticas
total_itens = quant1 + quant2 + quant3
ticket_medio = total_geral / total_itens

print("ESTATÍSTICAS")
print("-" * 70)
print(f"Total de itens vendidos: {total_itens}")
print(f"Ticket médio por item: R$ {ticket_medio:.2f}")
print(f"Produto mais caro: {produto1} (R$ {preco1:,.2f})")
print(f"Maior faturamento: {produto1} (R$ {total1:,.2f})")
print()

# Rodapé
print("=" * 70)
print(f"Relatório gerado em: {ano}-05-17")
print("=" * 70)
````

## Boas Práticas com print()

### 1. Prefira F-strings
```python
# ❌ EVITE - Concatenação com +
nome = "João"
idade = 25
print("Nome: " + nome + ", Idade: " + str(idade))

# ❌ ANTIGO - Método .format()
print("Nome: {}, Idade: {}".format(nome, idade))

# ✅ RECOMENDADO - F-strings (Python 3.6+)
print(f"Nome: {nome}, Idade: {idade}")
```

### 2. Use sep e end Apropriadamente
```python
# ✅ BOM - Separador customizado
print("2026", "05", "17", sep="-")  # 2026-05-17

# ✅ BOM - Evitar quebra de linha
print("Carregando", end="...")
print(" Concluído!")
```

### 3. Formate Números Adequadamente
```python
valor = 1234.5678

# ✅ Para moeda
print(f"R$ {valor:,.2f}")  # R$ 1,234.57

# ✅ Para porcentagem
taxa = 0.156
print(f"Taxa: {taxa:.1%}")  # Taxa: 15.6%

# ✅ Para notação científica
grande = 1234567890
print(f"{grande:.2e}")  # 1.23e+09
```

### 4. Use print() para Debug
```python
# ✅ Inclua nome da variável
x = 42
print(f"x = {x}")  # Mais claro que print(x)

# ✅ Use f-strings com = (Python 3.8+)
print(f"{x = }")  # x = 42 (mostra nome e valor)
```

### 5. Evite print() em Produção
* Use `logging` para aplicações profissionais
* `print()` é ótimo para:
  * Scripts simples
  * Aprendizado
  * Debug rápido
  * Notebooks interativos

### 6. Organize Saídas Complexas
```python
# ✅ Use linhas separadoras
print("=" * 50)
print("TÍTULO")
print("=" * 50)

# ✅ Alinhe tabelas
print(f"{'Nome':<20} | {'Idade':>5}")
print("-" * 28)
```

### 7. Evite Prints Excessivos
```python
# ❌ EVITE - Muitos prints seguidos
print("Nome:", nome)
print("Idade:", idade)
print("Cidade:", cidade)

# ✅ PREFIRA - Combine em um print
print(f"Nome: {nome}\nIdade: {idade}\nCidade: {cidade}")

# ✅ OU - Use print com múltiplos argumentos
print(
    f"Nome: {nome}",
    f"Idade: {idade}",
    f"Cidade: {cidade}",
    sep="\n"
)
```

### 8. Caracteres Especiais
```python
# ✅ Use raw strings para caminhos Windows
caminho = r"C:\Users\Documents\arquivo.txt"

# ✅ Use triple quotes para texto multilinha
mensagem = """Linha 1
Linha 2
Linha 3"""
```

### 9. Performance
* `print()` é relativamente lento (I/O)
* Para muitas saídas, considere:
  * Acumular em lista e printar uma vez
  * Usar `sys.stdout.write()` para controle fino
  * Desabilitar prints em produção

### 10. Debugging Avançado
```python
# ✅ Mostre tipo e valor
var = [1, 2, 3]
print(f"{type(var).__name__} = {var}")

# ✅ Use repr() para strings com caracteres especiais
texto = "Linha 1\nLinha 2"
print(f"{texto!r}")  # Mostra \\n literal
```
