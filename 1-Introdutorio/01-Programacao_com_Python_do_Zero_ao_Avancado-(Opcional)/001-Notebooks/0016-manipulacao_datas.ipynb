## 1️⃣ Módulo datetime - Conceitos Básicos

### 📖 Conceito
O módulo `datetime` fornece classes para manipular datas e horários. É a ferramenta principal para trabalhar com tempo em Python, oferecendo objetos imutáveis e com suporte a fusos horários.

### 🔧 Classes Principais
```python
import datetime

datetime.date      # Data (ano, mês, dia)
datetime.time      # Hora (hora, minuto, segundo, microsegundo)
datetime.datetime  # Data + Hora combinados
datetime.timedelta # Diferença entre datas/horas
datetime.timezone  # Fuso horário (UTC offset)
```

### 📊 Constantes Importantes
* **MINYEAR**: Menor ano suportado (1)
* **MAXYEAR**: Maior ano suportado (9999)
* **date.min**: Data mínima (0001-01-01)
* **date.max**: Data máxima (9999-12-31)

### ✅ Características
* **Objetos imutáveis**: Não podem ser alterados após criação
* **Comparabilidade**: Podem ser comparados com <, >, ==
* **Operações aritméticas**: Suportam adição/subtração com timedelta

````python
import datetime
from datetime import date, time, datetime as dt, timedelta

# Exemplo 1: Constantes do módulo 📊
print(f"🔢 Ano mínimo: {datetime.MINYEAR}")
print(f"🔢 Ano máximo: {datetime.MAXYEAR}")

# Exemplo 2: Limites de date 📅
print(f"\n📅 Data mínima: {date.min}")
print(f"📅 Data máxima: {date.max}")

# Exemplo 3: Verificação de tipos 🔍
data_hoje = date.today()
hora_agora = dt.now()
print(f"\n🔍 Tipo de date.today(): {type(data_hoje)}")
print(f"🔍 Tipo de datetime.now(): {tipo_hora_agora)}")

# Exemplo 4: Imutabilidade 🔒
data_original = date(2026, 5, 26)
print(f"\n🔒 Data original: {data_original}")
# data_original.year = 2027  # Isso causaria AttributeError
data_nova = date(2027, data_original.month, data_original.day)
print(f"✅ Nova data criada: {data_nova}")

# Exemplo 5: Comparabilidade ⚖️
data1 = date(2026, 1, 1)
data2 = date(2026, 12, 31)
print(f"\n⚖️ {data1} < {data2}? {data1 < data2}")
print(f"⚖️ {data1} == {data2}? {data1 == data2}")

# Exemplo 6: Operações aritméticas ➕➖
data_inicial = date(2026, 5, 26)
dias_adicionar = timedelta(days=100)
data_futura = data_inicial + dias_adicionar
diferenca = data_futura - data_inicial
print(f"\n📆 Data inicial: {data_inicial}")
print(f"➕ Adicionar {dias_adicionar.days} dias: {data_futura}")
print(f"➖ Diferença: {diferenca.days} dias")
````

## 2️⃣ Criação de Datas e Horas

### 📖 Conceito
Existem várias formas de criar objetos de data e hora: especificando valores manualmente, obtendo o momento atual, ou fazendo parse de strings.

### 🔧 Métodos de Criação

**date (apenas data):**
```python
date(year, month, day)
date.today()              # Data atual
date.fromisoformat("2026-05-26")
```

**datetime (data + hora):**
```python
datetime(year, month, day, hour=0, minute=0, second=0, microsecond=0)
datetime.now()            # Data e hora atuais (local)
datetime.today()          # Sinônimo de now()
datetime.utcnow()         # Data e hora UTC
datetime.fromisoformat("2026-05-26T15:30:00")
```

**time (apenas hora):**
```python
time(hour=0, minute=0, second=0, microsecond=0)
```

### ⚠️ Diferenças Importantes
* **now() vs today()**: Ambos retornam data/hora local (hoje são sinônimos)
* **now() vs utcnow()**: now() usa fuso local, utcnow() usa UTC
* **date vs datetime**: date só tem data, datetime tem data + hora

````python
from datetime import date, datetime, time

# Exemplo 1: Criando datas manualmente 📅
data_natal = date(2026, 12, 25)
data_ano_novo = date(2027, 1, 1)
print(f"🎄 Natal 2026: {data_natal}")
print(f"🎆 Ano Novo 2027: {data_ano_novo}")

# Exemplo 2: Data atual 📆
hoje = date.today()
print(f"\n📆 Hoje é: {hoje}")
print(f"   Ano: {hoje.year} | Mês: {hoje.month} | Dia: {hoje.day}")

# Exemplo 3: Criando datetime completos ⏰
reuniao = datetime(2026, 6, 15, 14, 30, 0)
voo = datetime(2026, 7, 10, 8, 45, 30)
print(f"\n💼 Reunião: {reuniao}")
print(f"✈️ Voo: {voo}")

# Exemplo 4: Momento atual - diferentes formas 🕒
agora_now = datetime.now()
agora_today = datetime.today()
agora_utc = datetime.utcnow()
print(f"\n🕒 datetime.now():    {agora_now}")
print(f"🕒 datetime.today(): {agora_today}")
print(f"🌍 datetime.utcnow(): {agora_utc} (UTC)")

# Exemplo 5: Criando objetos time (apenas hora) ⏱️
abertura_loja = time(9, 0, 0)
fechamento_loja = time(18, 30, 0)
print(f"\n🏪 Horário de funcionamento:")
print(f"   Abertura: {abertura_loja}")
print(f"   Fechamento: {fechamento_loja}")

# Exemplo 6: Parse de strings ISO format 📝
data_iso = date.fromisoformat("2026-12-31")
datetime_iso = datetime.fromisoformat("2026-12-31T23:59:59")
print(f"\n📝 Parse ISO 8601:")
print(f"   Data: {data_iso}")
print(f"   Datetime: {datetime_iso}")
````

## 3️⃣ Propriedades e Atributos

### 📖 Conceito
Objetos `date`, `time` e `datetime` possuem atributos somente-leitura que permitem acessar componentes individuais. Esses atributos são úteis para extração e validação de dados.

### 🔧 Atributos Disponíveis

**date e datetime:**
```python
.year       # Ano (MINYEAR - MAXYEAR)
.month      # Mês (1-12)
.day        # Dia (1-31)
```

**datetime e time:**
```python
.hour       # Hora (0-23)
.minute     # Minuto (0-59)
.second     # Segundo (0-59)
.microsecond # Microsegundo (0-999999)
```

**Métodos úteis:**
```python

.isoweekday() # Dia da semana ISO (1=Segunda, 7=Domingo)
.isocalendar() # Tupla (ano_iso, semana, dia_semana)
.timetuple()  # Tupla time.struct_time
```

### ⚠️ Importante
* Todos os atributos são **somente leitura** (imutáveis)
* `.weekday()` retorna 0-6 (Segunda a Domingo)
* `.isoweekday()` retorna 1-7 (Segunda a Domingo)

````python
from datetime import date, datetime

# Exemplo 1: Acessando componentes de date 📅
data_evento = date(2026, 12, 25)
print(f"📅 Data: {data_evento}")
print(f"   Ano: {data_evento.year}")
print(f"   Mês: {data_evento.month}")
print(f"   Dia: {data_evento.day}")

# Exemplo 2: Acessando componentes de datetime ⏰
horario_consulta = datetime(2026, 6, 15, 14, 30, 45, 123456)
print(f"\n⏰ Data/Hora: {horario_consulta}")
print(f"   Data: {horario_consulta.year}-{horario_consulta.month:02d}-{horario_consulta.day:02d}")
print(f"   Hora: {horario_consulta.hour:02d}:{horario_consulta.minute:02d}:{horario_consulta.second:02d}")
print(f"   Microsegundos: {horario_consulta.microsecond}")

# Exemplo 3: Dia da semana 📆
hoje = date.today()
dia_semana_num = hoje.weekday()
dia_semana_iso = hoje.isoweekday()
dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"]
print(f"\n📆 Hoje ({hoje}):")
print(f"   weekday(): {dia_semana_num} = {dias[dia_semana_num]}")
print(f"   isoweekday(): {dia_semana_iso}")

# Exemplo 4: isocalendar() - Ano, semana e dia ISO 📊
data_qualquer = date(2026, 5, 26)
iso_ano, iso_semana, iso_dia = data_qualquer.isocalendar()
print(f"\n📊 ISO Calendar para {data_qualquer}:")
print(f"   Ano ISO: {iso_ano}")
print(f"   Semana do ano: {iso_semana}")
print(f"   Dia da semana: {iso_dia} ({dias[iso_dia-1]})")

# Exemplo 5: Validando horário comercial 🏪
horario_atual = datetime.now()
hora = horario_atual.hour
dia_semana = horario_atual.weekday()

if 0 <= dia_semana <= 4 and 9 <= hora < 18:
    status = "✅ Aberto"
elif dia_semana == 5 and 9 <= hora < 14:
    status = "✅ Aberto (meio expediente)"
else:
    status = "❌ Fechado"

print(f"\n🏪 Status da loja ({horario_atual.strftime('%A %H:%M')}): {status}")

# Exemplo 6: Extrair data e hora separadamente ✂️
datetime_completo = datetime(2026, 12, 31, 23, 59, 59)
apenas_data = datetime_completo.date()
apenas_hora = datetime_completo.time()
print(f"\n✂️ Separando {datetime_completo}:")
print(f"   Apenas data: {apenas_data} (tipo: {type(apenas_data).__name__})")
print(f"   Apenas hora: {apenas_hora} (tipo: {type(apenas_hora).__name__})")
````

## 4️⃣ Formatação de Datas e Horas

### 📖 Conceito
Formatação converte objetos datetime em strings legíveis (strftime) ou parseia strings em objetos datetime (strptime). Essencial para apresentação e processamento de dados temporais.

### 🔧 Métodos Principais
```python
.strftime(formato)           # datetime → string ("string from time")
datetime.strptime(str, fmt)  # string → datetime ("string parse time")
```

### 📋 Códigos de Formato Comuns
| Código | Descrição | Exemplo |
|--------|-------------|----------|
| **%Y** | Ano com 4 dígitos | 2026 |
| **%y** | Ano com 2 dígitos | 26 |
| **%m** | Mês (01-12) | 05 |
| **%B** | Nome do mês completo | May |
| **%b** | Nome do mês abreviado | May |
| **%d** | Dia do mês (01-31) | 26 |
| **%A** | Nome do dia completo | Monday |
| **%a** | Nome do dia abreviado | Mon |
| **%H** | Hora 24h (00-23) | 14 |
| **%I** | Hora 12h (01-12) | 02 |
| **%M** | Minuto (00-59) | 30 |
| **%S** | Segundo (00-59) | 45 |
| **%p** | AM/PM | PM |
| **%f** | Microsegundo | 123456 |

### 🌍 Formatos Internacionais Comuns
* **ISO 8601**: `%Y-%m-%dT%H:%M:%S` → 2026-05-26T14:30:00
* **Brasil**: `%d/%m/%Y %H:%M:%S` → 26/05/2026 14:30:00
* **EUA**: `%m/%d/%Y %I:%M:%S %p` → 05/26/2026 02:30:00 PM
* **Europa**: `%d.%m.%Y %H:%M` → 26.05.2026 14:30

````python
from datetime import datetime

# Exemplo 1: strftime - Formatação básica 📝
agora = datetime.now()
print("📝 Formatos diferentes da mesma data/hora:")
print(f"   ISO 8601:     {agora.strftime('%Y-%m-%dT%H:%M:%S')}")
print(f"   Brasil:       {agora.strftime('%d/%m/%Y %H:%M:%S')}")
print(f"   EUA:          {agora.strftime('%m/%d/%Y %I:%M:%S %p')}")
print(f"   Europa:       {agora.strftime('%d.%m.%Y %H:%M')}")

# Exemplo 2: Formatos legíveis para usuário 👤
evento = datetime(2026, 12, 25, 19, 30, 0)
print(f"\n👤 Formatos amigáveis:")
print(f"   Completo: {evento.strftime('%A, %d de %B de %Y às %H:%M')}")
print(f"   Médio:    {evento.strftime('%d/%m/%Y %H:%M')}")
print(f"   Curto:    {evento.strftime('%d/%m/%y')}")

# Exemplo 3: strptime - Parse de strings 🔄
str_data_br = "26/05/2026"
str_data_usa = "05-26-2026 02:30 PM"
str_iso = "2026-05-26T14:30:00"

data_br = datetime.strptime(str_data_br, "%d/%m/%Y")
data_usa = datetime.strptime(str_data_usa, "%m-%d-%Y %I:%M %p")
data_iso = datetime.strptime(str_iso, "%Y-%m-%dT%H:%M:%S")

print(f"\n🔄 Parse de strings:")
print(f"   '{str_data_br}' → {data_br}")
print(f"   '{str_data_usa}' → {data_usa}")
print(f"   '{str_iso}' → {data_iso}")

# Exemplo 4: Formatação para relatórios 📊
vendas_data = datetime(2026, 3, 15, 10, 45, 30)
print(f"\n📊 Relatório de Vendas")
print(f"Data: {vendas_data.strftime('%d/%m/%Y')}")
print(f"Hora: {vendas_data.strftime('%H:%M:%S')}")
print(f"Trimestre: Q{(vendas_data.month-1)//3 + 1}")
print(f"Semana do ano: {vendas_data.strftime('%W')}")

# Exemplo 5: Timestamps para arquivos 💾
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
arquivo_backup = f"backup_{timestamp}.sql"
arquivo_log = f"log_{timestamp}.txt"
print(f"\n💾 Nomes de arquivos com timestamp:")
print(f"   {arquivo_backup}")
print(f"   {arquivo_log}")

# Exemplo 6: Tratamento de erros em parse ⚠️
strings_invalidas = [
    ("32/13/2026", "%d/%m/%Y"),  # Dia/mês inválidos
    ("2026-02-30", "%Y-%m-%d"),  # 30 de fevereiro
    ("not-a-date", "%Y-%m-%d")   # Formato errado
]

print(f"\n⚠️ Tratamento de erros:")
for string, formato in strings_invalidas:
    try:
        resultado = datetime.strptime(string, formato)
        print(f"   ✅ '{string}' parseado com sucesso")
    except ValueError as e:
        print(f"   ❌ '{string}' - Erro: {str(e)[:50]}...")
````

## 5️⃣ Operações com Timedelta

### 📖 Conceito
`timedelta` representa uma duração ou diferença entre duas datas/horas. Permite realizar aritmética temporal: adicionar/subtrair dias, horas, minutos, etc.

### 🔧 Criando Timedelta
```python
timedelta(days=0, seconds=0, microseconds=0, milliseconds=0,
          minutes=0, hours=0, weeks=0)
```

### 📊 Atributos de Timedelta
* **.days**: Número de dias
* **.seconds**: Segundos (0-86399, não inclui dias)
* **.microseconds**: Microsegundos (0-999999)
* **.total_seconds()**: Duração total em segundos

### ⚠️ Importante
* Timedelta armazena apenas `days`, `seconds` e `microseconds`
* Horas são convertidas em seconds: `hours=2` → `seconds=7200`
* Para obter horas: `timedelta.seconds // 3600`
* Para obter minutos: `(timedelta.seconds % 3600) // 60`

### ➕ Operações Suportadas
* `date + timedelta` → date
* `datetime + timedelta` → datetime
* `date - date` → timedelta
* `timedelta * int` → timedelta
* `timedelta / int` → timedelta

````python
from datetime import datetime, date, timedelta

# Exemplo 1: Criando timedeltas 🕒
dia = timedelta(days=1)
semana = timedelta(weeks=1)
hora = timedelta(hours=1)
minuto = timedelta(minutes=30)

print("🕒 Durações criadas:")
print(f"   1 dia = {dia}")
print(f"   1 semana = {semana} = {semana.days} dias")
print(f"   1 hora = {hora} = {hora.seconds} segundos")
print(f"   30 minutos = {minuto} = {minuto.seconds} segundos")

# Exemplo 2: Acessando atributos corretamente 📊
duracao = timedelta(days=5, hours=3, minutes=45, seconds=30)
print(f"\n📊 Duração: {duracao}")
print(f"   days: {duracao.days}")
print(f"   seconds: {duracao.seconds}")
print(f"   total_seconds(): {duracao.total_seconds()}")
# Convertendo seconds de volta para componentes
horas_componente = duracao.seconds // 3600
minutos_componente = (duracao.seconds % 3600) // 60
segundos_componente = duracao.seconds % 60
print(f"   Decomposto: {duracao.days}d {horas_componente}h {minutos_componente}m {segundos_componente}s")

# Exemplo 3: Adicionar tempo a datas ➕
hoje = date.today()
daqui_30_dias = hoje + timedelta(days=30)
daqui_90_dias = hoje + timedelta(weeks=13) # Aproximadamente 90 dias
print(f"\n➕ Adicionar tempo:")
print(f"   Hoje: {hoje}")
print(f"   Daqui a 30 dias: {daqui_30_dias}")
print(f"   Daqui a 13 semanas: {daqui_90_dias}")

# Exemplo 4: Subtrair tempo de datas ➖
agora = datetime.now()
ontem = agora - timedelta(days=1)
semana_passada = agora - timedelta(weeks=1)
print(f"\n➖ Subtrair tempo:")
print(f"   Agora: {agora.strftime('%d/%m/%Y %H:%M')}")
print(f"   Ontem: {ontem.strftime('%d/%m/%Y %H:%M')}")
print(f"   Semana passada: {semana_passada.strftime('%d/%m/%Y %H:%M')}")

# Exemplo 5: Calculando diferenças entre datas 📆
nascimento = date(1990, 5, 15)
hoje = date.today()
idade_timedelta = hoje - nascimento
idade_anos = idade_timedelta.days // 365
print(f"\n📆 Cálculo de idade:")
print(f"   Nascimento: {nascimento}")
print(f"   Hoje: {hoje}")
print(f"   Diferença: {idade_timedelta.days} dias")
print(f"   Idade aproximada: {idade_anos} anos")

# Exemplo 6: Operações aritméticas com timedelta 🧮
prazo_base = timedelta(days=30)
prazo_estendido = prazo_base * 2
prazo_reduzido = prazo_base / 2
print(f"\n🧮 Operações aritméticas:")
print(f"   Prazo base: {prazo_base.days} dias")
print(f"   Dobrado: {prazo_estendido.days} dias")
print(f"   Metade: {prazo_reduzido.days} dias")

# Exemplo 7: Prazo de vencimento com timedelta 💳
data_compra = datetime.now()
prazo_pagamento = timedelta(days=42)
data_vencimento = data_compra + prazo_pagamento
print(f"\n💳 Boleto:")
print(f"   Compra: {data_compra.strftime('%d/%m/%Y')}")
print(f"   Vencimento: {data_vencimento.strftime('%d/%m/%Y')} ({prazo_pagamento.days} dias)")
````

## 6️⃣ Módulo time

### 📖 Conceito
O módulo `time` trabalha com timestamps (segundos desde Epoch: 01/01/1970 UTC) e funções de sistema. É mais baixo nível que datetime, ideal para medições de performance e timestamps.

### 🔧 Funções Principais
```python
import time

time.time()           # Timestamp atual (float, segundos desde Epoch)
time.gmtime(secs)     # Timestamp → struct_time UTC
time.localtime(secs)  # Timestamp → struct_time local
time.ctime(secs)      # Timestamp → string legível
time.strftime(fmt, t) # struct_time → string formatada
time.strptime(str, fmt) # String → struct_time
time.sleep(secs)      # Pausa execução por N segundos
```

### 📊 struct_time
Tupla nomeada com 9 campos:
* `tm_year`, `tm_mon`, `tm_mday` (data)
* `tm_hour`, `tm_min`, `tm_sec` (hora)
* `tm_wday` (dia da semana: 0=Segunda)
* `tm_yday` (dia do ano: 1-366)
* `tm_isdst` (horário de verão: -1/0/1)

### ⏱️ Epoch (Unix Timestamp)
* **Início**: 1970-01-01 00:00:00 UTC
* **Formato**: Segundos (float) desde o início
* **Exemplo**: 1779752025.5 ≈ 2026-05-26

````python
import time
from datetime import datetime

# Exemplo 1: Timestamp atual 🕒
timestamp_agora = time.time()
print(f"🕒 Timestamp atual: {timestamp_agora}")
print(f"   Equivale a: {datetime.fromtimestamp(timestamp_agora)}")

# Exemplo 2: Epoch (timestamp 0) 🌍
epoch_utc = time.gmtime(0)
epoch_str = time.strftime("%Y-%m-%d %H:%M:%S UTC", epoch_utc)
print(f"\n🌍 Epoch (timestamp 0):")
print(f"   struct_time: {epoch_utc}")
print(f"   Formatado: {epoch_str}")

# Exemplo 3: gmtime vs localtime ⏰
timestamp = time.time()
utc_time = time.gmtime(timestamp)
local_time = time.localtime(timestamp)
print(f"\n⏰ Timestamp {timestamp}:")
print(f"   UTC:   {time.strftime('%Y-%m-%d %H:%M:%S', utc_time)}")
print(f"   Local: {time.strftime('%Y-%m-%d %H:%M:%S', local_time)}")

# Exemplo 4: ctime - String legível rápida 📝
agora_ctime = time.ctime()
ontem_timestamp = time.time() - 86400
ontem_ctime = time.ctime(ontem_timestamp)
print(f"\n📝 Formato ctime():")
print(f"   Agora: {agora_ctime}")
print(f"   Ontem: {ontem_ctime}")

# Exemplo 5: Formatação com strftime 🎨
struct_agora = time.localtime()
formatos = [
    ("%d/%m/%Y %H:%M:%S", "Brasil"),
    ("%Y-%m-%d %H:%M:%S", "ISO 8601"),
    ("%A, %B %d, %Y", "Completo"),
    ("%H:%M:%S", "Apenas hora")
]
print(f"\n🎨 Formatações com time.strftime():")
for formato, descricao in formatos:
    resultado = time.strftime(formato, struct_agora)
    print(f"   {descricao:15s}: {resultado}")

# Exemplo 6: Medição de performance ⏱️
print(f"\n⏱️ Medindo performance de operação:")
inicio = time.time()
# Simular processamento
total = sum(range(1000000))
fim = time.time()
duracao = fim - inicio
print(f"   Operação: soma de 1 a 1.000.000")
print(f"   Resultado: {total}")
print(f"   Tempo: {duracao:.6f} segundos ({duracao*1000:.2f}ms)")

# Exemplo 7: Conversão datetime ↔ timestamp 🔄
dt = datetime(2026, 12, 25, 18, 0, 0)
timestamp_natal = dt.timestamp()
dt_recuperado = datetime.fromtimestamp(timestamp_natal)
print(f"\n🔄 Conversão datetime ↔ timestamp:")
print(f"   datetime: {dt}")
print(f"   timestamp: {timestamp_natal}")
print(f"   de volta: {dt_recuperado}")
````

## 7️⃣ Boas Práticas com Datas e Horas

### 1️⃣ Sempre Use datetime, Não Strings
* Armazene datas como objetos `datetime`, não como strings
* Converta para string apenas para apresentação
* Objetos permitem comparações, operações e validações

### 2️⃣ Use ISO 8601 para Armazenamento
* Formato: `YYYY-MM-DD` ou `YYYY-MM-DDTHH:MM:SS`
* Ordenável lexicograficamente
* Padrão internacional reconhecido
* Use `.isoformat()` para gerar

### 3️⃣ Cuidado com Fusos Horários
* Use `datetime.now(tz=...)` com timezone aware
* Armazene sempre em UTC, converta na apresentação
* Biblioteca `pytz` ou `zoneinfo` (Python 3.9+) para timezones
* Objetos naive (sem timezone) podem causar bugs sutis

### 4️⃣ Valide Entradas de Usuário
* Use `try/except` com `strptime()` para parse
* Valide intervalos lógicos (data futura, idade razoável)
* Não assuma formatos - sempre especifique

### 5️⃣ Use timedelta para Operações
* Não manipule dias/meses manualmente
* `timedelta` trata anos bissextos, fins de mês, etc.
* Para operações complexas, use `dateutil.relativedelta`

### 6️⃣ Prefer datetime.now() a date.today()
* `datetime.now()` dá mais informações (hora)
* Pode extrair `.date()` se precisar apenas data
* Evita conversões desnecessárias

### 7️⃣ Timestamps para Logs e Métricas
* Use `time.time()` para medições de performance
* Use `datetime.now()` para logs legíveis
* Timestamps são ideais para ordenação e cálculos

### 8️⃣ Documente Formatos Esperados
* Comente o formato em variáveis: `data_br # formato: DD/MM/YYYY`
* Use constantes para formatos: `FORMATO_BR = "%d/%m/%Y"`
* Evita confusão entre DD/MM vs MM/DD

### 9️⃣ Teste Casos Limite
* Início/fim de mês, ano
* Anos bissextos (29 de fevereiro)
* Mudanças de horário de verão
* Timestamps próximos ao Epoch
