# 🎯 Sets (Conjuntos) em Python

## 📋 Visão Geral
Este notebook explora sets, uma estrutura de dados fundamental que armazena coleções **não ordenadas** de elementos **únicos**. Sets são otimizados para operações de conjunto e testes de pertinência.

## 🎯 Objetivos de Aprendizagem
* Criar e manipular sets com elementos únicos
* Adicionar e remover elementos de forma eficiente
* Realizar operações de conjunto: união, interseção, diferença
* Usar métodos de comparação: subconjunto, superconjunto, disjuntos
* Trabalhar com frozensets (conjuntos imutáveis)
* Criar sets com set comprehension
* Aplicar sets em problemas práticos: remoção de duplicatas, filtragem, análise
* Seguir boas práticas em manipulação de conjuntos

## 1️⃣ Criação e Conceitos Básicos

### 📖 Conceito
Sets são coleções **não ordenadas** de elementos **únicos**. Não permitem duplicatas e não mantêm ordem de inserção. São otimizados para testes de pertinência (verificação de existência) e operações de conjunto.

### 🔧 Criação de Sets
```python
# Set vazio (CUIDADO: {} cria dicionário!)
vazio = set()  # ✅ Correto

# Com elementos
frutas = {"maçã", "banana", "laranja"}

# Com set()
frutas2 = set(["maçã", "banana", "laranja"])

# De string (separa caracteres)
letras = set("python")  # {'p', 'y', 't', 'h', 'o', 'n'}
```

### 📊 Características
* **Elementos únicos**: Duplicatas são automaticamente removidas
* **Não ordenados**: Ordem não é garantida (sem índices)
* **Mutáveis**: Podemos adicionar/remover elementos
* **Elementos imutáveis**: Só aceita tipos hasháveis (strings, números, tuplas)
* **Acesso rápido**: O(1) para teste de pertinência

### ✅ Quando Usar
* Remover duplicatas de coleções
* Testar existência de elementos rapidamente
* Operações matemáticas de conjuntos
* Verificar elementos comuns entre grupos
* Filtrar valores únicos

### ⚠️ Limitações
* Não suporta indexação (sem `set[0]`)
* Não pode conter listas ou dicionários
* Ordem não é preservada

````python
# Exemplo 1: Criando sets de diferentes formas 🎓
planetas_anoes = {"Plutão", "Ceres", "Eris", "Haumea", "Makemake"}
print(f"🪐 Planetas anões: {planetas_anoes}")
print(f"🔢 Total: {len(planetas_anoes)}")

# De lista
linguagens = set(["Python", "Java", "JavaScript", "Python"])
print(f"\n💻 Linguagens (duplicata removida): {linguagens}")

# De string
palavra = set("python")
print(f"\n🔤 Letras em 'python': {palavra}")

# Set vazio (IMPORTANTE!)
vazio_errado = {}  # Isto cria dicionário!
vazio_correto = set()  # ✅ Isto cria set!
print(f"\n🔍 Tipo de {{}}: {type(vazio_errado)}")
print(f"✅ Tipo de set(): {type(vazio_correto)}")

# Exemplo 2: Elementos únicos (remoção automática de duplicatas) 🎯
notas = [85, 90, 85, 78, 90, 92, 85, 88]
notas_unicas = set(notas)
print(f"\n📊 Lista original: {notas}")
print(f"🎯 Notas únicas: {sorted(notas_unicas)}")
print(f"🔢 De {len(notas)} notas para {len(notas_unicas)} únicas")

# Exemplo 3: Teste de pertinência (in) - Muito rápido! ⚡
metais_preciosos = {"ouro", "prata", "platina", "paládio", "rôdio"}
print(f"\n💎 Metais preciosos: {metais_preciosos}")
print(f"\n✅ 'ouro' é precioso? {'ouro' in metais_preciosos}")
print(f"❌ 'ferro' é precioso? {'ferro' in metais_preciosos}")
print(f"✅ 'cobre' NÃO é precioso? {'cobre' not in metais_preciosos}")

# Exemplo 4: Iteração (ordem não garantida) 🔁
cores_primarias = {"vermelho", "azul", "amarelo"}
print(f"\n🎨 Cores primárias:")
for cor in cores_primarias:
    print(f"   • {cor.capitalize()}")

# Exemplo 5: Tipos de elementos permitidos ✅❌
# ✅ Elementos hasháveis (imutáveis)
valido = {1, 2.5, "texto", (1, 2), True}
print(f"\n✅ Set válido: {valido}")
print(f"   Tipos: int, float, str, tuple, bool")

# ❌ Não pode conter listas (mutáveis)
try:
    invalido = {[1, 2], [3, 4]}  # Erro!
except TypeError as e:
    print(f"\n❌ Erro ao criar set com listas: {e}")

# Exemplo 6: Contagem de elementos únicos 📊
visitantes = ["Ana", "Bruno", "Carla", "Ana", "Bruno", "Daniel", "Ana"]
visitantes_unicos = set(visitantes)

print(f"\n👥 Visitas totais: {len(visitantes)}")
print(f"👤 Visitantes únicos: {len(visitantes_unicos)}")
print(f"📋 Lista de únicos: {sorted(visitantes_unicos)}")

# Exemplo 7: Conversão entre tipos 🔄
lista = [1, 2, 2, 3, 3, 3]
set_unico = set(lista)  # Remove duplicatas
lista_sem_dup = list(set_unico)  # Volta para lista
lista_ordenada = sorted(set_unico)  # Lista ordenada sem duplicatas

print(f"\n🔄 Conversões:")
print(f"   Original: {lista}")
print(f"   Como set: {set_unico}")
print(f"   Lista sem dup: {lista_sem_dup}")
print(f"   Lista ordenada: {lista_ordenada}")
````

## 2️⃣ Adição e Remoção de Elementos

### 📖 Conceito
Sets são mutáveis, permitindo adicionar e remover elementos dinamicamente. Python oferece vários métodos com comportamentos diferentes.

### 🔧 Métodos de Adição

**add()** - Adiciona um elemento:
```python
set.add(elemento)  # Adiciona se não existir
```

**update()** - Adiciona múltiplos elementos:
```python
set.update([1, 2, 3])        # De lista
set.update({4, 5}, {6, 7})   # Múltiplos iteráveis
set.update("abc")            # De string
```

### 🔧 Métodos de Remoção

**remove()** - Remove elemento (erro se não existir):
```python
set.remove(elemento)  # Lança KeyError se não encontrar
```

**discard()** - Remove elemento (silencioso se não existir):
```python
set.discard(elemento)  # Não lança erro
```

**pop()** - Remove e retorna elemento aleatório:
```python
elemento = set.pop()  # Remove elemento arbitrário
```

**clear()** - Remove todos os elementos:
```python
set.clear()  # Set fica vazio
```

### ⚠️ Diferenças Importantes
* **add() vs update()**: add() adiciona 1 elemento, update() adiciona vários
* **remove() vs discard()**: remove() lança erro, discard() é silencioso
* **pop()**: Remove elemento aleatório (não é previsível!)

### ✅ Quando Usar Cada Método
* **add()**: Adicionar um único elemento
* **update()**: Adicionar múltiplos elementos de uma vez
* **discard()**: Remover se existir (mais seguro)
* **remove()**: Remover com validação (erro = elemento esperado)
* **pop()**: Processar elementos até esvaziar o set

````python
# Exemplo 1: add() - Adicionando elementos individuais ➕
estacoes = {"Verão", "Inverno"}
print(f"🌱 Estações iniciais: {estacoes}")

estacoes.add("Outono")
estacoes.add("Primavera")
print(f"➕ Após add(): {estacoes}")

# Tentar adicionar duplicata (não tem efeito)
estacoes.add("Verão")
print(f"🚫 Adicionar duplicata (sem efeito): {estacoes}")

# Exemplo 2: update() - Adicionando múltiplos elementos 🔄
vogais = {"a", "e"}
print(f"\n🔤 Vogais iniciais: {vogais}")

vogais.update(["i", "o", "u"])
print(f"🔄 Após update(): {sorted(vogais)}")

# update() com múltiplos iteráveis
numeros = {1, 2}
numeros.update([3, 4], {5, 6}, (7, 8))
print(f"\n🔢 Números após múltiplas updates: {sorted(numeros)}")

# Exemplo 3: remove() - Remover com validação ❌
planetas = {"Mercúrio", "Vênus", "Terra", "Marte", "Júpiter"}
print(f"\n🪐 Planetas: {planetas}")

planetas.remove("Marte")
print(f"❌ Removeu Marte: {planetas}")

# Tentar remover elemento inexistente lança erro
try:
    planetas.remove("Plutão")  # Erro!
except KeyError as e:
    print(f"⚠️ Erro ao remover 'Plutão': {e}")

# Exemplo 4: discard() - Remover com segurança 🚫
metais = {"ferro", "cobre", "alumínio", "zinco"}
print(f"\n🧫 Metais: {metais}")

metais.discard("zinco")
print(f"🚫 Removeu zinco: {metais}")

# discard() não lança erro se não existir
metais.discard("ouro")  # Silencioso, sem erro
print(f"✅ discard('ouro') - sem erro: {metais}")

# Exemplo 5: pop() - Remover elemento aleatório 🎲
cores = {"vermelho", "azul", "verde", "amarelo", "roxo"}
print(f"\n🎨 Cores originais: {cores}")

removida1 = cores.pop()
print(f"🎲 pop() removeu: {removida1}")
print(f"📊 Cores restantes: {cores}")

removida2 = cores.pop()
print(f"🎲 pop() removeu: {removida2}")
print(f"📊 Cores restantes: {cores}")

# Exemplo 6: clear() - Limpar tudo 🧹
temp_set = {1, 2, 3, 4, 5}
print(f"\n📊 Antes: {temp_set}")
temp_set.clear()
print(f"🧹 Após clear(): {temp_set}")
print(f"✅ Set vazio? {len(temp_set) == 0}")

# Exemplo 7: Padrão de processamento com pop() 🔄
tarefas = {"Estudar", "Exercício", "Leitura", "Código"}
print(f"\n✅ Tarefas pendentes: {tarefas}")

completadas = []
while tarefas:
    tarefa = tarefas.pop()
    completadas.append(tarefa)
    print(f"   ✔️ Completou: {tarefa}")

print(f"\n🎉 Todas as tarefas completadas: {completadas}")
print(f"📊 Tarefas restantes: {tarefas}")
````

## 3️⃣ Operações de Conjunto

### 📖 Conceito
Sets suportam operações matemáticas de conjuntos: união, interseção, diferença e diferença simétrica. Podem ser realizadas com operadores ou métodos.

### 🔧 Operações Principais

**União** - Todos os elementos de ambos:
```python
A | B           # Operador
A.union(B)      # Método
# {1, 2, 3, 4, 5} se A={1,2,3} e B={3,4,5}
```

**Interseção** - Elementos comuns:
```python
A & B                # Operador
A.intersection(B)    # Método
# {3} se A={1,2,3} e B={3,4,5}
```

**Diferença** - Em A mas não em B:
```python
A - B              # Operador
A.difference(B)    # Método
# {1, 2} se A={1,2,3} e B={3,4,5}
```

**Diferença Simétrica** - Em A ou B, mas não em ambos:
```python
A ^ B                        # Operador
A.symmetric_difference(B)    # Método
# {1, 2, 4, 5} se A={1,2,3} e B={3,4,5}
```

### 📊 Métodos de Atualização (In-place)
* `A.update(B)` = `A |= B` (adiciona elementos de B em A)
* `A.intersection_update(B)` = `A &= B` (mantém só comuns)
* `A.difference_update(B)` = `A -= B` (remove elementos de B)
* `A.symmetric_difference_update(B)` = `A ^= B`

### ✅ Casos de Uso
* **União**: Combinar listas de convidados, tags, categorias
* **Interseção**: Encontrar interesses comuns, clientes em comum
* **Diferença**: Remover itens processados, filtrar exclusões
* **Diferença Simétrica**: Detectar mudanças, elementos exclusivos

````python
# Exemplo 1: União (|) - Combinar conjuntos 🔗
grupo_a = {"Ana", "Bruno", "Carla"}
grupo_b = {"Carla", "Daniel", "Elena"}

uniao = grupo_a | grupo_b
print(f"👥 Grupo A: {grupo_a}")
print(f"👥 Grupo B: {grupo_b}")
print(f"🔗 União (A | B): {uniao}")
print(f"✅ Equivalente: {grupo_a.union(grupo_b)}")

# Exemplo 2: Interseção (&) - Elementos comuns 🎯
interesses_maria = {"leitura", "cinema", "viagem", "fotografia"}
interesses_joao = {"cinema", "games", "viagem", "esportes"}

comuns = interesses_maria & interesses_joao
print(f"\n📚 Maria: {interesses_maria}")
print(f"🎮 João: {interesses_joao}")
print(f"🎯 Interesses comuns (M & J): {comuns}")
print(f"✅ Equivalente: {interesses_maria.intersection(interesses_joao)}")

# Exemplo 3: Diferença (-) - Em A mas não em B ➡️
produtos_loja1 = {"notebook", "mouse", "teclado", "monitor", "webcam"}
produtos_loja2 = {"mouse", "teclado", "headset", "mousepad"}

exclusivos_loja1 = produtos_loja1 - produtos_loja2
exclusivos_loja2 = produtos_loja2 - produtos_loja1

print(f"\n🏪 Loja 1: {produtos_loja1}")
print(f"🏪 Loja 2: {produtos_loja2}")
print(f"➡️ Exclusivos da Loja 1: {exclusivos_loja1}")
print(f"➡️ Exclusivos da Loja 2: {exclusivos_loja2}")

# Exemplo 4: Diferença Simétrica (^) - Elementos exclusivos 🔄
alunos_manha = {"Ana", "Bruno", "Carla", "Daniel"}
alunos_tarde = {"Carla", "Daniel", "Elena", "Fernando"}

exclusivos = alunos_manha ^ alunos_tarde
print(f"\n☀️ Manhã: {alunos_manha}")
print(f"🌙 Tarde: {alunos_tarde}")
print(f"🔄 Só em um turno: {exclusivos}")
print(f"✅ Equivalente: {alunos_manha.symmetric_difference(alunos_tarde)}")

# Exemplo 5: Operações múltiplas (encadeadas) 🔗
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
C = {5, 6, 7, 8}

# União de três conjuntos
uniao_abc = A | B | C
print(f"\n🔢 A: {A}")
print(f"🔢 B: {B}")
print(f"🔢 C: {C}")
print(f"🔗 A | B | C: {uniao_abc}")

# Interseção de três conjuntos
intersecao_abc = A & B & C
print(f"🎯 A & B & C: {intersecao_abc}")

# Exemplo 6: Métodos de atualização (in-place) ✏️
tags_post1 = {"python", "tutorial", "iniciante"}
tags_post2 = {"tutorial", "avançado", "projetos"}

print(f"\n🏷️ Post 1 original: {tags_post1}")

# Adicionar tags do post 2 ao post 1
tags_post1.update(tags_post2)  # ou tags_post1 |= tags_post2
print(f"✏️ Post 1 após update: {tags_post1}")

# difference_update
palavras = {"o", "a", "de", "para", "python", "dados"}
stopwords = {"o", "a", "de", "para"}

print(f"\n📝 Palavras: {palavras}")
palavras.difference_update(stopwords)  # Remove stopwords
print(f"✏️ Sem stopwords: {palavras}")

# Exemplo 7: Caso prático - Análise de clientes 📈
clientes_janeiro = {"C001", "C002", "C003", "C004", "C005"}
clientes_fevereiro = {"C003", "C004", "C005", "C006", "C007"}

print(f"\n📅 Clientes em Janeiro: {clientes_janeiro}")
print(f"📅 Clientes em Fevereiro: {clientes_fevereiro}")

recorrentes = clientes_janeiro & clientes_fevereiro
novo_em_fev = clientes_fevereiro - clientes_janeiro
perdidos = clientes_janeiro - clientes_fevereiro
total_atendidos = clientes_janeiro | clientes_fevereiro

print(f"\n📈 Análise:")
print(f"   🔄 Recorrentes (ambos meses): {recorrentes}")
print(f"   🎉 Novos em Fevereiro: {novo_em_fev}")
print(f"   🚨 Perdidos (só Janeiro): {perdidos}")
print(f"   📆 Total de clientes únicos: {len(total_atendidos)}")
````

## 4️⃣ Métodos de Comparação

### 📖 Conceito
Sets podem ser comparados para verificar relações de subconjunto, superconjunto e disjunção. Essas operações são fundamentais em lógica de conjuntos.

### 🔧 Métodos de Comparação

**Igualdade:**
```python
A == B    # True se têm os mesmos elementos
A != B    # True se são diferentes
```

**Subconjunto** (A está contido em B):
```python
A <= B              # Subconjunto (pode ser igual)
A < B               # Subconjunto próprio (menor que)
A.issubset(B)       # Método equivalente
```

**Superconjunto** (A contém B):
```python
A >= B              # Superconjunto (pode ser igual)
A > B               # Superconjunto próprio (maior que)
A.issuperset(B)     # Método equivalente
```

**Disjuntos** (sem elementos comuns):
```python
A.isdisjoint(B)     # True se não há interseção
```

### 📊 Relações
* **A ⊆ B** (A <= B): Todo elemento de A está em B
* **A ⊂ B** (A < B): A ⊆ B e A ≠ B
* **A ⊇ B** (A >= B): Todo elemento de B está em A
* **A ⊃ B** (A > B): A ⊇ B e A ≠ B
* **A ∩ B = ∅**: A e B são disjuntos

### ✅ Casos de Uso
* Verificar permissões (usuário tem todas permissões necessárias?)
* Validar requisitos (projeto atende todos requisitos?)
* Detectar conflitos (horários não se sobrepõem?)
* Hierarquias (categoria contém subcategoria?)

````python
# Exemplo 1: Igualdade de sets ⚖️
set1 = {1, 2, 3, 4}
set2 = {4, 3, 2, 1}  # Ordem não importa
set3 = {1, 2, 3}

print(f"🔢 Set 1: {set1}")
print(f"🔢 Set 2: {set2}")
print(f"🔢 Set 3: {set3}")
print(f"\n✅ set1 == set2? {set1 == set2}")
print(f"❌ set1 == set3? {set1 == set3}")
print(f"✅ set1 != set3? {set1 != set3}")

# Exemplo 2: Subconjunto (issubset / <=) 🔽
vogais = {"a", "e", "i", "o", "u"}
vogais_parcial = {"a", "e", "i"}
vogais_com_extra = {"a", "e", "i", "x"}

print(f"\n🔤 Vogais completas: {vogais}")
print(f"🔤 Vogais parciais: {vogais_parcial}")
print(f"🔤 Com extra: {vogais_com_extra}")

print(f"\n✅ parcial ⊆ completas? {vogais_parcial <= vogais}")
print(f"✅ parcial ⊂ completas? {vogais_parcial < vogais}")
print(f"❌ com_extra ⊆ completas? {vogais_com_extra <= vogais}")
print(f"\n✅ Método: {vogais_parcial.issubset(vogais)}")

# Exemplo 3: Superconjunto (issuperset / >=) 🔼
permissoes_admin = {"ler", "escrever", "excluir", "administrar"}
permissoes_usuario = {"ler", "escrever"}

print(f"\n🔒 Admin: {permissoes_admin}")
print(f"👤 Usuário: {permissoes_usuario}")

print(f"\n✅ Admin ⊇ Usuario? {permissoes_admin >= permissoes_usuario}")
print(f"✅ Admin ⊃ Usuario? {permissoes_admin > permissoes_usuario}")
print(f"❌ Usuario ⊇ Admin? {permissoes_usuario >= permissoes_admin}")
print(f"\n✅ Método: {permissoes_admin.issuperset(permissoes_usuario)}")

# Exemplo 4: Disjuntos (isdisjoint) 🚫
turno_manha = {"Ana", "Bruno", "Carla"}
turno_tarde = {"Daniel", "Elena", "Fernando"}
turno_misto = {"Carla", "Daniel", "Gabriel"}

print(f"\n☀️ Manhã: {turno_manha}")
print(f"🌙 Tarde: {turno_tarde}")
print(f"🔄 Misto: {turno_misto}")

print(f"\n✅ Manhã e Tarde disjuntos? {turno_manha.isdisjoint(turno_tarde)}")
print(f"❌ Manhã e Misto disjuntos? {turno_manha.isdisjoint(turno_misto)}")
print(f"❌ Tarde e Misto disjuntos? {turno_tarde.isdisjoint(turno_misto)}")

# Exemplo 5: Validação de requisitos ✅
requisitos_projeto = {"Python", "SQL", "Git", "Docker"}
habilidades_dev1 = {"Python", "SQL", "Git", "Docker", "AWS"}
habilidades_dev2 = {"Python", "SQL", "JavaScript"}

print(f"\n📄 Requisitos: {requisitos_projeto}")
print(f"👨‍💻 Dev 1: {habilidades_dev1}")
print(f"👨‍💻 Dev 2: {habilidades_dev2}")

dev1_qualificado = requisitos_projeto <= habilidades_dev1
dev2_qualificado = requisitos_projeto <= habilidades_dev2

print(f"\n✅ Dev 1 qualificado? {dev1_qualificado}")
print(f"❌ Dev 2 qualificado? {dev2_qualificado}")

if not dev2_qualificado:
    faltando = requisitos_projeto - habilidades_dev2
    print(f"   ⚠️ Dev 2 precisa aprender: {faltando}")

# Exemplo 6: Comparação de versões de recursos 🔄
recursos_v1 = {"login", "dashboard", "relatorios"}
recursos_v2 = {"login", "dashboard", "relatorios", "api", "notificacoes"}

print(f"\n📦 Versão 1.0: {recursos_v1}")
print(f"📦 Versão 2.0: {recursos_v2}")

print(f"\n✅ v1 ⊂ v2 (v2 adiciona recursos)? {recursos_v1 < recursos_v2}")

novos_recursos = recursos_v2 - recursos_v1
print(f"🎉 Novos na v2: {novos_recursos}")

# Exemplo 7: Verificação de conflitos de horário ⏰
aula_a = {"8h", "9h", "10h"}
aula_b = {"10h", "11h", "12h"}
aula_c = {"14h", "15h", "16h"}

print(f"\n📚 Aula A: {sorted(aula_a)}")
print(f"📚 Aula B: {sorted(aula_b)}")
print(f"📚 Aula C: {sorted(aula_c)}")

conflito_ab = not aula_a.isdisjoint(aula_b)
conflito_ac = not aula_a.isdisjoint(aula_c)
conflito_bc = not aula_b.isdisjoint(aula_c)

print(f"\n⚠️ Conflito A-B? {conflito_ab}")
if conflito_ab:
    print(f"   Horários conflitantes: {aula_a & aula_b}")

print(f"✅ Conflito A-C? {conflito_ac}")
print(f"✅ Conflito B-C? {conflito_bc}")
````

## 5️⃣ Frozenset (Conjunto Imutável)

### 📖 Conceito
Frozenset é uma versão **imutável** de set. Uma vez criado, não pode ser modificado. Por ser imutável, pode ser usado como chave de dicionário ou elemento de outro set.

### 🔧 Criação
```python
# De lista/tupla
fs = frozenset([1, 2, 3, 4])

# De set
fs = frozenset({1, 2, 3, 4})

# Vazio
fs_vazio = frozenset()
```

### 📊 Características
* **Imutável**: Não permite add(), remove(), discard(), etc.
* **Hashável**: Pode ser chave de dicionário ou elemento de set
* **Operações de conjunto**: Suporta união, interseção, etc.
* **Todas operações de leitura**: len(), in, iteração funcionam

### ✅ Quando Usar
* Como chave de dicionário (set normal não pode)
* Como elemento de outro set (set de sets)
* Quando precisa garantir imutabilidade
* Em estruturas de dados complexas

### ⚠️ Diferenças vs Set Normal
```python
# Set normal - mutável
s = {1, 2, 3}
s.add(4)  # ✅ OK

# Frozenset - imutável
fs = frozenset([1, 2, 3])
fs.add(4)  # ❌ AttributeError

# Frozenset como chave
dict_key = {frozenset([1,2]): "valor"}  # ✅ OK
dict_key = {{1, 2}: "valor"}  # ❌ TypeError
```

````python
# Exemplo 1: Criando frozensets ❄️
fs1 = frozenset([1, 2, 3, 4, 5])
fs2 = frozenset({"a", "b", "c"})
fs3 = frozenset("python")  # De string
fs_vazio = frozenset()

print(f"❄️ Frozenset 1: {fs1}")
print(f"❄️ Frozenset 2: {fs2}")
print(f"❄️ Frozenset 3: {fs3}")
print(f"❄️ Vazio: {fs_vazio}")
print(f"\n🔍 Tipo: {type(fs1)}")

# Exemplo 2: Frozenset é imutável 🚫
fs = frozenset([1, 2, 3])
print(f"\n❄️ Frozenset original: {fs}")

# Tentar modificar lança erro
try:
    fs.add(4)
except AttributeError as e:
    print(f"❌ Erro ao tentar add(): {e}")

try:
    fs.remove(1)
except AttributeError as e:
    print(f"❌ Erro ao tentar remove(): {e}")

# Exemplo 3: Operações de conjunto funcionam ✅
fs_a = frozenset([1, 2, 3, 4])
fs_b = frozenset([3, 4, 5, 6])

print(f"\n❄️ A: {fs_a}")
print(f"❄️ B: {fs_b}")

print(f"\n🔗 União: {fs_a | fs_b}")
print(f"🎯 Interseção: {fs_a & fs_b}")
print(f"➡️ Diferença: {fs_a - fs_b}")
print(f"🔄 Dif. Simétrica: {fs_a ^ fs_b}")

# Exemplo 4: Frozenset como chave de dicionário 🔑
# Set normal NÃO pode ser chave
try:
    d = {{1, 2}: "valor"}  # Erro!
except TypeError as e:
    print(f"\n❌ Set como chave: {e}")

# Frozenset PODE ser chave
grupos = {
    frozenset(["Ana", "Bruno"]): "Grupo A",
    frozenset(["Carla", "Daniel"]): "Grupo B",
    frozenset(["Elena"]): "Grupo C"
}

print(f"\n✅ Frozensets como chaves:")
for grupo, nome in grupos.items():
    print(f"   {nome}: {grupo}")

# Buscar valor por frozenset
chave = frozenset(["Ana", "Bruno"])
print(f"\n🔍 Buscar {chave}: {grupos[chave]}")

# Exemplo 5: Set de frozensets (conjunto de conjuntos) 📦
conjuntos = {
    frozenset([1, 2]),
    frozenset([3, 4]),
    frozenset([5, 6])
}

print(f"\n📦 Set de frozensets: {conjuntos}")

# Adicionar novo frozenset
conjuntos.add(frozenset([7, 8]))
print(f"➕ Após adicionar: {conjuntos}")

# Verificar existência
print(f"\n✅ {frozenset([1,2])} existe? {frozenset([1,2]) in conjuntos}")
print(f"❌ {frozenset([9,10])} existe? {frozenset([9,10]) in conjuntos}")

# Exemplo 6: Operações de leitura 📊
fs = frozenset(["python", "java", "javascript", "c++", "ruby"])

print(f"\n💻 Linguagens: {fs}")
print(f"🔢 Quantidade: {len(fs)}")
print(f"✅ 'python' existe? {'python' in fs}")
print(f"❌ 'php' existe? {'php' in fs}")

print(f"\n🔁 Iterando:")
for lang in sorted(fs):
    print(f"   • {lang}")

# Exemplo 7: Caso prático - Permissões por papel 🔒
# Usar frozenset garante que permissões não sejam alteradas acidentalmente
permissoes_por_papel = {
    "admin": frozenset(["ler", "escrever", "excluir", "admin"]),
    "editor": frozenset(["ler", "escrever"]),
    "leitor": frozenset(["ler"])
}

print(f"\n🔒 Permissões por papel:")
for papel, perms in permissoes_por_papel.items():
    print(f"   {papel.capitalize()}: {perms}")

# Verificar permissões de usuário
usuario_papel = "editor"
usuario_perms = permissoes_por_papel[usuario_papel]

print(f"\n👤 Usuário ({usuario_papel}):")
print(f"   Pode escrever? {'escrever' in usuario_perms}")
print(f"   Pode excluir? {'excluir' in usuario_perms}")
print(f"   Pode administrar? {'admin' in usuario_perms}")
````

## 6️⃣ Set Comprehension (Compreensão de Sets)

### 📖 Conceito
Set comprehension é uma forma concisa de criar sets usando sintaxe similar a list comprehension. Permite transformar, filtrar e criar sets em uma única linha, com remoção automática de duplicatas.

### 🔧 Sintaxe Básica
```python
# Forma geral
{expressao for item in iteravel}

# Com condicional
{expressao for item in iteravel if condicao}

# Aninhado (evite se possível)
{x for lista in listas for x in lista}
```

### 📊 Exemplos de Padrões

**Transformar valores:**
```python
quadrados = {x**2 for x in range(5)}
# {0, 1, 4, 9, 16}
```

**Filtrar e transformar:**
```python
pares_quadrados = {x**2 for x in range(10) if x % 2 == 0}
# {0, 4, 16, 36, 64}
```

**Extrair elementos:**
```python
palavras = ["Python", "Java", "C++"]
primeiras = {p[0] for p in palavras}
# {'P', 'J', 'C'}
```

**Remover duplicatas com transformação:**
```python
nomes = ["ana", "Ana", "ANA", "bruno", "Bruno"]
unicos = {nome.lower() for nome in nomes}
# {'ana', 'bruno'}
```

### ✅ Vantagens
* Sintaxe concisa e legível
* Remoção automática de duplicatas
* Mais rápido que loop com add()
* Permite filtragem inline

### ⚠️ Quando Evitar
* Lógica muito complexa (use loop tradicional)
* Múltiplos níveis de aninhamento
* Quando legibilidade é comprometida

````python
# Exemplo 1: Criando sets com comprehension básica 🔢
quadrados = {x**2 for x in range(1, 11)}
print(f"🔢 Quadrados de 1 a 10: {sorted(quadrados)}")

cubos = {x**3 for x in range(1, 6)}
print(f"📦 Cubos de 1 a 5: {sorted(cubos)}")

# Exemplo 2: Filtrando com condição 🔍
pares = {x for x in range(20) if x % 2 == 0}
impares = {x for x in range(20) if x % 2 != 0}

print(f"\n➕ Pares (0-19): {sorted(pares)}")
print(f"➖ Ímpares (0-19): {sorted(impares)}")

# Múltiplos de 3
multiplos_3 = {x for x in range(1, 31) if x % 3 == 0}
print(f"\n🔢 Múltiplos de 3 (1-30): {sorted(multiplos_3)}")

# Exemplo 3: Transformando strings 🔤
nomes = ["ana", "Bruno", "CARLA", "daniel", "Elena"]
nomes_lower = {nome.lower() for nome in nomes}
nomes_upper = {nome.upper() for nome in nomes}
nomes_title = {nome.title() for nome in nomes}

print(f"\n👤 Nomes originais: {nomes}")
print(f"🔡 Lowercase: {sorted(nomes_lower)}")
print(f"🔠 Uppercase: {sorted(nomes_upper)}")
print(f"🎯 Title case: {sorted(nomes_title)}")

# Exemplo 4: Removendo duplicatas com transformação 🧼
emails = [
    "ana@email.com",
    "Ana@email.com",
    "ANA@EMAIL.COM",
    "bruno@email.com",
    "Bruno@email.com"
]

emails_unicos = {email.lower() for email in emails}
print(f"\n📧 Emails originais ({len(emails)}): {emails}")
print(f"✅ Emails únicos ({len(emails_unicos)}): {sorted(emails_unicos)}")

# Exemplo 5: Extraindo partes de strings 🔍
palavras = ["Python", "Java", "JavaScript", "C++", "Ruby"]

primeiras_letras = {palavra[0] for palavra in palavras}
ultimas_letras = {palavra[-1] for palavra in palavras}
comprimentos = {len(palavra) for palavra in palavras}

print(f"\n💻 Palavras: {palavras}")
print(f"🔤 Primeiras letras: {sorted(primeiras_letras)}")
print(f"🔤 Últimas letras: {sorted(ultimas_letras)}")
print(f"🔢 Comprimentos: {sorted(comprimentos)}")

# Exemplo 6: De lista aninhada (flatten) 📦
listas = [[1, 2, 3], [3, 4, 5], [5, 6, 7]]
elementos_unicos = {num for lista in listas for num in lista}

print(f"\n📦 Listas aninhadas: {listas}")
print(f"🎯 Elementos únicos: {sorted(elementos_unicos)}")

# Exemplo 7: Filtrando palavras por tamanho 📊
texto = "Python é uma linguagem de programação poderosa e versátil"
palavras = texto.split()

palavras_curtas = {p for p in palavras if len(p) <= 4}
palavras_longas = {p for p in palavras if len(p) > 7}

print(f"\n📝 Texto: {texto}")
print(f"👌 Palavras curtas (≤ 4): {palavras_curtas}")
print(f"👍 Palavras longas (> 7): {palavras_longas}")

# Exemplo 8: Caso prático - IDs válidos ✅
ids_raw = ["A001", "B002", "C003", "INVALID", "D004", "E", "F005"]

# Filtrar IDs válidos (começa com letra e tem 4 caracteres)
ids_validos = {id for id in ids_raw if len(id) == 4 and id[0].isalpha()}
ids_invalidos = {id for id in ids_raw if len(id) != 4 or not id[0].isalpha()}

print(f"\n🏷️ IDs originais: {ids_raw}")
print(f"✅ IDs válidos: {sorted(ids_validos)}")
print(f"❌ IDs inválidos: {sorted(ids_invalidos)}")

# Exemplo 9: Domínios de emails 🌐
contatos = [
    "ana@gmail.com",
    "bruno@yahoo.com",
    "carla@gmail.com",
    "daniel@outlook.com",
    "elena@gmail.com"
]

dominios = {email.split('@')[1] for email in contatos}
print(f"\n📧 Contatos: {contatos}")
print(f"🌐 Domínios únicos: {sorted(dominios)}")

# Exemplo 10: Valores absolutos únicos 🔢
numeros = [-5, -3, -1, 0, 1, 3, 5, -5, 3]
absolutos = {abs(n) for n in numeros}

print(f"\n🔢 Números: {numeros}")
print(f"🎯 Valores absolutos únicos: {sorted(absolutos)}")
````

## 7️⃣ Boas Práticas com Sets

### 🎯 10 Recomendações Essenciais

#### 1️⃣ Use Sets para Remover Duplicatas
```python
# ✅ Bom - Rápido e conciso
lista_unica = list(set(lista))

# ❌ Evite - Loop desnecessário
lista_unica = []
for item in lista:
    if item not in lista_unica:
        lista_unica.append(item)
```

#### 2️⃣ Prefira `in` para Testes de Pertinência
```python
# ✅ Bom - O(1) com set
if item in meu_set:  # Muito rápido
    ...

# ❌ Evite - O(n) com lista
if item in minha_lista:  # Lento para listas grandes
    ...
```

#### 3️⃣ Use Operações de Conjunto Apropriadas
```python
# ✅ Bom - Operadores claros
comuns = set_a & set_b
todos = set_a | set_b

# ❌ Evite - Loop manual
comuns = set()
for item in set_a:
    if item in set_b:
        comuns.add(item)
```

#### 4️⃣ Cuidado com Set Vazio
```python
# ❌ Errado - Cria dicionário!
vazio = {}

# ✅ Correto - Cria set
vazio = set()
```

#### 5️⃣ Use `discard()` Quando Remoção É Opcional
```python
# ✅ Bom - Não lança erro
meu_set.discard(item_que_pode_nao_existir)

# ❌ Evite - Requer try/except
try:
    meu_set.remove(item_que_pode_nao_existir)
except KeyError:
    pass
```

#### 6️⃣ Converta para Lista Quando Precisar de Ordem
```python
# ✅ Bom - Lista ordenada sem duplicatas
lista_ordenada = sorted(set(lista))

# Sets não garantem ordem
meu_set = {3, 1, 2}  # Ordem não é garantida
```

#### 7️⃣ Use Frozenset para Imutabilidade
```python
# ✅ Bom - Imutável, pode ser chave
permissoes = frozenset(['ler', 'escrever'])
config = {permissoes: 'editor'}

# ❌ Erro - Set normal não pode ser chave
config = {{1, 2}: 'valor'}  # TypeError
```

#### 8️⃣ Use Set Comprehension para Transformações
```python
# ✅ Bom - Conciso e eficiente
quadrados = {x**2 for x in numeros if x > 0}

# ❌ Evite - Mais verboso
quadrados = set()
for x in numeros:
    if x > 0:
        quadrados.add(x**2)
```

#### 9️⃣ Documente Quando Ordem Não Importa
```python
# ✅ Bom - Documenta que ordem não importa
def obter_tags_unicas(posts):
    """Retorna set de tags únicas (sem ordem)."""
    return {tag for post in posts for tag in post.tags}
```

#### 🔟 Performance: Quando Usar Sets
* **Use set**: Testes de pertinência frequentes
* **Use set**: Remover duplicatas
* **Use set**: Operações de conjunto
* **Use lista**: Quando ordem importa
* **Use lista**: Quando precisa de índices
* **Use lista**: Elementos podem repetir

### 📊 Complexidade de Operações
* **in / not in**: O(1) - Muito rápido
* **add()**: O(1)
* **remove()/discard()**: O(1)
* **União/Interseção**: O(len(set1) + len(set2))
* **Iteração**: O(n)
