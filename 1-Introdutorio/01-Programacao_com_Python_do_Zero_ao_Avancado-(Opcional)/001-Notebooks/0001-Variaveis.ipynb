# Variáveis em Python

Este notebook cobre os conceitos fundamentais sobre variáveis em Python, incluindo:
* Declaração e atribuição de variáveis
* Tipos de dados primitivos
* Verificação de tipos
* Operações básicas

## 1. Declaração e Atribuição de Variáveis

Em Python, não é necessário declarar o tipo da variável explicitamente. O tipo é inferido automaticamente com base no valor atribuído.

```python
# Atribuição simples de variável
nome_usuario = "Bob Esponja"
print(f"Nome: {nome_usuario}")

# Múltiplas atribuições
nome, idade = "Patrick Estrela", 45
print(f"Nome: {nome}, Idade: {idade}")

# Atribuição múltipla com mesmo valor
x = y = z = 0
print(f"x={x}, y={y}, z={z}")
```

## 2. Tipos de Dados Primitivos

Python possui vários tipos de dados primitivos:
* **int**: Números inteiros
* **float**: Números decimais
* **str**: Cadeias de caracteres (texto)
* **bool**: Valores booleanos (True/False)
* **complex**: Números complexos

Use a função `type()` para verificar o tipo de uma variável.

```python
# Inteiro (int)
media = 0
print(f"media = {media}, tipo: {type(media)}")

# Ponto flutuante (float)
preco = 19.99
print(f"preco = {preco}, tipo: {type(preco)}")

# String (str)
nome = "Bob Esponja"
print(f"nome = '{nome}', tipo: {type(nome)}")

# Booleano (bool)
esta_ativo = True
print(f"esta_ativo = {esta_ativo}, tipo: {type(esta_ativo)}")

# Número complexo (complex)
numero_complexo = 1 + 2j
print(f"numero_complexo = {numero_complexo}, tipo: {type(numero_complexo)}")
```

## 3. Verificação de Tipos com isinstance()

A função `isinstance()` verifica se uma variável é de um tipo específico:
* Retorna `True` se a variável for do tipo especificado
* Retorna `False` caso contrário
* Pode verificar múltiplos tipos usando uma tupla

**Sintaxe:** `isinstance(variavel, tipo)` ou `isinstance(variavel, (tipo1, tipo2, ...))`

```python
# Definindo variáveis para teste
a = 10
b = "Estrela"
c = 3.14

# Verificando tipo específico
print(f"{a} é int? {isinstance(a, int)}")
print(f"{a} é float? {isinstance(a, float)}")

# Verificando múltiplos tipos (int OU float)
print(f"{a} é int ou float? {isinstance(a, (int, float))}")
print(f"{c} é int ou float? {isinstance(c, (int, float))}")

# Verificando strings
print(f"{b} é str? {isinstance(b, str)}")
```

## 4. Operações com Variáveis

Variáveis podem ser usadas em operações aritméticas e outros cálculos:
* **Adição:** `+`
* **Subtração:** `-`
* **Multiplicação:** `*`
* **Divisão:** `/`
* **Divisão inteira:** `//`
* **Módulo (resto):** `%`
* **Potência:** `**`

```python
# Definindo variáveis
a = 40
b = 3

# Operações aritméticas
print(f"Adição: {a} + {b} = {a + b}")
print(f"Subtração: {a} - {b} = {a - b}")
print(f"Multiplicação: {a} * {b} = {a * b}")
print(f"Divisão: {a} / {b} = {a / b}")
print(f"Divisão inteira: {a} // {b} = {a // b}")
print(f"Módulo (resto): {a} % {b} = {a % b}")
print(f"Potência: {a} ** 2 = {a ** 2}")
```

## Boas Práticas ao Trabalhar com Variáveis

### Nomeação de Variáveis
* Use nomes descritivos: `nome_usuario` em vez de `n`
* Siga o padrão snake_case: `idade_cliente` em vez de `idadeCliente`
* Evite nomes reservados do Python: `list`, `str`, `type`, etc.
* Use nomes em inglês para código profissional (opcional)

### Tipos de Dados
* Python é dinamicamente tipado, mas seja consistente
* Use `isinstance()` em vez de `type()` para verificações
* Documente tipos esperados em funções usando type hints
