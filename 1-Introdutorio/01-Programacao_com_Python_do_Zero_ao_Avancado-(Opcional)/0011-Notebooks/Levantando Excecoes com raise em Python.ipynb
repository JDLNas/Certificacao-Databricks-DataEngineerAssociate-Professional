# 🚀 Levantando Exceções com raise em Python

## 📋 Visão Geral
Este notebook explora o comando **raise**, que permite levantar exceções intencionalmente no código. Raise é fundamental para validação de dados, controle de fluxo e criação de exceções personalizadas que tornam o código mais expressivo e robusto.

* Criar exceções personalizadas simples

* Seguir boas práticas no uso de raise

## 1️⃣ Conceitos Básicos do raise

### 📖 Conceito
**raise** é a palavra-chave usada para **levantar** (ou lançar) exceções intencionalmente. Enquanto try/except **captura** erros, raise **cria** erros propositadamente para sinalizar condições excepcionais.

### 🔧 Sintaxe Básica
```python
# Levantar exceção sem mensagem
raise ValueError

# Levantar exceção com mensagem
raise ValueError("Mensagem descritiva")

# Relevantar exceção capturada
except ValueError:
    raise  # Relevanta a mesma exceção
```

### 📊 Quando Usar raise
* **Validação de entrada**: Dados inválidos ou fora do esperado
* **Pré-condições**: Verificar estado antes de operação
* **Pós-condições**: Garantir resultado válido
* **Limites de negócio**: Regras de domínio violadas
* **Recursos indisponíveis**: Arquivo/conexão não encontrados

### ✅ Benefícios
* **Fail Fast**: Detecta problemas cedo
* **Código defensivo**: Previne estados inválidos
* **Clareza**: Sinaliza explicitamente condições de erro
* **Manutenção**: Facilita debugging
* **Documentação**: Comunica expectativas

### 📊 Exceções Built-in Comuns para raise
* **ValueError**: Valor inadequado/inválido
* **TypeError**: Tipo incorreto
* **KeyError**: Chave não existe
* **IndexError**: Índice fora do intervalo
* **AttributeError**: Atributo não encontrado
* **RuntimeError**: Erro genérico de execução
* **NotImplementedError**: Funcionalidade não implementada

````python
# Exemplo 1: raise com ValueError ⚠️
def calcular_desconto(preco, percentual):
    """Calcula desconto com validação."""
    if preco < 0:
        raise ValueError("Preço não pode ser negativo")
    if percentual < 0 or percentual > 100:
        raise ValueError("Percentual deve estar entre 0 e 100")
    
    desconto = preco * (percentual / 100)
    return preco - desconto

print("⚠️ Validação com ValueError:")
testes = [(100, 10), (100, -5), (-50, 10), (200, 25)]
for preco, perc in testes:
    try:
        resultado = calcular_desconto(preco, perc)
        print(f"   R$ {preco} - {perc}% = R$ {resultado:.2f}")
    except ValueError as e:
        print(f"   ❌ R$ {preco} - {perc}%: {e}")
print()

# Exemplo 2: raise com TypeError 🔢
def multiplicar(a, b):
    """Multiplica dois números."""
    if not isinstance(a, (int, float)):
        raise TypeError(f"Primeiro argumento deve ser número, recebido: {type(a).__name__}")
    if not isinstance(b, (int, float)):
        raise TypeError(f"Segundo argumento deve ser número, recebido: {type(b).__name__}")
    
    return a * b

print("🔢 Validação com TypeError:")
testes = [(5, 3), ("5", 3), (5, [1, 2]), (10, 2.5)]
for a, b in testes:
    try:
        resultado = multiplicar(a, b)
        print(f"   {a} × {b} = {resultado}")
    except TypeError as e:
        print(f"   ❌ {repr(a)} × {repr(b)}: {e}")
print()

# Exemplo 3: raise com mensagens descritivas 📝
def dividir_lista(lista, tamanho_grupo):
    """Divide lista em grupos de tamanho especificado."""
    if not isinstance(lista, list):
        raise TypeError(f"Esperado list, recebido {type(lista).__name__}")
    
    if tamanho_grupo <= 0:
        raise ValueError(f"Tamanho do grupo deve ser positivo, recebido: {tamanho_grupo}")
    
    if len(lista) == 0:
        raise ValueError("Lista não pode estar vazia")
    
    grupos = [lista[i:i+tamanho_grupo] for i in range(0, len(lista), tamanho_grupo)]
    return grupos

print("📝 Mensagens descritivas:")
lista_nums = [1, 2, 3, 4, 5, 6, 7, 8, 9]
testes = [
    (lista_nums, 3),
    (lista_nums, -2),
    ([], 2),
    ("123", 2)
]

for lst, tam in testes:
    try:
        resultado = dividir_lista(lst, tam)
        print(f"   {lst[:3]}... dividido em {tam}: {resultado}")
    except (ValueError, TypeError) as e:
        print(f"   ❌ {e}")
print()

# Exemplo 4: Validação de intervalo 📊
def definir_volume(volume):
    """Define volume entre 0 e 100."""
    if volume < 0:
        raise ValueError(f"Volume {volume} abaixo do mínimo (0)")
    if volume > 100:
        raise ValueError(f"Volume {volume} acima do máximo (100)")
    
    return f"✅ Volume ajustado para {volume}"

print("📊 Validação de intervalo:")
volumes = [50, 0, 100, -10, 150, 75]
for vol in volumes:
    try:
        resultado = definir_volume(vol)
        print(f"   {resultado}")
    except ValueError as e:
        print(f"   ❌ {e}")
print()

# Exemplo 5: Validação de formato 📧
def validar_email(email):
    """Valida formato básico de email."""
    if not isinstance(email, str):
        raise TypeError("Email deve ser uma string")
    
    if "@" not in email:
        raise ValueError(f"Email '{email}' deve conter @")
    
    if "." not in email.split("@")[1]:
        raise ValueError(f"Email '{email}' deve ter domínio válido")
    
    partes = email.split("@")
    if len(partes[0]) == 0:
        raise ValueError(f"Email '{email}' deve ter nome de usuário")
    
    return f"✅ Email '{email}' válido"

print("📧 Validação de email:")
emails = [
    "usuario@dominio.com",
    "semdominio@com",
    "semarroba.com",
    "@dominio.com",
    123
]

for email in emails:
    try:
        resultado = validar_email(email)
        print(f"   {resultado}")
    except (ValueError, TypeError) as e:
        print(f"   ❌ {e}")
print()

# Exemplo 6: Validação de pré-condições 🚫
def sacar_dinheiro(saldo, valor):
    """Realiza saque com validações."""
    if valor <= 0:
        raise ValueError(f"Valor {valor} inválido para saque")
    
    if valor > saldo:
        raise ValueError(f"Saldo insuficiente: R$ {saldo:.2f} < R$ {valor:.2f}")
    
    if valor % 10 != 0:
        raise ValueError(f"Valor R$ {valor} deve ser múltiplo de 10")
    
    novo_saldo = saldo - valor
    return novo_saldo

print("🚫 Validação de pré-condições:")
saldo_conta = 1000.00
saques = [100, 0, 1500, 55, 200]

for valor in saques:
    try:
        novo_saldo = sacar_dinheiro(saldo_conta, valor)
        print(f"   Saque R$ {valor:.2f}: ✅ Novo saldo R$ {novo_saldo:.2f}")
        saldo_conta = novo_saldo
    except ValueError as e:
        print(f"   Saque R$ {valor:.2f}: ❌ {e}")
````

## 2️⃣ Exceções Personalizadas Simples

### 📖 Conceito
Exceções personalizadas são classes que herdam de `Exception` (ou suas subclasses). Elas tornam o código mais expressivo e permitem captura seletiva de erros específicos do domínio.

### 🔧 Criando Exceção Personalizada
```python
# Simples
class MeuErro(Exception):
    pass

raise MeuErro("Algo deu errado")

# Com docstring
class IdadeInvalidaError(Exception):
    """Levantada quando idade é inválida."""
    pass

raise IdadeInvalidaError("Idade deve ser positiva")
```

### 📊 Convenções de Nomenclatura
* Termine com **Error** ou **Exception**
* Use CamelCase (IdadeInvalidaError)
* Nome descritivo da condição de erro
* Agrupe por domínio (ValidacaoError, ConexaoError)

### ✅ Vantagens
* **Expressividade**: Nome comunica o tipo de erro
* **Captura seletiva**: Tratar erros específicos
* **Organização**: Agrupa erros relacionados
* **Documentação**: Self-documenting code
* **Manutenção**: Mais fácil de localizar/tratar

### 🎯 Quando Criar Exceções Personalizadas
* Erros específicos do seu domínio/aplicação
* Quando built-ins não são suficientemente descritivas
* Para adicionar comportamento/atributos extras
* Quando precisar capturar seletivamente

````python
# Exemplo 1: Exceção personalizada básica ✨
class IdadeInvalidaError(Exception):
    """Levantada quando idade é inválida."""
    pass

def validar_idade_motorista(idade):
    """Valida idade mínima para motorista."""
    if idade < 18:
        raise IdadeInvalidaError(f"Idade {idade} insuficiente (mínimo: 18 anos)")
    if idade > 100:
        raise IdadeInvalidaError(f"Idade {idade} inválida (máximo: 100 anos)")
    return f"✅ Idade {idade} válida para CNH"

print("✨ Exceção personalizada básica:")
idades = [25, 16, 100, 18, 105, 50]
for idade in idades:
    try:
        resultado = validar_idade_motorista(idade)
        print(f"   {resultado}")
    except IdadeInvalidaError as e:
        print(f"   ❌ {e}")
print()

# Exemplo 2: Exceção para domínio de negócio 🏢
class SaldoInsuficienteError(Exception):
    """Levantada quando saldo é insuficiente."""
    pass

class LimiteDiarioExcedidoError(Exception):
    """Levantada quando limite diário é excedido."""
    pass

def processar_compra(valor, saldo, limite_diario, gasto_hoje):
    """Processa compra com validações."""
    if valor > saldo:
        raise SaldoInsuficienteError(f"Saldo R$ {saldo:.2f} insuficiente para R$ {valor:.2f}")
    
    if gasto_hoje + valor > limite_diario:
        raise LimiteDiarioExcedidoError(f"Limite diário R$ {limite_diario:.2f} seria excedido")
    
    return f"✅ Compra aprovada: R$ {valor:.2f}"

print("🏢 Exceções de domínio:")
saldo = 1000.00
limite = 500.00
gasto = 200.00

compras = [100, 350, 1500]
for valor in compras:
    try:
        resultado = processar_compra(valor, saldo, limite, gasto)
        print(f"   {resultado}")
        gasto += valor
    except (SaldoInsuficienteError, LimiteDiarioExcedidoError) as e:
        print(f"   ❌ R$ {valor:.2f}: {e}")
print()

# Exemplo 3: Exceções para validação de dados 📝
class CpfInvalidoError(Exception):
    """CPF inválido."""
    pass

class EmailInvalidoError(Exception):
    """Email inválido."""
    pass

def validar_cadastro(cpf, email):
    """Valida dados de cadastro."""
    # Valida CPF (formato simples)
    cpf_limpo = cpf.replace(".", "").replace("-", "")
    if len(cpf_limpo) != 11 or not cpf_limpo.isdigit():
        raise CpfInvalidoError(f"CPF '{cpf}' inválido")
    
    # Valida email
    if "@" not in email or "." not in email:
        raise EmailInvalidoError(f"Email '{email}' inválido")
    
    return "✅ Cadastro válido"

print("📝 Validação de dados:")
cadastros = [
    ("123.456.789-01", "user@email.com"),
    ("123.456", "user@email.com"),
    ("123.456.789-01", "useremail.com"),
    ("987.654.321-00", "admin@company.com")
]

for cpf, email in cadastros:
    try:
        resultado = validar_cadastro(cpf, email)
        print(f"   {resultado}")
    except (CpfInvalidoError, EmailInvalidoError) as e:
        print(f"   ❌ {e}")
print()

# Exemplo 4: Exceção para recursos não encontrados 🔍
class ProdutoNaoEncontradoError(Exception):
    """Produto não encontrado."""
    pass

class EstoqueInsuficienteError(Exception):
    """Estoque insuficiente."""
    pass

estoque = {
    "notebook": 5,
    "mouse": 10,
    "teclado": 0
}

def verificar_disponibilidade(produto, quantidade):
    """Verifica disponibilidade de produto."""
    if produto not in estoque:
        raise ProdutoNaoEncontradoError(f"Produto '{produto}' não encontrado")
    
    if estoque[produto] < quantidade:
        raise EstoqueInsuficienteError(
            f"Estoque de '{produto}': {estoque[produto]} < {quantidade} solicitado(s)"
        )
    
    return f"✅ {quantidade} unidade(s) de '{produto}' disponível(is)"

print("🔍 Recursos não encontrados:")
pedidos = [
    ("notebook", 3),
    ("monitor", 1),
    ("teclado", 1),
    ("mouse", 5)
]

for prod, qtd in pedidos:
    try:
        resultado = verificar_disponibilidade(prod, qtd)
        print(f"   {resultado}")
    except (ProdutoNaoEncontradoError, EstoqueInsuficienteError) as e:
        print(f"   ❌ {e}")
print()

# Exemplo 5: Exceção para estado inválido 🚫
class ConexaoFechadaError(Exception):
    """Operação em conexão fechada."""
    pass

class Conexao:
    """Simula conexão com banco de dados."""
    def __init__(self, nome):
        self.nome = nome
        self.aberta = False
    
    def abrir(self):
        self.aberta = True
        return f"✅ Conexão '{self.nome}' aberta"
    
    def fechar(self):
        self.aberta = False
        return f"🔒 Conexão '{self.nome}' fechada"
    
    def executar_query(self, query):
        if not self.aberta:
            raise ConexaoFechadaError(f"Conexão '{self.nome}' está fechada")
        return f"✅ Query executada: {query}"

print("🚫 Estado inválido:")
conn = Conexao("PostgreSQL")

operacoes = [
    ("executar", "SELECT * FROM users"),
    ("abrir", None),
    ("executar", "SELECT * FROM products"),
    ("fechar", None),
    ("executar", "SELECT * FROM orders")
]

for op, query in operacoes:
    try:
        if op == "executar":
            resultado = conn.executar_query(query)
            print(f"   {resultado}")
        elif op == "abrir":
            resultado = conn.abrir()
            print(f"   {resultado}")
        elif op == "fechar":
            resultado = conn.fechar()
            print(f"   {resultado}")
    except ConexaoFechadaError as e:
        print(f"   ❌ {e}")
````

## 3️⃣ Exceções com Atributos Customizados

### 📖 Conceito
Exceções podem armazenar dados adicionais como atributos. Isso permite capturar contexto do erro para tratamento mais sofisticado ou logging detalhado.

### 🔧 Criando com Construtor
```python
class SaldoInsuficienteError(Exception):
    def __init__(self, saldo, valor):
        self.saldo = saldo
        self.valor = valor
        self.deficit = valor - saldo
        mensagem = f"Saldo R$ {saldo} < Valor R$ {valor}"
        super().__init__(mensagem)

# Uso
try:
    raise SaldoInsuficienteError(100, 150)
except SaldoInsuficienteError as e:
    print(f"Déficit: R$ {e.deficit}")
```

### 📊 Vantagens de Atributos
* **Contexto rico**: Armazena dados relevantes do erro
* **Tratamento inteligente**: Decisões baseadas em atributos
* **Logging detalhado**: Mais informações para debug
* **Recovery**: Dados para tentar recuperação
* **Análise**: Métricas sobre erros

### ✅ O Que Armazenar
* Valores que causaram o erro
* Estado do sistema no momento
* IDs de recursos relacionados
* Timestamps
* Códigos de erro
* URLs ou paths
* Mensagens traduzíveis

````python
# Exemplo 1: Exceção com valores do erro 💰
class SaldoInsuficienteError(Exception):
    """Saldo insuficiente para operação."""
    def __init__(self, saldo_atual, valor_solicitado):
        self.saldo_atual = saldo_atual
        self.valor_solicitado = valor_solicitado
        self.deficit = valor_solicitado - saldo_atual
        mensagem = f"Saldo R$ {saldo_atual:.2f} < R$ {valor_solicitado:.2f}"
        super().__init__(mensagem)

def sacar(saldo, valor):
    """Realiza saque."""
    if valor > saldo:
        raise SaldoInsuficienteError(saldo, valor)
    return saldo - valor

print("💰 Exceção com atributos:")
saldo_conta = 1000.00
saques = [200, 500, 1500, 300]

for valor in saques:
    try:
        novo_saldo = sacar(saldo_conta, valor)
        print(f"   Saque R$ {valor:.2f}: ✅ Novo saldo R$ {novo_saldo:.2f}")
        saldo_conta = novo_saldo
    except SaldoInsuficienteError as e:
        print(f"   Saque R$ {valor:.2f}: ❌ {e}")
        print(f"      Faltam R$ {e.deficit:.2f}")
print()

# Exemplo 2: Exceção com timestamp e contexto ⏰
from datetime import datetime

class ErroProcessamento(Exception):
    """Erro durante processamento."""
    def __init__(self, mensagem, registro_id, timestamp=None):
        super().__init__(mensagem)
        self.registro_id = registro_id
        self.timestamp = timestamp or datetime.now()
        self.mensagem = mensagem

def processar_registro(registro_id, dados):
    """Processa registro com validação."""
    if not dados:
        raise ErroProcessamento(
            "Dados vazios",
            registro_id=registro_id
        )
    
    if "email" not in dados:
        raise ErroProcessamento(
            "Campo 'email' obrigatório",
            registro_id=registro_id
        )
    
    return f"✅ Registro {registro_id} processado"

print("⏰ Exceção com timestamp:")
registros = [
    (1001, {"nome": "Ana", "email": "ana@email.com"}),
    (1002, {}),
    (1003, {"nome": "Carlos"}),
    (1004, {"nome": "Diana", "email": "diana@email.com"})
]

for reg_id, dados in registros:
    try:
        resultado = processar_registro(reg_id, dados)
        print(f"   {resultado}")
    except ErroProcessamento as e:
        print(f"   ❌ Registro {e.registro_id}: {e.mensagem}")
        print(f"      Timestamp: {e.timestamp.strftime('%H:%M:%S')}")
print()

# Exemplo 3: Exceção com código de erro 🔢
class ErroValidacao(Exception):
    """Erro de validação com código."""
    def __init__(self, mensagem, codigo, campo=None):
        super().__init__(mensagem)
        self.codigo = codigo
        self.campo = campo
        self.mensagem = mensagem

def validar_usuario(dados):
    """Valida dados de usuário."""
    if "username" not in dados:
        raise ErroValidacao(
            "Nome de usuário obrigatório",
            codigo="VAL001",
            campo="username"
        )
    
    if len(dados.get("username", "")) < 3:
        raise ErroValidacao(
            "Nome de usuário deve ter mínimo 3 caracteres",
            codigo="VAL002",
            campo="username"
        )
    
    if "senha" not in dados:
        raise ErroValidacao(
            "Senha obrigatória",
            codigo="VAL003",
            campo="senha"
        )
    
    if len(dados.get("senha", "")) < 8:
        raise ErroValidacao(
            "Senha deve ter mínimo 8 caracteres",
            codigo="VAL004",
            campo="senha"
        )
    
    return "✅ Usuário válido"

print("🔢 Códigos de erro:")
usuarios = [
    {"username": "ana", "senha": "senha123"},
    {},
    {"username": "ab", "senha": "senha123"},
    {"username": "carlos", "senha": "123"}
]

for dados in usuarios:
    try:
        resultado = validar_usuario(dados)
        print(f"   {resultado}")
    except ErroValidacao as e:
        print(f"   ❌ Erro {e.codigo} ({e.campo}): {e.mensagem}")
print()

# Exemplo 4: Exceção com métodos auxiliares 🛠️
class ErroPedido(Exception):
    """Erro ao processar pedido."""
    def __init__(self, mensagem, pedido_id, itens_problema=None):
        super().__init__(mensagem)
        self.pedido_id = pedido_id
        self.itens_problema = itens_problema or []
        self.timestamp = datetime.now()
    
    def to_dict(self):
        """Converte para dicionário."""
        return {
            "pedido_id": self.pedido_id,
            "mensagem": str(self),
            "itens_problema": self.itens_problema,
            "timestamp": self.timestamp.isoformat()
        }
    
    def resumo(self):
        """Retorna resumo do erro."""
        return f"Pedido {self.pedido_id}: {len(self.itens_problema)} item(ns) com problema"

def processar_pedido(pedido_id, itens):
    """Processa pedido validando itens."""
    if not itens:
        raise ErroPedido(
            "Pedido sem itens",
            pedido_id=pedido_id
        )
    
    itens_invalidos = [item for item in itens if item.get("preco", 0) <= 0]
    if itens_invalidos:
        raise ErroPedido(
            "Itens com preço inválido",
            pedido_id=pedido_id,
            itens_problema=itens_invalidos
        )
    
    return f"✅ Pedido {pedido_id} processado: {len(itens)} itens"

print("🛠️ Métodos auxiliares:")
pedidos = [
    (5001, [{"nome": "Mouse", "preco": 50}]),
    (5002, []),
    (5003, [{"nome": "Teclado", "preco": 0}, {"nome": "Monitor", "preco": 500}])
]

for ped_id, itens in pedidos:
    try:
        resultado = processar_pedido(ped_id, itens)
        print(f"   {resultado}")
    except ErroPedido as e:
        print(f"   ❌ {e.resumo()}")
        print(f"      Detalhes: {e.to_dict()['mensagem']}")
````

## 4️⃣ Relevantar Exceções (Re-raising)

### 📖 Conceito
**Relevantar** significa capturar uma exceção, fazer algo (logging, limpeza), e então lançá-la novamente para que seja tratada em nível superior.

### 🔧 Sintaxe
```python
try:
    operacao()
except ValueError as e:
    print(f"LOG: {e}")
    raise  # Relevanta a mesma exceção
```

**Relevantar com contexto:**
```python
try:
    operacao()
except ValueError as e:
    raise RuntimeError("Falha na operação") from e
```

### 📊 Quando Relevantar
* **Logging**: Registrar erro antes de propagar
* **Limpeza**: Liberar recursos antes de propagar
* **Enriquecimento**: Adicionar contexto
* **Transformação**: Converter tipo de exceção
* **Monitoramento**: Métricas/alertas

### ✅ raise vs raise from
* **raise**: Relevanta mesma exceção, preserva traceback
* **raise ... from e**: Nova exceção, mantendo causa original
* **raise ... from None**: Suprime contexto original

### ⚠️ Cuidados
```python
# ❌ Não faça isso
except Exception as e:
    raise e  # Perde informação do traceback

# ✅ Faça isso
except Exception:
    raise  # Preserva traceback completo
```

````python
# Exemplo 1: Relevantar após logging 📝
def processar_com_log(dados):
    """Processa dados e loga erros."""
    try:
        resultado = 100 / dados
        return resultado
    except ZeroDivisionError as e:
        print(f"   📝 LOG: ZeroDivisionError capturado")
        print(f"   📝 LOG: Tipo: {type(e).__name__}")
        print(f"   📝 LOG: Mensagem: {e}")
        raise  # Relevanta para nível superior

print("📝 Relevantar após logging:")
try:
    processar_com_log(0)
except ZeroDivisionError:
    print("   ❌ Erro capturado no nível superior")
print()

# Exemplo 2: Relevantar após limpeza 🧹
class Recurso:
    """Simula recurso que precisa ser liberado."""
    def __init__(self, nome):
        self.nome = nome
        self.alocado = False
    
    def alocar(self):
        self.alocado = True
        print(f"   🔓 Recurso '{self.nome}' alocado")
    
    def liberar(self):
        self.alocado = False
        print(f"   🔒 Recurso '{self.nome}' liberado")

def usar_recurso(falhar=False):
    """Usa recurso com limpeza garantida."""
    recurso = Recurso("Database")
    
    try:
        recurso.alocar()
        
        if falhar:
            raise ValueError("Operação falhou")
        
        print(f"   ✅ Operação concluída")
        return "Sucesso"
    
    except ValueError:
        print(f"   ⚠️ Limpando antes de propagar erro")
        recurso.liberar()  # Limpeza
        raise  # Relevanta

print("🧹 Relevantar após limpeza:")
print("\nCaso 1: Sucesso")
usar_recurso(falhar=False)

print("\nCaso 2: Falha")
try:
    usar_recurso(falhar=True)
except ValueError:
    print("   ❌ Erro tratado no nível superior")
print()

# Exemplo 3: Transformar tipo de exceção 🔄
import json

class ErroConfiguracao(Exception):
    """Erro ao carregar configuração."""
    pass

def carregar_config(json_string):
    """Carrega configuração de JSON."""
    try:
        config = json.loads(json_string)
        return config
    except json.JSONDecodeError as e:
        # Transforma JSONDecodeError em ErroConfiguracao
        raise ErroConfiguracao(f"JSON inválido: {e}") from e

print("🔄 Transformar tipo de exceção:")
configs = [
    '{"host": "localhost", "port": 5432}',
    '{host: localhost}',
    '[1, 2, 3]'
]

for cfg in configs:
    try:
        resultado = carregar_config(cfg)
        if isinstance(resultado, dict):
            print(f"   ✅ Config carregada: {list(resultado.keys())}")
        else:
            print(f"   ✅ JSON válido (tipo: {type(resultado).__name__})")
    except ErroConfiguracao as e:
        print(f"   ❌ {e}")
print()

# Exemplo 4: Enriquecer com contexto 📊
class ErroProcessamentoDados(Exception):
    """Erro ao processar dados."""
    pass

def processar_lote(registros):
    """Processa lote de registros."""
    for i, registro in enumerate(registros, 1):
        try:
            # Simula processamento
            if "valor" not in registro:
                raise KeyError("Campo 'valor' ausente")
            
            if registro["valor"] < 0:
                raise ValueError("Valor negativo")
            
            print(f"   ✅ Registro {i} processado")
        
        except (KeyError, ValueError) as e:
            # Enriquece com contexto
            raise ErroProcessamentoDados(
                f"Erro no registro {i}: {e}"
            ) from e

print("📊 Enriquecer com contexto:")
registros = [
    {"valor": 100},
    {"nome": "Item 2"},
    {"valor": -50}
]

try:
    processar_lote(registros)
except ErroProcessamentoDados as e:
    print(f"   ❌ {e}")
    print(f"   🔍 Causa: {e.__cause__}")
print()

# Exemplo 5: Relevantar seletivamente 🎯
def operacao_critica(valor):
    """Operação com tratamento seletivo."""
    try:
        if valor == 0:
            raise ZeroDivisionError("Divisão por zero")
        elif valor < 0:
            raise ValueError("Valor negativo")
        
        resultado = 100 / valor
        return resultado
    
    except ZeroDivisionError:
        print(f"   📝 LOG: ZeroDivisionError - valor={valor}")
        raise  # Sempre relevanta
    
    except ValueError as e:
        print(f"   ⚠️ WARN: ValueError - valor={valor}")
        # Só relevanta se valor muito negativo
        if valor < -100:
            print(f"   ⚠️ Valor crítico, relevantando")
            raise
        else:
            print(f"   🔧 Valor aceitável, retornando None")
            return None

print("🎯 Relevantar seletivamente:")
valores = [10, 0, -50, -150]

for val in valores:
    try:
        resultado = operacao_critica(val)
        if resultado is not None:
            print(f"   100 / {val} = {resultado}")
    except (ZeroDivisionError, ValueError) as e:
        print(f"   ❌ Erro propagado: {type(e).__name__}")
print()
````

## 5️⃣ Boas Práticas com raise

### 🎯 10 Recomendações Essenciais

#### 1️⃣ Use Tipo de Exceção Apropriado
```python
# ❌ Ruim - muito genérico
if idade < 0:
    raise Exception("Idade inválida")

# ✅ Bom - tipo específico
if idade < 0:
    raise ValueError("Idade não pode ser negativa")
```

#### 2️⃣ Forneça Mensagens Descritivas
```python
# ❌ Ruim
raise ValueError("Erro")

# ✅ Bom
raise ValueError(f"Idade {idade} inválida: deve estar entre 0 e 150")
```

#### 3️⃣ Valide Cedo (Fail Fast)
```python
# ✅ Bom - valida no início
def processar(valor):
    if valor < 0:
        raise ValueError("Valor deve ser positivo")
    if not isinstance(valor, (int, float)):
        raise TypeError("Valor deve ser numérico")
    
    # Resto do processamento
```

#### 4️⃣ Documente Exceções Levantadas
```python
def dividir(a, b):
    """
    Divide dois números.
    
    Args:
        a (float): Dividendo
        b (float): Divisor
    
    Returns:
        float: Resultado da divisão
    
    Raises:
        ZeroDivisionError: Se b for zero
        TypeError: Se a ou b não forem numéricos
    """
    return a / b
```

#### 5️⃣ Crie Exceções Personalizadas para Seu Domínio
```python
# ✅ Bom - expressivo
class SaldoInsuficienteError(Exception):
    pass

if saldo < valor:
    raise SaldoInsuficienteError(f"Saldo {saldo} < {valor}")
```

#### 6️⃣ Não Use raise para Fluxo Normal
```python
# ❌ Ruim - exceção para controle de fluxo
try:
    valor = dicionario[chave]
except KeyError:
    valor = default

# ✅ Bom - método apropriado
valor = dicionario.get(chave, default)
```

#### 7️⃣ Use raise Sem Argumentos para Relevantar
```python
# ❌ Ruim - perde traceback
except ValueError as e:
    raise e

# ✅ Bom - preserva traceback
except ValueError:
    raise
```

#### 8️⃣ Use raise from para Encadear Exceções
```python
# ✅ Bom - mantém contexto
try:
    processar_json(texto)
except JSONDecodeError as e:
    raise ErroConfiguracao("Config inválida") from e
```

#### 9️⃣ Agrupe Exceções Relacionadas
```python
# ✅ Bom - hierarquia clara
class ErroValidacao(Exception):
    pass

class CpfInvalidoError(ErroValidacao):
    pass

class EmailInvalidoError(ErroValidacao):
    pass
```

#### 🔟 Inclua Dados Relevantes em Atributos
```python
# ✅ Bom - contexto rico
class ErroRequisicao(Exception):
    def __init__(self, mensagem, url, status_code):
        super().__init__(mensagem)
        self.url = url
        self.status_code = status_code
```
