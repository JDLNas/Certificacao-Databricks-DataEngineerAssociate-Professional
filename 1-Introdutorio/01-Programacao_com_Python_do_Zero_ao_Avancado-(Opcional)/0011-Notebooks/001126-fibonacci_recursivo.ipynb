# Sequência de Fibonacci - Implementações e Otimizações

## Conceitos Matemáticos

### O que é a Sequência de Fibonacci?

A sequência de Fibonacci é uma sequência matemática onde cada número é a soma dos dois anteriores.

**Definição matemática:**
* F(0) = 0
* F(1) = 1
* F(n) = F(n-1) + F(n-2) para n ≥ 2

### Primeiros números da sequência:

| n | F(n) |
|---|------|
| 0 | 0 |
| 1 | 1 |
| 2 | 1 |
| 3 | 2 |
| 4 | 3 |
| 5 | 5 |
| 6 | 8 |
| 7 | 13 |
| 8 | 21 |
| 9 | 34 |
| 10 | 55 |

### Contexto histórico:
A sequência foi descrita por Leonardo Fibonacci no século XIII ao estudar o crescimento de populações de coelhos. Aparece frequentemente na natureza (conchas, plantas, galáxias).

## Implementação Recursiva Ingênua

A abordagem mais direta é traduzir a definição matemática diretamente em código recursivo.

````python
def fibonacci_recursivo(n):
    """
    Calcula o n-ésimo número de Fibonacci usando recursão simples.
    
    ATENÇÃO: Esta implementação é MUITO INEFICIENTE para n >= 30!
    
    Parâmetros:
        n (int): Posição na sequência (n >= 0)
    
    Retorna:
        int: N-ésimo número de Fibonacci
    """
    # Casos base
    if n <= 1:
        return n
    
    # Caso recursivo
    return fibonacci_recursivo(n - 1) + fibonacci_recursivo(n - 2)

# Exemplos de uso
print("=" * 50)
print("FIBONACCI RECURSIVO - Primeiros números")
print("=" * 50)

for i in range(11):
    print(f"F({i}) = {fibonacci_recursivo(i)}")
````

## ❌ O Problema: Ineficiência Exponencial

### Por que a recursão simples é ineficiente?

A implementação recursiva ingênua recalcula os mesmos valores **múltiplas vezes**.

### Exemplo: Calculando F(5)

```
fibonacci(5)
├── fibonacci(4)
│   ├── fibonacci(3)
│   │   ├── fibonacci(2)
│   │   │   ├── fibonacci(1) → 1
│   │   │   └── fibonacci(0) → 0
│   │   └── fibonacci(1) → 1
│   └── fibonacci(2)          ← RECALCULADO!
│       ├── fibonacci(1) → 1
│       └── fibonacci(0) → 0
└── fibonacci(3)              ← RECALCULADO!
    ├── fibonacci(2)          ← RECALCULADO!
    │   ├── fibonacci(1) → 1
    │   └── fibonacci(0) → 0
    └── fibonacci(1) → 1
```

### Estatísticas:
* **F(5)**: 15 chamadas recursivas
* **F(10)**: 177 chamadas recursivas
* **F(20)**: 21.891 chamadas recursivas
* **F(35)**: 29.860.703 chamadas recursivas! 🔥

**Complexidade:** O(2^n) - exponencial!

````python
# Vamos contar quantas chamadas são feitas
chamadas = 0

def fibonacci_com_contador(n):
    """Fibonacci recursivo que conta as chamadas."""
    global chamadas
    chamadas += 1
    
    if n <= 1:
        return n
    return fibonacci_com_contador(n - 1) + fibonacci_com_contador(n - 2)

print("=" * 50)
print("DEMONSTRAÇÃO: Cálculos Duplicados")
print("=" * 50)

for num in [5, 10, 15, 20]:
    chamadas = 0
    resultado = fibonacci_com_contador(num)
    print(f"F({num}) = {resultado:>5} | Chamadas: {chamadas:>8}")

print("\n⚠️ Note como o número de chamadas cresce exponencialmente!")
````

## ✅ Solução 1: Abordagem Iterativa

Em vez de recursão, podemos usar um loop para calcular a sequência de baixo para cima.

**Vantagens:**
* Tempo: O(n) - linear
* Espaço: O(1) - constante
* Não há risco de RecursionError
* Muito mais rápido!

````python
def fibonacci_iterativo(n):
    """
    Calcula o n-ésimo número de Fibonacci usando iteração.
    
    Complexidade: O(n) tempo, O(1) espaço
    
    Parâmetros:
        n (int): Posição na sequência (n >= 0)
    
    Retorna:
        int: N-ésimo número de Fibonacci
    """
    # Casos base
    if n <= 1:
        return n
    
    # Inicializar os dois primeiros números
    anterior = 0
    atual = 1
    
    # Calcular iterativamente
    for _ in range(2, n + 1):
        proximo = anterior + atual
        anterior = atual
        atual = proximo
    
    return atual

# Exemplos de uso
print("=" * 50)
print("FIBONACCI ITERATIVO")
print("=" * 50)

for i in [0, 1, 5, 10, 20, 30, 40]:
    print(f"F({i}) = {fibonacci_iterativo(i)}")
````

## ✅ Solução 2: Recursão com Memoization

Memoization é uma técnica que **armazena resultados já calculados** para evitar cálculos duplicados.

Python oferece o decorador `@lru_cache` que implementa memoization automaticamente.

**Vantagens:**
* Mantém a elegância da recursão
* Tempo: O(n) - linear
* Espaço: O(n) - para armazenar cache
* Cada valor é calculado apenas uma vez

````python
from functools import lru_cache

@lru_cache(maxsize=None)
def fibonacci_memoization(n):
    """
    Calcula o n-ésimo número de Fibonacci usando recursão com cache.
    
    O decorador @lru_cache armazena resultados já calculados.
    Complexidade: O(n) tempo, O(n) espaço
    
    Parâmetros:
        n (int): Posição na sequência (n >= 0)
    
    Retorna:
        int: N-ésimo número de Fibonacci
    """
    # Casos base
    if n <= 1:
        return n
    
    # Caso recursivo - resultados são armazenados automaticamente
    return fibonacci_memoization(n - 1) + fibonacci_memoization(n - 2)

# Exemplos de uso
print("=" * 50)
print("FIBONACCI COM MEMOIZATION")
print("=" * 50)

for i in [0, 1, 5, 10, 20, 30, 40, 50]:
    print(f"F({i}) = {fibonacci_memoization(i)}")

# Informações sobre o cache
print(f"\nCache info: {fibonacci_memoization.cache_info()}")
````

## ⚡ Comparação de Performance

Vamos medir o tempo de execução de cada abordagem.

````python
import time

def medir_tempo(funcao, n):
    """Mede o tempo de execução de uma função."""
    inicio = time.time()
    resultado = funcao(n)
    fim = time.time()
    tempo = fim - inicio
    return resultado, tempo

print("=" * 70)
print("COMPARAÇÃO DE PERFORMANCE")
print("=" * 70)

# Testar com n = 30 (recursivo simples fica lento)
n_teste = 30

print(f"\nCalculando F({n_teste}):\n")

# 1. Recursivo simples (LENTO!)
res1, tempo1 = medir_tempo(fibonacci_recursivo, n_teste)
print(f"1. Recursivo simples:   {res1:>10} | Tempo: {tempo1:.6f}s")

# 2. Iterativo (RÁPIDO!)
res2, tempo2 = medir_tempo(fibonacci_iterativo, n_teste)
print(f"2. Iterativo:           {res2:>10} | Tempo: {tempo2:.6f}s")

# 3. Memoization (RÁPIDO!)
# Limpar cache antes de medir
fibonacci_memoization.cache_clear()
res3, tempo3 = medir_tempo(fibonacci_memoization, n_teste)
print(f"3. Recursão + Cache:    {res3:>10} | Tempo: {tempo3:.6f}s")

# Speedup
print(f"\n📊 Speedup:")
print(f"   Iterativo é {tempo1/tempo2:.0f}x mais rápido que recursivo simples")
print(f"   Memoization é {tempo1/tempo3:.0f}x mais rápido que recursivo simples")
````

## 💡 Boas Práticas e Recomendações

### Quando usar cada abordagem?

| Abordagem | Quando usar | Quando NÃO usar |
|-----------|-------------|----------------|
| **Recursivo Simples** | Nunca para Fibonacci! | Valores n > 20 |
| **Iterativo** | Produção, valores grandes | - |
| **Memoization** | Quando recursão é necessária | Espaço limitado |

### 🎯 Recomendações:

1. **Para Fibonacci: prefira iterativo**
   - Mais rápido e usa menos memória
   - Simples de entender e manter
   - Sem risco de RecursionError

2. **Use memoization quando:**
   - A recursão é natural ao problema
   - Há muitos subproblemas sobrepostos
   - Exemplo: programação dinâmica

3. **Evite recursão simples para Fibonacci:**
   - É um exemplo clássico de **quando NÃO usar recursão**
   - Apenas útil para fins educacionais

### ⚠️ Validação de entrada:

Em código de produção, sempre valide:
* `n` deve ser inteiro não-negativo
* Trate valores inválidos com exceções apropriadas
* Documente limitações (ex: RecursionError para n > 1000)

````python
def fibonacci(n, metodo='iterativo'):
    """
    Calcula o n-ésimo número de Fibonacci com validação.
    
    Parâmetros:
        n (int): Posição na sequência (n >= 0)
        metodo (str): 'iterativo' ou 'recursivo'
    
    Retorna:
        int: N-ésimo número de Fibonacci
    
    Raises:
        ValueError: Se n for negativo ou não for inteiro
    """
    # Validação
    if not isinstance(n, int):
        raise ValueError("n deve ser um número inteiro")
    
    if n < 0:
        raise ValueError("n deve ser não-negativo")
    
    # Escolher método
    if metodo == 'iterativo':
        return fibonacci_iterativo(n)
    elif metodo == 'recursivo':
        if n > 35:
            raise ValueError("n muito grande para recursão simples. Use metodo='iterativo'.")
        return fibonacci_recursivo(n)
    else:
        raise ValueError("metodo deve ser 'iterativo' ou 'recursivo'")

# Testes
print("=" * 50)
print("FUNÇÃO COM VALIDAÇÃO")
print("=" * 50)

try:
    print(f"F(10) = {fibonacci(10)}")
    print(f"F(50) = {fibonacci(50)}")
except ValueError as e:
    print(f"Erro: {e}")

try:
    print(f"\nF(-5) = {fibonacci(-5)}")
except ValueError as e:
    print(f"Erro: {e}")

try:
    print(f"\nF(100, 'recursivo') = {fibonacci(100, 'recursivo')}")
except ValueError as e:
    print(f"Erro: {e}")
````
