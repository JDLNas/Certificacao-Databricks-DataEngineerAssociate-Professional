# Estruturas Condicionais em Python

Este notebook cobre as estruturas condicionais que permitem executar código baseado em condições:

* **if** - Executa código se uma condição for verdadeira
* **if-else** - Executa um bloco ou outro dependendo da condição
* **if-elif-else** - Testa múltiplas condições em sequência
* **Operadores lógicos** - Combina condições (and, or, not)
* **Condicionais aninhados** - if dentro de if
* **Operador ternário** - Condicional em uma linha

Estruturas condicionais são fundamentais para:
* Tomada de decisões no código
* Validações e verificações
* Controle de fluxo do programa
* Lógica de negócios

## 1. Estrutura if Simples

O `if` executa um bloco de código **somente se** a condição for verdadeira (`True`).

**Sintaxe:**
```python
if condicao:
    # código executado se condição for True
    comando1
    comando2
```

**Importante:**
* A identação (4 espaços ou 1 tab) é **obrigatória** em Python
* O sinal `:` após a condição é **obrigatório**
* Se a condição for `False`, o bloco é ignorado

````python
# Exemplo 1: Verificando idade
idade = 20

if idade >= 18:
    print("Você é maior de idade")
    print("✅ Pode votar!")

print(f"Idade verificada: {idade} anos\n")

# Exemplo 2: Verificando nota
nota = 8.5

if nota >= 7.0:
    print(f"Nota {nota}: Aprovado!")
    print("🎉 Parabéns!\n")

# Exemplo 3: Verificando estoque
estoque = 5
estoque_minimo = 10

if estoque < estoque_minimo:
    print(f"⚠️ Alerta: Estoque baixo!")
    print(f"Atual: {estoque}, Mínimo: {estoque_minimo}")
    print("Ação: Solicitar reposição\n")

# Exemplo 4: Condição falsa (bloco não executado)
temperatura = 15

if temperatura > 30:
    print("Está muito quente!")  # Não será executado

print(f"Temperatura atual: {temperatura}°C")
````

## 2. Estrutura if-else

O `if-else` permite executar um bloco de código quando a condição é verdadeira e **outro bloco** quando é falsa.

**Sintaxe:**
```python
if condicao:
    # executado se condição for True
    comando1
else:
    # executado se condição for False
    comando2
```

**Características:**
* Sempre executa um dos dois blocos
* Use quando há apenas duas possibilidades
* O `else` não tem condição (captura tudo que não foi True no if)

````python
# Exemplo 1: Par ou Ímpar
numero = 7

if numero % 2 == 0:
    print(f"{numero} é par")
else:
    print(f"{numero} é ímpar")

print()

# Exemplo 2: Maior ou menor de idade
idade = 16

if idade >= 18:
    print(f"Idade {idade}: Maior de idade")
    print("✅ Pode dirigir")
else:
    print(f"Idade {idade}: Menor de idade")
    print("❌ Não pode dirigir")

print()

# Exemplo 3: Aprovado ou reprovado
nota = 5.5
nota_minima = 7.0

if nota >= nota_minima:
    print(f"Nota {nota}: Aprovado! 🎉")
else:
    print(f"Nota {nota}: Reprovado")
    print(f"Faltaram {nota_minima - nota:.1f} pontos")

print()

# Exemplo 4: Verificando saldo
saldo = 150.00
valor_compra = 200.00

if saldo >= valor_compra:
    print(f"✅ Compra aprovada!")
    print(f"Saldo restante: R$ {saldo - valor_compra:.2f}")
else:
    print(f"❌ Saldo insuficiente")
    print(f"Saldo: R$ {saldo:.2f}, Valor: R$ {valor_compra:.2f}")
    print(f"Faltam: R$ {valor_compra - saldo:.2f}")
````

## 3. Estrutura if-elif-else

O `elif` (else if) permite testar **múltiplas condições** em sequência.

**Sintaxe:**
```python
if condicao1:
    # executado se condicao1 for True
    comando1
elif condicao2:
    # executado se condicao1 for False e condicao2 for True
    comando2
elif condicao3:
    # executado se condicao1 e condicao2 forem False e condicao3 for True
    comando3
else:
    # executado se todas as condições forem False
    comando4
```

**Características:**
* Pode ter quantos `elif` forem necessários
* Testa as condições **em ordem** até encontrar uma verdadeira
* Apenas **um bloco** é executado (o primeiro que for True)
* O `else` é opcional

````python
# Exemplo 1: Classificação de nota
nota = 8.5

if nota >= 9.0:
    print(f"Nota {nota}: Excelente! 🌟")
elif nota >= 7.0:
    print(f"Nota {nota}: Bom! 👍")
elif nota >= 5.0:
    print(f"Nota {nota}: Regular")
else:
    print(f"Nota {nota}: Insuficiente")

print()

# Exemplo 2: Faixa etária
idade = 35

if idade < 12:
    print(f"Idade {idade}: Criança")
elif idade < 18:
    print(f"Idade {idade}: Adolescente")
elif idade < 60:
    print(f"Idade {idade}: Adulto")
else:
    print(f"Idade {idade}: Idoso")

print()

# Exemplo 3: Sistema de aprovação (similar ao código original)
n1 = 7.0
n2 = 8.0
media = (n1 + n2) / 2
faltas = 3

print(f"Notas: {n1} e {n2}")
print(f"Média: {media}")
print(f"Faltas: {faltas}\n")

if faltas >= 10:
    print("❌ Resultado: Reprovado por falta!")
    print(f"Limite de faltas excedido ({faltas}/10)")
elif media >= 7.0:
    print("✅ Resultado: Aprovado!")
    print("🎉 Parabéns!")
elif media >= 5.0:
    print("⚠️ Resultado: Recuperação")
    print(f"Faltam {7.0 - media:.1f} pontos para aprovação")
else:
    print("❌ Resultado: Reprovado por nota")
    print(f"Nota insuficiente: {media}")

print()

# Exemplo 4: Classificação de IMC
peso = 75  # kg
altura = 1.75  # metros
imc = peso / (altura ** 2)

print(f"Peso: {peso}kg, Altura: {altura}m")
print(f"IMC: {imc:.1f}\n")

if imc < 18.5:
    print("Classificação: Abaixo do peso")
elif imc < 25:
    print("Classificação: Peso normal ✅")
elif imc < 30:
    print("Classificação: Sobrepeso")
else:
    print("Classificação: Obesidade")
````

## 4. Operadores Lógicos em Condicionais

Operadores lógicos permitem **combinar múltiplas condições**:

### Operador `and`
* Retorna `True` se **todas** as condições forem verdadeiras
* `condicao1 and condicao2` → ambas devem ser True

### Operador `or`
* Retorna `True` se **pelo menos uma** condição for verdadeira
* `condicao1 or condicao2` → uma ou ambas devem ser True

### Operador `not`
* Inverte o valor booleano
* `not condicao` → True vira False, False vira True

### Tabela Verdade
```
A     | B     | A and B | A or B | not A
------|-------|---------|--------|-------
True  | True  | True    | True   | False
True  | False | False   | True   | False
False | True  | False   | True   | True
False | False | False   | False  | True
```

````python
# Exemplo 1: Operador AND - Todas as condições devem ser True
idade = 25
tem_cnh = True

if idade >= 18 and tem_cnh:
    print("✅ Pode alugar carro")
    print(f"Idade: {idade}, CNH: {tem_cnh}")
else:
    print("❌ Não pode alugar carro")

print()

# Exemplo 2: Operador OR - Pelo menos uma condição deve ser True
dia_semana = "sábado"
feriado = False

if dia_semana == "sábado" or dia_semana == "domingo" or feriado:
    print("🎉 É fim de semana ou feriado!")
    print("Não precisa trabalhar")
else:
    print("💼 Dia de trabalho")

print()

# Exemplo 3: Operador NOT - Inverte a condição
esta_chovendo = False

if not esta_chovendo:
    print("☀️ Não está chovendo")
    print("Pode sair sem guarda-chuva")
else:
    print("☔ Está chovendo")

print()

# Exemplo 4: Combinando operadores
nota = 8.0
faltas = 5
participacao = True

if nota >= 7.0 and faltas < 10 and participacao:
    print("✅ Aprovado com todos os critérios!")
    print(f"Nota: {nota}, Faltas: {faltas}, Participação: {participacao}")
elif nota >= 7.0 and faltas < 10:
    print("✅ Aprovado (por nota e frequência)")
    print("⚠️ Melhorar participação")
else:
    print("❌ Reprovado")

print()

# Exemplo 5: Intervalo de valores com AND
temperatura = 22

if temperatura >= 18 and temperatura <= 25:
    print(f"Temperatura {temperatura}°C: Confortável! 🌡️")
elif temperatura < 18:
    print(f"Temperatura {temperatura}°C: Frio! ❄️")
else:
    print(f"Temperatura {temperatura}°C: Quente! 🔥")

print()

# Exemplo 6: Múltiplas condições com OR
forma_pagamento = "cartão"

if forma_pagamento == "dinheiro" or forma_pagamento == "cartão" or forma_pagamento == "pix":
    print(f"✅ Forma de pagamento aceita: {forma_pagamento}")
else:
    print(f"❌ Forma de pagamento não aceita: {forma_pagamento}")
````

## 5. Condicionais Aninhados

Condicionais aninhados são estruturas `if` **dentro de outras** estruturas `if`.

**Sintaxe:**
```python
if condicao1:
    if condicao2:
        # executado se ambas forem True
        comando
    else:
        # executado se condicao1 True e condicao2 False
        comando
else:
    # executado se condicao1 False
    comando
```

**Quando usar:**
* Quando a segunda condição só faz sentido se a primeira for verdadeira
* Para decisões hierárquicas
* **Cuidado:** Muitos níveis de aninhamento dificultam a leitura

````python
# Exemplo 1: Verificação de idade e CNH
idade = 20
tem_cnh = True

if idade >= 18:
    print("Maior de idade")
    if tem_cnh:
        print("✅ Pode dirigir")
    else:
        print("❌ Não pode dirigir (sem CNH)")
else:
    print("Menor de idade")
    print("❌ Não pode dirigir (idade insuficiente)")

print()

# Exemplo 2: Sistema de desconto progressivo
valor_compra = 550.00
cliente_vip = True

print(f"Valor da compra: R$ {valor_compra:.2f}")
print(f"Cliente VIP: {cliente_vip}\n")

if valor_compra >= 500:
    if cliente_vip:
        desconto = 0.20  # 20% para VIP acima de 500
        print("Desconto VIP especial: 20%")
    else:
        desconto = 0.15  # 15% para não-VIP acima de 500
        print("Desconto padrão: 15%")
elif valor_compra >= 100:
    if cliente_vip:
        desconto = 0.10  # 10% para VIP acima de 100
        print("Desconto VIP: 10%")
    else:
        desconto = 0.05  # 5% para não-VIP acima de 100
        print("Desconto básico: 5%")
else:
    desconto = 0
    print("Sem desconto")

valor_final = valor_compra * (1 - desconto)
print(f"Valor final: R$ {valor_final:.2f}")
print(f"Economia: R$ {valor_compra - valor_final:.2f}")

print()

# Exemplo 3: Aprovação de empréstimo
salario = 5000.00
tem_restricao = False
tempo_emprego = 2  # anos

print(f"Salário: R$ {salario:.2f}")
print(f"Tempo de emprego: {tempo_emprego} anos")
print(f"Restrição de crédito: {tem_restricao}\n")

if salario >= 3000:
    if not tem_restricao:
        if tempo_emprego >= 1:
            print("✅ Empréstimo APROVADO")
            print("Todos os critérios atendidos")
        else:
            print("❌ Empréstimo NEGADO")
            print("Motivo: Tempo de emprego insuficiente")
    else:
        print("❌ Empréstimo NEGADO")
        print("Motivo: Restrição de crédito")
else:
    print("❌ Empréstimo NEGADO")
    print("Motivo: Salário insuficiente")
````

## 6. Operador Ternário (Condicional Inline)

O operador ternário permite escrever uma **condicional simples em uma única linha**.

**Sintaxe:**
```python
variavel = valor_se_true if condicao else valor_se_false
```

**Equivalente a:**
```python
if condicao:
    variavel = valor_se_true
else:
    variavel = valor_se_false
```

**Quando usar:**
* Para atribuições simples baseadas em condições
* Quando há apenas dois resultados possíveis
* Para código mais conciso e legível

**Evite:**
* Operadores ternários aninhados (dificulta leitura)
* Lógica complexa (use if-else tradicional)

````python
# Exemplo 1: Par ou ímpar
numero = 8
resultado = "par" if numero % 2 == 0 else "ímpar"
print(f"{numero} é {resultado}")
print()

# Exemplo 2: Maior ou menor de idade
idade = 17
status = "maior de idade" if idade >= 18 else "menor de idade"
print(f"Idade {idade}: {status}")
print()

# Exemplo 3: Aprovado ou reprovado
nota = 7.5
situacao = "Aprovado" if nota >= 7.0 else "Reprovado"
print(f"Nota {nota}: {situacao}")
print()

# Exemplo 4: Comparando com if-else tradicional
# Forma tradicional
temperatura = 28
if temperatura > 25:
    clima = "quente"
else:
    clima = "frio"
print(f"Forma tradicional - Temperatura {temperatura}°C: {clima}")

# Forma ternária (mais concisa)
clima = "quente" if temperatura > 25 else "frio"
print(f"Forma ternária - Temperatura {temperatura}°C: {clima}")
print()

# Exemplo 5: Definindo valor padrão
nome_usuario = ""  # vazio
nome_exibir = nome_usuario if nome_usuario else "Usuário Anônimo"
print(f"Nome: {nome_exibir}")

nome_usuario = "João"
nome_exibir = nome_usuario if nome_usuario else "Usuário Anônimo"
print(f"Nome: {nome_exibir}")
print()

# Exemplo 6: Cálculo de desconto
valor = 120.00
tem_cupom = True
desconto = 10 if tem_cupom else 0
valor_final = valor * (1 - desconto/100)

print(f"Valor original: R$ {valor:.2f}")
print(f"Cupom: {'Sim' if tem_cupom else 'Não'}")
print(f"Desconto: {desconto}%")
print(f"Valor final: R$ {valor_final:.2f}")
print()

# Exemplo 7: Mensagem personalizada
hora = 14
mensagem = "Bom dia" if hora < 12 else "Boa tarde" if hora < 18 else "Boa noite"
print(f"Hora {hora}h: {mensagem}")
print("Nota: Ternários aninhados (como acima) são válidos mas podem dificultar leitura")
````

## 7. Exemplo Consolidado: Sistema Completo de Avaliação

Vamos criar um sistema completo que combina várias estruturas condicionais.

````python
# Sistema de Avaliação de Aluno
# Combina várias estruturas condicionais

# Dados do aluno
aluno = "Maria Silva"
n1 = 8.0
n2 = 7.5
n3 = 9.0
faltas = 3
max_faltas = 10
participacao_ativa = True

# Cálculo da média
media = (n1 + n2 + n3) / 3

print("=" * 60)
print("SISTEMA DE AVALIAÇÃO ESCOLAR")
print("=" * 60)
print(f"\nAluno: {aluno}")
print(f"Notas: {n1}, {n2}, {n3}")
print(f"Média: {media:.2f}")
print(f"Faltas: {faltas}/{max_faltas}")
print(f"Participação Ativa: {'Sim' if participacao_ativa else 'Não'}")
print("\n" + "-" * 60)

# Verificação de faltas (prioridade máxima)
if faltas >= max_faltas:
    resultado = "REPROVADO POR FALTA"
    emoji = "❌"
    mensagem = f"Limite de faltas excedido ({faltas}/{max_faltas})"
    bonus = 0
    
# Verificação de nota com participação
elif media >= 7.0:
    resultado = "APROVADO"
    emoji = "✅"
    
    # Verifica bônus por participação (condicional aninhado)
    if participacao_ativa and media >= 9.0:
        bonus = 0.5
        mensagem = "Excelente desempenho! Bônus de 0.5 pontos"
    elif participacao_ativa:
        bonus = 0.3
        mensagem = "Bom desempenho! Bônus de 0.3 pontos"
    else:
        bonus = 0
        mensagem = "Aprovado sem bônus"
    
    media_final = media + bonus
    
elif media >= 5.0:
    resultado = "RECUPERAÇÃO"
    emoji = "⚠️"
    mensagem = f"Necessário {7.0 - media:.2f} pontos na prova de recuperação"
    bonus = 0
    media_final = media
    
else:
    resultado = "REPROVADO POR NOTA"
    emoji = "❌"
    mensagem = f"Média insuficiente: {media:.2f} (mínimo 5.0)"
    bonus = 0
    media_final = media

# Exibição dos resultados
print(f"\nRESULTADO: {emoji} {resultado}")
print(mensagem)

if bonus > 0:
    print(f"\nMédia Original: {media:.2f}")
    print(f"Bônus: +{bonus:.2f}")
    print(f"Média Final: {media_final:.2f}")

# Mensagem final com operador ternário
mensagem_final = "🎉 Parabéns!" if resultado == "APROVADO" else "📚 Continue estudando!"
print(f"\n{mensagem_final}")
print("\n" + "=" * 60)
````

## Boas Práticas com Estruturas Condicionais

### 1. Identação Consistente
* Use **4 espaços** (padrão Python) ou **1 tab** consistentemente
* Nunca misture espaços e tabs
* A identação é **obrigatória** e define o escopo do bloco

### 2. Condições Claras e Legíveis
```python
# ❌ EVITE - Condição confusa
if not (x < 10 and y > 5 or z != 0):
    pass

# ✅ PREFIRA - Condição clara
esta_no_intervalo = x >= 10
y_valido = y <= 5
z_zero = z == 0

if not esta_no_intervalo and (y_valido or z_zero):
    pass
```

### 3. Ordem das Condições
* Coloque as condições **mais específicas primeiro**
* Teste **casos mais prováveis** antes (otimização)
* Condições **mutuamente exclusivas** facilitam compreensão

### 4. Evite Aninhamento Excessivo
```python
# ❌ EVITE - Muitos níveis de aninhamento
if a:
    if b:
        if c:
            if d:
                print("Difícil de ler")

# ✅ PREFIRA - Use operadores lógicos
if a and b and c and d:
    print("Mais legível")

# ✅ OU - Early return/continue
if not a:
    return
if not b:
    return
if not c:
    return
print("Todas as condições atendidas")
```

### 5. Use elif em vez de if repetido
```python
# ❌ EVITE - Múltiplos if
if nota >= 9:
    print("A")
if nota >= 7 and nota < 9:
    print("B")
if nota < 7:
    print("C")

# ✅ PREFIRA - elif
if nota >= 9:
    print("A")
elif nota >= 7:
    print("B")
else:
    print("C")
```

### 6. Valores Booleanos
```python
# ❌ DESNECESSÁRIO
if ativo == True:
    pass
if lista == []:
    pass

# ✅ MAIS PYTHONICA
if ativo:
    pass
if not lista:  # lista vazia é False
    pass
```

### 7. Operador Ternário com Moderação
```python
# ✅ BOM - Simples e claro
status = "ativo" if habilitado else "inativo"

# ❌ EVITE - Ternário complexo/aninhado
resultado = "A" if x > 90 else "B" if x > 80 else "C" if x > 70 else "D"
# Prefira if-elif-else tradicional para casos complexos
```

### 8. Comentários em Condições Complexas
* Documente **por que** a condição existe, não o que ela faz
* Use comentários para explicar lógica de negócio
* Condições simples raramente precisam de comentários

### 9. Validações no Início (Guard Clauses)
```python
# ✅ Validar e retornar cedo
def processar_pedido(pedido):
    if not pedido:
        return "Pedido inválido"
    if pedido.valor <= 0:
        return "Valor inválido"
    if not pedido.cliente:
        return "Cliente não informado"
    
    # Lógica principal aqui
    return "Pedido processado"
```

### 10. Constantes para Valores Mágicos
```python
# ❌ EVITE - Números mágicos
if idade >= 18:
    pass
if idade >= 65:
    pass

# ✅ PREFIRA - Constantes nomeadas
IDADE_MAIOR = 18
IDADE_IDOSO = 65

if idade >= IDADE_MAIOR:
    pass
if idade >= IDADE_IDOSO:
    pass
```
