# Equações de Segundo Grau - Cálculo de Raízes

Este notebook demonstra como calcular as raízes de equações de segundo grau usando Python.

**Objetivo:** Implementar e entender o cálculo de raízes usando a fórmula de Bhaskara.

**Tópicos abordados:**
* Conceitos matemáticos fundamentais
* Discriminante (Delta)
* Implementação em Python
* Tratamento de casos especiais
* Exemplos práticos

## Conceitos Matemáticos

Uma equação de segundo grau (ou equação quadrática) tem a forma geral:

```
ax² + bx + c = 0
```

Onde:
* `a`, `b` e `c` são coeficientes (com a ≠ 0)
* `x` é a variável que queremos encontrar

### Fórmula de Bhaskara

As raízes são calculadas pela fórmula:

```
x = (-b ± √Δ) / 2a
```

Onde Δ (delta) é o **discriminante**:

```
Δ = b² - 4ac
```

## Análise do Discriminante (Delta)

O valor de Δ determina quantas e quais raízes existem:

| Condição | Resultado |
|----------|----------|
| Δ > 0 | Duas raízes reais e diferentes |
| Δ = 0 | Duas raízes reais e iguais |
| Δ < 0 | Não há raízes reais (raízes complexas) |

````python
def calcular_raizes(a, b, c):
    """
    Calcula as raízes de uma equação de segundo grau.
    
    Parâmetros:
        a (float): Coeficiente de x²
        b (float): Coeficiente de x
        c (float): Termo independente
    
    Retorna:
        tuple: (x1, x2) se existirem raízes reais, ou (None, None) caso contrário
    """
    # Calcular o discriminante
    delta = b**2 - 4*a*c
    
    # Verificar se existem raízes reais
    if delta < 0:
        print(f"Equação: {a}x² + {b}x + {c} = 0")
        print(f"Delta = {delta} (negativo)")
        print("Não existem raízes reais\n")
        return None, None
    
    # Calcular as raízes usando a fórmula de Bhaskara
    raiz_delta = delta ** 0.5
    x1 = (-b + raiz_delta) / (2 * a)
    x2 = (-b - raiz_delta) / (2 * a)
    
    # Exibir resultados
    print(f"Equação: {a}x² + {b}x + {c} = 0")
    print(f"Delta = {delta}")
    print(f"Raiz x₁ = {x1:.4f}")
    print(f"Raiz x₂ = {x2:.4f}")
    
    # Indicar o tipo de raízes
    if delta == 0:
        print("(Raízes iguais)\n")
    else:
        print("(Raízes diferentes)\n")
    
    return x1, x2
````

## Exemplos Práticos

Vamos testar a função com diferentes tipos de equações para demonstrar todos os casos possíveis.

````python
# Exemplo 1: Duas raízes reais diferentes (Δ > 0)
print("=" * 50)
print("EXEMPLO 1: Duas raízes reais diferentes")
print("=" * 50)

# Equação: x² - 5x + 6 = 0
# Esperado: x₁ = 3 e x₂ = 2
x1, x2 = calcular_raizes(1, -5, 6)
````

````python
# Exemplo 2: Duas raízes reais iguais (Δ = 0)
print("=" * 50)
print("EXEMPLO 2: Duas raízes reais iguais")
print("=" * 50)

# Equação: x² - 4x + 4 = 0
# Esperado: x₁ = x₂ = 2
x1, x2 = calcular_raizes(1, -4, 4)
````

````python
# Exemplo 3: Sem raízes reais (Δ < 0)
print("=" * 50)
print("EXEMPLO 3: Sem raízes reais")
print("=" * 50)

# Equação: x² + 2x + 5 = 0
# Esperado: Nenhuma raiz real
x1, x2 = calcular_raizes(1, 2, 5)
````

````python
# Exemplo 4: Equação com coeficientes maiores
print("=" * 50)
print("EXEMPLO 4: Coeficientes maiores")
print("=" * 50)

# Equação: 2x² - 8x + 6 = 0
# Esperado: x₁ = 3 e x₂ = 1
x1, x2 = calcular_raizes(2, -8, 6)
````

## Boas Práticas

### Validação de Entrada

Em aplicações reais, é importante validar os dados de entrada:

* **Verificar se `a ≠ 0`**: Se a = 0, não é uma equação de segundo grau
* **Validar tipos**: Garantir que a, b, c sejam números
* **Tratamento de exceções**: Usar try/except para capturar erros

### Extensões Possíveis

* Adicionar suporte para números complexos quando Δ < 0
* Implementar visualização gráfica da parábola
* Criar interface interativa para entrada de valores
* Adicionar testes unitários para validação

````python
def calcular_raizes_validado(a, b, c):
    """
    Versão melhorada com validação de entrada.
    """
    # Validar se a é diferente de zero
    if a == 0:
        raise ValueError("O coeficiente 'a' não pode ser zero em uma equação de segundo grau")
    
    # Validar tipos
    try:
        a, b, c = float(a), float(b), float(c)
    except (ValueError, TypeError):
        raise ValueError("Os coeficientes devem ser números")
    
    # Calcular usando a função original
    return calcular_raizes(a, b, c)

# Teste da função validada
print("=" * 50)
print("EXEMPLO COM VALIDAÇÃO")
print("=" * 50)

try:
    calcular_raizes_validado(1, -7, 10)
except ValueError as e:
    print(f"Erro: {e}")
````
