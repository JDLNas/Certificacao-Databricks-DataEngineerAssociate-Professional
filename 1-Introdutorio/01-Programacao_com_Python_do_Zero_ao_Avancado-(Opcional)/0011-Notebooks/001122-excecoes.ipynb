## 1️⃣ Conceitos Básicos de Exceções

### 📖 Conceito
**Exceções** são eventos que interrompem o fluxo normal de execução quando ocorre um erro. Sem tratamento, exceções fazem o programa terminar abruptamente. Com tratamento adequado, podemos detectar erros e tomar ações apropriadas.

### 🔧 Estrutura Try/Except
```python
try:
    # Código que pode lançar exceção
    resultado = 10 / 0
except ZeroDivisionError:
    # Código executado se exceção ocorrer
    print("Erro: divisão por zero")
```

### 📊 Exceções Comuns
* **ZeroDivisionError**: Divisão por zero
* **ValueError**: Valor inadequado (ex: int("abc"))
* **TypeError**: Tipo incorreto para operação
* **KeyError**: Chave não existe em dicionário
* **IndexError**: Índice fora do intervalo de lista
* **FileNotFoundError**: Arquivo não encontrado
* **AttributeError**: Atributo não existe

### ✅ Por Que Tratar Exceções?
* **Robustez**: Programa não trava inesperadamente
* **Experiência do usuário**: Mensagens claras de erro
* **Recuperação**: Possibilidade de continuar execução
* **Logging**: Registrar erros para diagnóstico
* **Segurança**: Evitar exposição de informações internas

### ⚠️ O Que Acontece Sem Tratamento
```python
# Sem try/except
resultado = 10 / 0  # Programa termina aqui
print("Nunca executado")  # Esta linha nunca roda
```

````python
# Exemplo 1: Programa sem tratamento de exceção ❌
def dividir_sem_tratamento(a, b):
    """Divisão sem proteção - pode falhar."""
    return a / b

print("❌ Sem tratamento de exceção:")
try:
    resultado = dividir_sem_tratamento(10, 0)
    print(f"   Resultado: {resultado}")
except ZeroDivisionError:
    print("   ⚠️ Erro capturado: divisão por zero")
print()

# Exemplo 2: Programa com tratamento de exceção ✅
def dividir_com_tratamento(a, b):
    """Divisão com proteção contra erro."""
    try:
        resultado = a / b
        return resultado
    except ZeroDivisionError:
        return "❌ Erro: divisão por zero"

print("✅ Com tratamento de exceção:")
print(f"   10 / 2 = {dividir_com_tratamento(10, 2)}")
print(f"   10 / 0 = {dividir_com_tratamento(10, 0)}")
print(f"   Programa continua normalmente!")
print()

# Exemplo 3: ValueError com conversão de tipo 🔢
def converter_para_inteiro(texto):
    """Converte texto para inteiro com tratamento."""
    try:
        numero = int(texto)
        return numero
    except ValueError:
        return f"❌ '{texto}' não é um número válido"

print("🔢 Conversão de string para int:")
entradas = ["42", "abc", "3.14", "100"]
for entrada in entradas:
    resultado = converter_para_inteiro(entrada)
    print(f"   '{entrada}' → {resultado}")
print()

# Exemplo 4: KeyError com dicionários 🔑
produtos = {
    "notebook": 3500,
    "mouse": 50,
    "teclado": 250
}

def buscar_preco(produto):
    """Busca preço de produto com tratamento."""
    try:
        preco = produtos[produto]
        return f"R$ {preco:.2f}"
    except KeyError:
        return f"❌ Produto '{produto}' não encontrado"

print("📦 Buscando preços:")
buscas = ["notebook", "monitor", "mouse", "webcam"]
for item in buscas:
    resultado = buscar_preco(item)
    print(f"   {item}: {resultado}")
print()

# Exemplo 5: IndexError com listas 📊
numeros = [10, 20, 30, 40, 50]

def acessar_indice(lista, indice):
    """Acessa elemento por índice com proteção."""
    try:
        valor = lista[indice]
        return f"✅ Valor: {valor}"
    except IndexError:
        return f"❌ Índice {indice} fora do intervalo (0-{len(lista)-1})"

print(f"📊 Lista: {numeros}")
print("Acessando índices:")
for idx in [0, 2, 5, 10]:
    resultado = acessar_indice(numeros, idx)
    print(f"   Índice {idx}: {resultado}")
print()

# Exemplo 6: TypeError com operações 🔄
def somar_valores(a, b):
    """Soma valores com verificação de tipo."""
    try:
        resultado = a + b
        return resultado
    except TypeError:
        return f"❌ Não é possível somar {type(a).__name__} com {type(b).__name__}"

print("➕ Testes de soma:")
testes = [(10, 5), ("Python", " rocks"), (10, "5"), ([1,2], [3,4])]
for a, b in testes:
    resultado = somar_valores(a, b)
    print(f"   {repr(a)} + {repr(b)} = {resultado}")
print()

# Exemplo 7: Capturando objeto de exceção 📝
def processar_arquivo(nome_arquivo):
    """Simula processamento de arquivo."""
    try:
        # Simula erro de arquivo
        if nome_arquivo != "dados.txt":
            raise FileNotFoundError(f"Arquivo '{nome_arquivo}' não encontrado")
        return "✅ Arquivo processado com sucesso"
    except FileNotFoundError as e:
        # Captura objeto de exceção para acessar detalhes
        return f"❌ Erro: {e}"

print("💾 Processando arquivos:")
arquivos = ["dados.txt", "config.json", "relatorio.pdf"]
for arquivo in arquivos:
    resultado = processar_arquivo(arquivo)
    print(f"   {arquivo}: {resultado}")
````

## 2️⃣ Múltiplas Exceções

### 📖 Conceito
Código pode gerar diferentes tipos de exceções. Podemos capturá-las separadamente para fornecer tratamento específico para cada tipo de erro.

### 🔧 Capturando Múltiplas Exceções

**Blocos except separados:**
```python
try:
    resultado = operacao()
except ValueError:
    print("Erro de valor")
except TypeError:
    print("Erro de tipo")
except ZeroDivisionError:
    print("Divisão por zero")
```

**Múltiplas exceções em uma linha:**
```python
try:
    resultado = operacao()
except (ValueError, TypeError):
    print("Erro de valor ou tipo")
```

**Captura genérica (cuidado!):**
```python
try:
    resultado = operacao()
except Exception as e:
    print(f"Erro genérico: {e}")
```

### 📊 Ordem de Captura
* **Específica primeiro**: Coloque exceções mais específicas antes
* **Genérica por último**: Exception captura tudo
* **Hierarquia importa**: Subclasses antes de superclasses

### ✅ Boas Práticas
* Capture exceções específicas sempre que possível
* Evite `except:` vazio (captura tudo, inclusive Ctrl+C)
* Use `except Exception:` para genérico, não `except:`
* Fornecer mensagens de erro claras

### ⚠️ Cuidados
```python
# ❌ RUIM - Ordem errada
try:
    ...
except Exception:  # Captura tudo primeiro
    ...
except ValueError:  # Nunca executado!
    ...

# ✅ BOM - Ordem correta
try:
    ...
except ValueError:  # Específico primeiro
    ...
except Exception:   # Genérico por último
    ...
```

````python
# Exemplo 1: Múltiplas exceções separadas 📊
def calcular(operacao, a, b):
    """Executa operação matemática com tratamento."""
    try:
        if operacao == "+":
            return a + b
        elif operacao == "-":
            return a - b
        elif operacao == "*":
            return a * b
        elif operacao == "/":
            return a / b
        else:
            raise ValueError(f"Operação '{operacao}' inválida")
    except ZeroDivisionError:
        return "❌ Erro: divisão por zero"
    except TypeError:
        return f"❌ Erro: tipos incompatíveis ({type(a).__name__}, {type(b).__name__})"
    except ValueError as e:
        return f"❌ Erro: {e}"

print("🧮 Calculadora com múltiplas exceções:")
testes = [
    ("+", 10, 5),
    ("/", 10, 0),
    ("*", "5", 3),
    ("%", 10, 3)
]
for op, x, y in testes:
    resultado = calcular(op, x, y)
    print(f"   {x} {op} {y} = {resultado}")
print()

# Exemplo 2: Capturando múltiplas exceções em uma linha 🔗
def processar_entrada(valor):
    """Processa entrada com várias validações."""
    try:
        # Tenta converter e processar
        numero = int(valor)
        resultado = 100 / numero
        return f"✅ Resultado: {resultado:.2f}"
    except (ValueError, TypeError):
        return f"❌ Entrada inválida: '{valor}' não é um número"
    except ZeroDivisionError:
        return "❌ Não é possível dividir por zero"

print("📝 Processando entradas:")
entradas = ["50", "0", "abc", "25", None]
for entrada in entradas:
    resultado = processar_entrada(entrada)
    print(f"   {repr(entrada):10} → {resultado}")
print()

# Exemplo 3: Hierarquia de exceções 🎯
def acessar_dados(dados, chave, indice):
    """Acessa dados com múltiplos pontos de falha."""
    try:
        lista = dados[chave]  # Pode lançar KeyError
        valor = lista[indice]  # Pode lançar IndexError
        return f"✅ Valor: {valor}"
    except KeyError:
        return f"❌ Chave '{chave}' não encontrada"
    except IndexError:
        return f"❌ Índice {indice} fora do intervalo"
    except TypeError:
        return "❌ Tipo de dado incorreto"

dados = {
    "vendas": [100, 200, 300],
    "custos": [50, 75, 90]
}

print("📋 Acessando dados estruturados:")
testes = [
    ("vendas", 1),
    ("lucros", 0),
    ("vendas", 10),
    ("custos", 2)
]
for chave, idx in testes:
    resultado = acessar_dados(dados, chave, idx)
    print(f"   dados['{chave}'][{idx}]: {resultado}")
print()

# Exemplo 4: Exceção genérica como fallback 🛡️
def operacao_complexa(x, y, z):
    """Operação com tratamento específico e genérico."""
    try:
        resultado = (x + y) / z
        return f"✅ {resultado:.2f}"
    except ZeroDivisionError:
        return "❌ Divisão por zero"
    except TypeError:
        return "❌ Tipos incompatíveis"
    except Exception as e:
        # Captura qualquer outra exceção
        return f"⚠️ Erro inesperado: {type(e).__name__}: {e}"

print("🛡️ Operações com fallback:")
testes = [
    (10, 5, 3),
    (10, 5, 0),
    ("10", 5, 3),
    (10, 5, "3")
]
for x, y, z in testes:
    resultado = operacao_complexa(x, y, z)
    print(f"   ({x} + {y}) / {z} = {resultado}")
print()

# Exemplo 5: Validando dados de usuário 👤
def criar_usuario(dados):
    """Cria usuário com múltiplas validações."""
    try:
        nome = dados["nome"]
        idade = int(dados["idade"])
        email = dados["email"]
        
        if idade < 0:
            raise ValueError("Idade não pode ser negativa")
        if "@" not in email:
            raise ValueError("Email inválido")
        
        return f"✅ Usuário '{nome}' criado com sucesso"
    
    except KeyError as e:
        return f"❌ Campo obrigatório ausente: {e}"
    except ValueError as e:
        return f"❌ Valor inválido: {e}"
    except TypeError:
        return "❌ Dados devem ser um dicionário"

print("👤 Criando usuários:")
usuarios = [
    {"nome": "Ana", "idade": "28", "email": "ana@email.com"},
    {"nome": "Bruno", "idade": "abc", "email": "bruno@email.com"},
    {"nome": "Carla", "email": "carla@email.com"},
    {"nome": "Daniel", "idade": "30", "email": "daniel.email.com"},
]
for i, user in enumerate(usuarios, 1):
    resultado = criar_usuario(user)
    print(f"   Usuário {i}: {resultado}")
print()

# Exemplo 6: Ordem correta de captura ✅❌
def demonstrar_ordem():
    """Demonstra importância da ordem de except."""
    valores = [10, 0, "abc"]
    
    for valor in valores:
        print(f"\n   Processando: {repr(valor)}")
        
        # ✅ Ordem correta
        try:
            resultado = 100 / int(valor)
            print(f"      ✅ Resultado: {resultado}")
        except ZeroDivisionError:
            print("      ❌ ZeroDivisionError capturado")
        except ValueError:
            print("      ❌ ValueError capturado")
        except Exception as e:
            print(f"      ⚠️ Outro erro: {type(e).__name__}")

print("🔢 Demonstração de ordem:")
demonstrar_ordem()
````

## 3️⃣ Cláusulas Else e Finally

### 📖 Conceito
**else** e **finally** complementam try/except, permitindo controle mais preciso do fluxo de execução.

### 🔧 Cláusula Else

Executa **apenas se nenhuma exceção** ocorrer:
```python
try:
    resultado = 10 / 2
except ZeroDivisionError:
    print("Erro")
else:
    print("Sucesso!")  # Executa se não houver erro
    print(f"Resultado: {resultado}")
```

### 🔧 Cláusula Finally

Executa **sempre**, ocorra ou não exceção:
```python
try:
    arquivo = open("dados.txt")
    processar(arquivo)
except FileNotFoundError:
    print("Arquivo não encontrado")
finally:
    arquivo.close()  # Sempre executa (limpeza)
```

### 📊 Estrutura Completa
```python
try:
    # Código principal
    resultado = operacao()
except TipoErro:
    # Tratamento de erro
    print("Erro")
else:
    # Executa se sucesso
    print("Sucesso")
finally:
    # Sempre executa (limpeza)
    print("Finalização")
```

### ✅ Quando Usar
* **else**: Código que só deve rodar em caso de sucesso
* **finally**: Limpeza de recursos (fechar arquivos, conexões)
* **finally**: Garantir execução mesmo com return/break

### ⚠️ Diferenças Importantes
* **else**: Roda apenas se **sem** exceção
* **finally**: Roda **sempre** (com ou sem exceção)
* finally executa mesmo se houver return no try

````python
# Exemplo 1: Usando else para código de sucesso ✅
def dividir_com_else(a, b):
    """Divisão com cláusula else."""
    try:
        resultado = a / b
    except ZeroDivisionError:
        print(f"   ❌ Erro: não é possível dividir {a} por zero")
        return None
    else:
        print(f"   ✅ Divisão bem-sucedida")
        return resultado

print("🔢 Usando else:")
print(f"\n10 / 2:")
r1 = dividir_com_else(10, 2)
if r1:
    print(f"   Resultado: {r1}")

print(f"\n10 / 0:")
r2 = dividir_com_else(10, 0)
print()

# Exemplo 2: Usando finally para limpeza 🧹
def processar_dados_simulado(arquivo, processar=True):
    """Simula processamento de arquivo com finally."""
    print(f"\n   📂 Abrindo arquivo '{arquivo}'")
    
    try:
        if not processar:
            raise ValueError("Erro durante processamento")
        print(f"   ⚙️ Processando dados...")
        resultado = "Dados processados"
        return resultado
    except ValueError as e:
        print(f"   ❌ Erro: {e}")
        return None
    else:
        print(f"   ✅ Processamento concluído")
    finally:
        print(f"   🔒 Fechando arquivo '{arquivo}'")

print("🧹 Usando finally:")
processar_dados_simulado("dados.csv", processar=True)
processar_dados_simulado("config.json", processar=False)
print()

# Exemplo 3: Estrutura completa (try/except/else/finally) 🎯
def executar_transacao(valor, saldo_suficiente=True):
    """Simula transação bancária."""
    print(f"\n   💳 Iniciando transação de R$ {valor}")
    
    try:
        if not saldo_suficiente:
            raise ValueError("Saldo insuficiente")
        if valor <= 0:
            raise ValueError("Valor inválido")
        
        print(f"   ⚙️ Processando...")
        
    except ValueError as e:
        print(f"   ❌ Erro: {e}")
        status = "FALHOU"
    
    else:
        print(f"   ✅ Transação aprovada")
        status = "SUCESSO"
    
    finally:
        print(f"   📋 Status final: {status}")
        print(f"   🔒 Conexão encerrada")

print("🎯 Estrutura completa:")
executar_transacao(100, saldo_suficiente=True)
executar_transacao(100, saldo_suficiente=False)
print()

# Exemplo 4: Finally executa mesmo com return 🔄
def funcao_com_return():
    """Demonstra finally executando mesmo com return."""
    print("   Entrando na função")
    
    try:
        print("   Executando try")
        return "Retorno do try"
    except Exception:
        print("   Executando except")
    else:
        print("   Executando else")
    finally:
        print("   🎯 Finally SEMPRE executa (mesmo com return!)")

print("🔄 Finally com return:")
resultado = funcao_com_return()
print(f"   Resultado retornado: {resultado}")
print()

# Exemplo 5: Contadores de sucesso/falha 📊
def processar_lote(valores):
    """Processa lote de valores e conta sucessos/falhas."""
    sucessos = 0
    falhas = 0
    
    for valor in valores:
        try:
            resultado = 100 / valor
        except ZeroDivisionError:
            falhas += 1
        except TypeError:
            falhas += 1
        else:
            sucessos += 1
        finally:
            # Finally executa em cada iteração
            pass
    
    return sucessos, falhas

print("📊 Processamento em lote:")
valores = [10, 5, 0, 20, "abc", 2]
sucessos, falhas = processar_lote(valores)
print(f"   Valores: {valores}")
print(f"   ✅ Sucessos: {sucessos}")
print(f"   ❌ Falhas: {falhas}")
print(f"   📊 Total: {len(valores)}")
print()

# Exemplo 6: Gerenciamento de recursos 💾
class ConexaoBancoDados:
    """Simula conexão com banco de dados."""
    def __init__(self, nome):
        self.nome = nome
        self.conectado = False
    
    def conectar(self):
        print(f"   🔗 Conectando a {self.nome}")
        self.conectado = True
    
    def desconectar(self):
        print(f"   🔒 Desconectando de {self.nome}")
        self.conectado = False

def executar_consulta(query, falhar=False):
    """Executa consulta com gerenciamento de conexão."""
    conexao = ConexaoBancoDados("PostgreSQL")
    
    try:
        conexao.conectar()
        
        if falhar:
            raise Exception("Erro na consulta")
        
        print(f"   ⚙️ Executando: {query}")
        resultado = ["Linha 1", "Linha 2", "Linha 3"]
        return resultado
    
    except Exception as e:
        print(f"   ❌ Erro: {e}")
        return None
    
    finally:
        # Garante que conexão é fechada
        if conexao.conectado:
            conexao.desconectar()

print("💾 Gerenciamento de recursos:")
print("\nConsulta com sucesso:")
executar_consulta("SELECT * FROM usuarios", falhar=False)
print("\nConsulta com falha:")
executar_consulta("SELECT * FROM tabela_invalida", falhar=True)
````

## 4️⃣ Levantando Exceções com raise

### 📖 Conceito
**raise** permite que você **levante** (lançar) exceções intencionalmente. É usado para sinalizar erros ou condições excepcionais no seu código.

### 🔧 Sintaxe do raise

**Levantar exceção específica:**
```python
if idade < 0:
    raise ValueError("Idade não pode ser negativa")
```

**Relevantar exceção capturada:**
```python
try:
    operacao()
except Exception as e:
    print(f"Erro capturado: {e}")
    raise  # Relevanta a mesma exceção
```

**Levantar com contexto:**
```python
try:
    operacao1()
except Exception as e:
    raise ValueError("Falha na operação") from e
```

### 📊 Quando Usar raise
* **Validação**: Dados inválidos ou fora do esperado
* **Estado inconsistente**: Condições que não deveriam ocorrer
* **Propagação**: Deixar chamador tratar o erro
* **Transformação**: Converter um tipo de erro em outro

### ✅ Boas Práticas
* Use tipo de exceção apropriado
* Forneça mensagem descritiva
* Valide no início da função (fail fast)
* Documente exceções que a função pode lançar

### 📊 Exceções Built-in Comuns
* **ValueError**: Valor inadequado
* **TypeError**: Tipo incorreto
* **RuntimeError**: Erro genérico em runtime
* **NotImplementedError**: Funcionalidade não implementada
* **AssertionError**: Assert falhou

````python
# Exemplo 1: Validação com raise ✅
def calcular_desconto(preco, percentual):
    """Calcula desconto com validação."""
    if preco < 0:
        raise ValueError("Preço não pode ser negativo")
    if percentual < 0 or percentual > 100:
        raise ValueError("Percentual deve estar entre 0 e 100")
    
    desconto = preco * (percentual / 100)
    return preco - desconto

print("✅ Validação com raise:")
testes = [(100, 10), (100, -5), (-50, 10), (100, 150)]
for preco, perc in testes:
    try:
        resultado = calcular_desconto(preco, perc)
        print(f"   Preço R$ {preco} com {perc}% desc: R$ {resultado:.2f}")
    except ValueError as e:
        print(f"   ❌ Preço R$ {preco} com {perc}%: {e}")
print()

# Exemplo 2: Validação de idade 🎂
def cadastrar_idade(idade):
    """Cadastra idade com validações."""
    if not isinstance(idade, int):
        raise TypeError("Idade deve ser um número inteiro")
    if idade < 0:
        raise ValueError("Idade não pode ser negativa")
    if idade > 150:
        raise ValueError("Idade inválida (muito alta)")
    
    return f"✅ Idade {idade} cadastrada"

print("🎂 Validação de idade:")
idades = [25, -5, 200, "30", 100]
for idade in idades:
    try:
        resultado = cadastrar_idade(idade)
        print(f"   {idade}: {resultado}")
    except (ValueError, TypeError) as e:
        print(f"   {idade}: ❌ {e}")
print()

# Exemplo 3: Relevantar exceção (raise sem argumentos) 🔄
def processar_com_log(dados):
    """Processa dados e loga erros antes de relevantar."""
    try:
        resultado = 100 / dados
        return resultado
    except ZeroDivisionError:
        print("   📋 LOG: ZeroDivisionError capturado")
        print("   📋 LOG: Salvando erro no banco de dados...")
        raise  # Relevanta a mesma exceção

print("🔄 Relevantar exceção:")
try:
    processar_com_log(0)
except ZeroDivisionError:
    print("   ❌ Exceção capturada no nível superior")
print()

# Exemplo 4: Validação de CPF 🆔
def validar_cpf(cpf):
    """Valida formato de CPF."""
    # Remove caracteres não numéricos
    cpf_limpo = ''.join(c for c in cpf if c.isdigit())
    
    if len(cpf_limpo) != 11:
        raise ValueError(f"CPF deve ter 11 dígitos (fornecido: {len(cpf_limpo)})")
    
    if cpf_limpo == cpf_limpo[0] * 11:
        raise ValueError("CPF não pode ter todos os dígitos iguais")
    
    return f"✅ CPF {cpf} válido"

print("🆔 Validação de CPF:")
cpfs = ["123.456.789-01", "111.111.111-11", "123.456", "987.654.321-00"]
for cpf in cpfs:
    try:
        resultado = validar_cpf(cpf)
        print(f"   {cpf}: {resultado}")
    except ValueError as e:
        print(f"   {cpf}: ❌ {e}")
print()

# Exemplo 5: Função com múltiplas validações 🛡️
def criar_conta(usuario, senha, email):
    """Cria conta com validações múltiplas."""
    if not usuario or len(usuario) < 3:
        raise ValueError("Usuário deve ter pelo menos 3 caracteres")
    
    if not senha or len(senha) < 8:
        raise ValueError("Senha deve ter pelo menos 8 caracteres")
    
    if "@" not in email or "." not in email:
        raise ValueError("Email inválido")
    
    return f"✅ Conta '{usuario}' criada com sucesso"

print("🛡️ Criação de conta:")
contas = [
    ("ana", "senha123", "ana@email.com"),
    ("ab", "senha123", "ab@email.com"),
    ("carlos", "12345", "carlos@email.com"),
    ("maria", "senhasegura", "mariaemail.com"),
]
for user, pwd, email in contas:
    try:
        resultado = criar_conta(user, pwd, email)
        print(f"   {resultado}")
    except ValueError as e:
        print(f"   ❌ {user}: {e}")
print()

# Exemplo 6: Transformando exceções 🔄
def processar_json(texto_json):
    """Processa JSON e transforma erro."""
    import json
    try:
        dados = json.loads(texto_json)
        return dados
    except json.JSONDecodeError as e:
        # Transforma JSONDecodeError em ValueError
        raise ValueError(f"Formato JSON inválido: {e}")

print("🔄 Transformando exceções:")
jsons = ['{"nome": "Ana"}', '{nome: Ana}', '[1, 2, 3]']
for texto in jsons:
    try:
        resultado = processar_json(texto)
        print(f"   ✅ JSON válido: {resultado}")
    except ValueError as e:
        print(f"   ❌ {e}")
print()

# Exemplo 7: NotImplementedError 🚧
class Animal:
    """Classe base com métodos abstratos."""
    def fazer_som(self):
        raise NotImplementedError("Subclasse deve implementar fazer_som()")

class Cachorro(Animal):
    def fazer_som(self):
        return "Au au!"

class Gato(Animal):
    pass  # Não implementou fazer_som()

print("🚧 NotImplementedError:")
cachorro = Cachorro()
print(f"   Cachorro: {cachorro.fazer_som()}")

gato = Gato()
try:
    print(f"   Gato: {gato.fazer_som()}")
except NotImplementedError as e:
    print(f"   ❌ Gato: {e}")
````

## 5️⃣ Exceções Personalizadas

### 📖 Conceito
Você pode criar suas próprias classes de exceção herdando de `Exception` ou suas subclasses. Exceções personalizadas tornam código mais expressivo e permitem tratamento específico de erros do seu domínio.

### 🔧 Criando Exceção Personalizada

**Simples:**
```python
class MeuErro(Exception):
    pass

raise MeuErro("Algo deu errado")
```

**Com construtor:**
```python
class SaldoInsuficienteError(Exception):
    def __init__(self, saldo, valor):
        self.saldo = saldo
        self.valor = valor
        super().__init__(f"Saldo R$ {saldo} < Valor R$ {valor}")
```

**Hierarquia:**
```python
class ErroAplicacao(Exception):
    """Erro base da aplicação."""
    pass

class ErroBancoDados(ErroAplicacao):
    """Erro relacionado ao banco."""
    pass

class ErroValidacao(ErroAplicacao):
    """Erro de validação."""
    pass
```

### ✅ Quando Criar Exceções Personalizadas
* Erros específicos do seu domínio
* Quando built-ins não são expressivas o suficiente
* Para captura seletiva em diferentes níveis
* Para adicionar atributos/comportamento extra

### 📊 Convenções
* Nome termina com "Error" ou "Exception"
* Herda de `Exception` ou subclasse apropriada
* Documente o propósito da exceção
* Mantenha hierarquia clara

````python
# Exemplo 1: Exceção personalizada simples ✨
class IdadeInvalidaError(Exception):
    """Levantada quando idade é inválida."""
    pass

def validar_idade_motorista(idade):
    """Valida idade mínima para motorista."""
    if idade < 18:
        raise IdadeInvalidaError(f"Idade {idade} insuficiente (mínimo: 18)")
    return f"✅ Idade {idade} válida para dirigir"

print("✨ Exceção personalizada simples:")
idades = [25, 16, 18, 12]
for idade in idades:
    try:
        resultado = validar_idade_motorista(idade)
        print(f"   {resultado}")
    except IdadeInvalidaError as e:
        print(f"   ❌ {e}")
print()

# Exemplo 2: Exceção com atributos customizados 💰
class SaldoInsuficienteError(Exception):
    """Levantada quando saldo é insuficiente para operação."""
    def __init__(self, saldo_atual, valor_solicitado):
        self.saldo_atual = saldo_atual
        self.valor_solicitado = valor_solicitado
        self.deficit = valor_solicitado - saldo_atual
        mensagem = f"Saldo insuficiente: R$ {saldo_atual:.2f} < R$ {valor_solicitado:.2f}"
        super().__init__(mensagem)

def sacar(saldo, valor):
    """Realiza saque com validação de saldo."""
    if valor > saldo:
        raise SaldoInsuficienteError(saldo, valor)
    return saldo - valor

print("💰 Exceção com atributos:")
saldo_conta = 1000.00
saques = [200, 500, 1500]

for valor in saques:
    try:
        novo_saldo = sacar(saldo_conta, valor)
        print(f"   Saque R$ {valor:.2f}: ✅ Novo saldo R$ {novo_saldo:.2f}")
        saldo_conta = novo_saldo
    except SaldoInsuficienteError as e:
        print(f"   Saque R$ {valor:.2f}: ❌ {e}")
        print(f"      Déficit: R$ {e.deficit:.2f}")
print()

# Exemplo 3: Hierarquia de exceções 🎯
class ErroValidacao(Exception):
    """Erro base para validações."""
    pass

class CpfInvalidoError(ErroValidacao):
    """CPF inválido."""
    pass

class EmailInvalidoError(ErroValidacao):
    """Email inválido."""
    pass

class TelefoneInvalidoError(ErroValidacao):
    """Telefone inválido."""
    pass

def validar_dados_cadastro(cpf, email, telefone):
    """Valida dados de cadastro."""
    if len(cpf.replace(".", "").replace("-", "")) != 11:
        raise CpfInvalidoError(f"CPF inválido: {cpf}")
    
    if "@" not in email:
        raise EmailInvalidoError(f"Email inválido: {email}")
    
    if not telefone.replace("-", "").replace("(", "").replace(")", "").replace(" ", "").isdigit():
        raise TelefoneInvalidoError(f"Telefone inválido: {telefone}")
    
    return "✅ Todos os dados válidos"

print("🎯 Hierarquia de exceções:")
dados = [
    ("123.456.789-01", "ana@email.com", "(11) 98765-4321"),
    ("123.456", "bruno@email.com", "(11) 98765-4321"),
    ("987.654.321-00", "carla.email.com", "(11) 98765-4321"),
    ("111.222.333-44", "daniel@email.com", "abc-defg"),
]

for cpf, email, tel in dados:
    try:
        resultado = validar_dados_cadastro(cpf, email, tel)
        print(f"   {resultado}")
    except ErroValidacao as e:
        # Captura qualquer erro de validação
        print(f"   ❌ {type(e).__name__}: {e}")
print()

# Exemplo 4: Exceções de domínio de negócio 🏢
class ErroReserva(Exception):
    """Erro base para sistema de reservas."""
    pass

class QuartoIndisponivelError(ErroReserva):
    """Quarto não disponível."""
    pass

class DataInvalidaError(ErroReserva):
    """Data de reserva inválida."""
    pass

class LimiteReservasError(ErroReserva):
    """Limite de reservas atingido."""
    pass

def fazer_reserva(quarto, data, reservas_cliente):
    """Faz reserva com validações de negócio."""
    quartos_disponiveis = [101, 102, 103]
    
    if quarto not in quartos_disponiveis:
        raise QuartoIndisponivelError(f"Quarto {quarto} não disponível")
    
    if data < "2026-06-01":
        raise DataInvalidaError(f"Data {data} é passada")
    
    if reservas_cliente >= 3:
        raise LimiteReservasError("Cliente já possui 3 reservas ativas")
    
    return f"✅ Reserva confirmada: Quarto {quarto} em {data}"

print("🏢 Exceções de negócio:")
tentativas = [
    (101, "2026-06-15", 1),
    (105, "2026-06-20", 1),
    (102, "2026-05-01", 2),
    (103, "2026-06-25", 3),
]

for quarto, data, reservas in tentativas:
    try:
        resultado = fazer_reserva(quarto, data, reservas)
        print(f"   {resultado}")
    except ErroReserva as e:
        print(f"   ❌ {type(e).__name__}: {e}")
print()

# Exemplo 5: Exceção com métodos adicionais 🛠️
class ErroProcessamento(Exception):
    """Erro de processamento com contexto."""
    def __init__(self, mensagem, codigo_erro, detalhes=None):
        super().__init__(mensagem)
        self.codigo_erro = codigo_erro
        self.detalhes = detalhes or {}
    
    def to_dict(self):
        """Converte erro para dicionário."""
        return {
            "mensagem": str(self),
            "codigo": self.codigo_erro,
            "detalhes": self.detalhes
        }

def processar_pedido(pedido_id, itens):
    """Processa pedido com validações."""
    if not itens:
        raise ErroProcessamento(
            "Pedido sem itens",
            codigo_erro="PED001",
            detalhes={"pedido_id": pedido_id}
        )
    
    if len(itens) > 10:
        raise ErroProcessamento(
            "Pedido excede limite de itens",
            codigo_erro="PED002",
            detalhes={"pedido_id": pedido_id, "itens": len(itens)}
        )
    
    return f"✅ Pedido {pedido_id} processado: {len(itens)} itens"

print("🛠️ Exceção com métodos:")
pedidos = [
    (1001, ["item1", "item2"]),
    (1002, []),
    (1003, [f"item{i}" for i in range(15)]),
]

for ped_id, itens in pedidos:
    try:
        resultado = processar_pedido(ped_id, itens)
        print(f"   {resultado}")
    except ErroProcessamento as e:
        erro_dict = e.to_dict()
        print(f"   ❌ Erro {erro_dict['codigo']}: {erro_dict['mensagem']}")
        print(f"      Detalhes: {erro_dict['detalhes']}")
````

## 6️⃣ Boas Práticas com Exceções

### 🎯 10 Recomendações Essenciais

#### 1️⃣ Capture Exceções Específicas
```python
# ❌ Ruim - captura tudo
try:
    operacao()
except:
    pass

# ✅ Bom - específico
try:
    operacao()
except ValueError:
    print("Valor inválido")
```

#### 2️⃣ Não Use Exceções para Fluxo Normal
```python
# ❌ Ruim - usa exceção para controle
try:
    valor = dict[chave]
except KeyError:
    valor = default

# ✅ Bom - usa get()
valor = dict.get(chave, default)
```

#### 3️⃣ Forneça Mensagens Descritivas
```python
# ❌ Ruim
raise ValueError("Erro")

# ✅ Bom
raise ValueError(f"Idade {idade} inválida: deve ser entre 0 e 150")
```

#### 4️⃣ Use Finally para Limpeza
```python
# ✅ Bom - garante limpeza
try:
    arquivo = open("dados.txt")
    processar(arquivo)
finally:
    arquivo.close()
```

#### 5️⃣ Valide Cedo (Fail Fast)
```python
# ✅ Bom - valida no início
def processar(valor):
    if valor < 0:
        raise ValueError("Valor deve ser positivo")
    # restante do código
```

#### 6️⃣ Não Silencie Erros
```python
# ❌ Ruim - engole erro
try:
    operacao()
except Exception:
    pass  # Erro ignorado!

# ✅ Bom - loga erro
try:
    operacao()
except Exception as e:
    print(f"Erro: {e}")
    # ou log.error(e)
```

#### 7️⃣ Use Context Managers (with)
```python
# ✅ Bom - gerencia recurso automaticamente
with open("dados.txt") as arquivo:
    dados = arquivo.read()
# Arquivo fechado automaticamente
```

#### 8️⃣ Documente Exceções
```python
def dividir(a, b):
    """
    Divide dois números.
    
    Args:
        a: Dividendo
        b: Divisor
    
    Returns:
        float: Resultado da divisão
    
    Raises:
        ZeroDivisionError: Se b for zero
        TypeError: Se a ou b não forem números
    """
    return a / b
```

#### 9️⃣ Limpe Recursos em Finally
```python
# ✅ Bom
conexao = None
try:
    conexao = conectar_banco()
    processar(conexao)
finally:
    if conexao:
        conexao.fechar()
```

#### 🔟 Capture no Nível Apropriado
* Capture onde você pode **tratar** o erro
* Deixe propagar se não souber como tratar
* Não capture apenas para re-lançar

### 🎓 Resumo Final
* **try/except**: Captura e trata erros
* **else**: Executa se não houver erro

* **Personalizadas**: Crie para seu domínio
* **Específico**: Capture exceções específicas
* **Mensagens**: Descritivas e úteis
* **Documentação**: Documente exceções possíveis
