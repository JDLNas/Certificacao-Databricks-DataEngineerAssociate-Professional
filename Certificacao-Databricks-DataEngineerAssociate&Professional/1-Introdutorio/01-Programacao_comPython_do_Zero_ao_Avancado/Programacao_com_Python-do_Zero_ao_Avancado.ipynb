### Principais Tipos de Dados em Python

## 6. Operadores AritimĂŠticos

### Operadores AritmĂŠticos em Python

Python suporta os seguintes operadores aritmĂŠticos:

* **+** (adiĂ§ĂŁo)
* **-** (subtraĂ§ĂŁo)
* ***** (multiplicaĂ§ĂŁo)
* **/** (divisĂŁo com resultado decimal)
* **//** (divisĂŁo inteira - retorna apenas a parte inteira)
* **%** (mĂłdulo - retorna o resto da divisĂŁo)
* ****** (potenciaĂ§ĂŁo/exponenciaĂ§ĂŁo)

````python
# DeclaraĂ§ĂŁo de variĂĄveis para os exemplos
a = 10
b = 5
c = 11

# AdiĂ§ĂŁo
print(f"AdiĂ§ĂŁo: {a} + {b} = {a + b}")

# SubtraĂ§ĂŁo
print(f"SubtraĂ§ĂŁo: {a} - {b} = {a - b}")
print(f"SubtraĂ§ĂŁo inversa: {b} - {a} = {b - a}")

# DivisĂŁo (resultado decimal)
print(f"DivisĂŁo: {a} / {b} = {a / b}")
print(f"DivisĂŁo inversa: {b} / {a} = {b / a}")
````

### DivisĂŁo Inteira (//)

Retorna apenas a parte inteira da divisĂŁo, descartando a parte decimal.

### Operador MĂłdulo (%)

Retorna o **resto** da divisĂŁo entre dois nĂşmeros. Ătil para verificar se um nĂşmero ĂŠ par/Ă­mpar ou divisĂ­vel por outro.

````python
# VariĂĄveis para os exemplos
a = 10
b = 5
c = 11

# DivisĂŁo inteira (//) - retorna apenas a parte inteira
print(f"DivisĂŁo inteira: {a} // {b} = {a // b}")
print(f"DivisĂŁo inteira inversa: {b} // {a} = {b // a}")

# Operador mĂłdulo (%) - retorna o resto da divisĂŁo
print(f"MĂłdulo: {a} % {b} = {a % b}")  # 10 dividido por 5 tem resto 0
print(f"MĂłdulo: {c} % {b} = {c % b}")  # 11 dividido por 5 tem resto 1
print(f"MĂłdulo: {a} % {c} = {a % c}")  # 10 dividido por 11 tem resto 10
````

### PotenciaĂ§ĂŁo (**)

Eleva um nĂşmero a uma potĂŞncia.

### PrecedĂŞncia de Operadores

Python segue a ordem matemĂĄtica padrĂŁo:
1. **ParĂŞnteses** ()
2. **PotenciaĂ§ĂŁo** **
3. **MultiplicaĂ§ĂŁo, DivisĂŁo, MĂłdulo** *, /, //, %
4. **AdiĂ§ĂŁo e SubtraĂ§ĂŁo** +, -

````python
# VariĂĄveis para os exemplos
a = 10
b = 5
c = 11

# PotenciaĂ§ĂŁo (exponenciaĂ§ĂŁo)
print(f"PotĂŞncia: {a} ** 2 = {a ** 2}")  # 10 ao quadrado
print(f"PotĂŞncia: {a} ** {b} = {a ** b}")  # 10 elevado a 5

# PrecedĂŞncia de operadores - multiplicaĂ§ĂŁo antes da adiĂ§ĂŁo
resultado1 = a + b * c
print(f"Sem parĂŞnteses: {a} + {b} * {c} = {resultado1}")  # 5 * 11 = 55, depois 10 + 55 = 65

# ParĂŞnteses alteram a precedĂŞncia
resultado2 = (a + b) * c
print(f"Com parĂŞnteses: ({a} + {b}) * {c} = {resultado2}")  # 10 + 5 = 15, depois 15 * 11 = 165
````

## 8. Operadores de ComparaĂ§ĂŁo ou Relacionais

### Operadores de ComparaĂ§ĂŁo (Relacionais)

Operadores de comparaĂ§ĂŁo sĂŁo usados para comparar dois valores e **sempre retornam um valor booleano** (`True` ou `False`).

**Lista de operadores:**

* **==** (igual a)
* **!=** (diferente de)
* **>** (maior que)
* **<** (menor que)
* **>=** (maior ou igual a)
* **<=** (menor ou igual a)

````python
# DeclaraĂ§ĂŁo de variĂĄveis para os exemplos
a = 5
b = 5

# Operador de igualdade (==) - verifica se os valores sĂŁo iguais
print(f"a = {a}")
print(f"b = {b}")
print(f"a == 5? {a == 5}")  # True, pois a ĂŠ 5
print(f"b == 6? {b == 6}")  # False, pois b ĂŠ 5, nĂŁo 6
print(f"a == b? {a == b}")  # True, pois ambos valem 5

print("\n" + "=" * 50 + "\n")

# Operador de diferenĂ§a (!=) - verifica se os valores sĂŁo diferentes
print(f"a != b? {a != b}")  # False, pois sĂŁo iguais
print(f"b != (a + 1)? {b != a + 1}")  # True, pois 5 != 6
````

### Operadores de Magnitude (Maior/Menor)

Comparam o tamanho relativo entre dois valores:

* **>** verifica se o valor Ă  esquerda ĂŠ **maior** que o da direita
* **<** verifica se o valor Ă  esquerda ĂŠ **menor** que o da direita
* **>=** verifica se o valor Ă  esquerda ĂŠ **maior ou igual** ao da direita
* **<=** verifica se o valor Ă  esquerda ĂŠ **menor ou igual** ao da direita

````python
# VariĂĄveis para os exemplos
a = 5
b = 5
c = 10

# Operadores de maior e menor
print(f"a = {a}, b = {b}, c = {c}")
print(f"a > b? {a > b}")  # False, pois sĂŁo iguais
print(f"c > a? {c > a}")  # True, pois 10 > 5
print(f"a < c? {a < c}")  # True, pois 5 < 10

print("\n" + "=" * 50 + "\n")

# Operadores de maior ou igual / menor ou igual
print(f"a >= b? {a >= b}")  # True, pois sĂŁo iguais (5 >= 5)
print(f"a * 2 >= c? {a * 2 >= c}")  # True, pois 10 >= 10
print(f"a <= c? {a <= c}")  # True, pois 5 <= 10
````

### Armazenando Resultados em VariĂĄveis

Os resultados das comparaĂ§Ăľes (valores booleanos) podem ser armazenados em variĂĄveis para uso posterior em lĂłgica condicional.

````python
# DeclaraĂ§ĂŁo de variĂĄveis
n1 = 10
n2 = 20

# Armazenando o resultado de comparaĂ§Ăľes em variĂĄveis
iguais = n1 == n2
diferentes = n1 != n2
maior = n1 > n2

print(f"n1 = {n1}, n2 = {n2}")
print(f"SĂŁo iguais? {iguais}")  # False
print(f"SĂŁo diferentes? {diferentes}")  # True
print(f"n1 ĂŠ maior que n2? {maior}")  # False
````

### Comparador Interativo de NĂşmeros

Vamos criar um programa que solicita dois nĂşmeros ao usuĂĄrio e exibe o resultado de vĂĄrias comparaĂ§Ăľes entre eles.

````python
# InicializaĂ§ĂŁo das variĂĄveis de resultado
x = y = z = False
n1 = n2 = 0

# Solicitando entrada do usuĂĄrio com tratamento de erro
try:
    n1 = int(input("Digite um nĂşmero: "))
    n2 = int(input("Digite outro nĂşmero: "))
    
    print("\n" + "=" * 50)
    print(f"Comparando {n1} e {n2}")
    print("=" * 50 + "\n")
    
    # ComparaĂ§ĂŁo de igualdade
    x = n1 == n2
    print(f"SĂŁo iguais? {x}\n")
    
    # ComparaĂ§ĂŁo de diferenĂ§a
    y = n1 != n2
    print(f"SĂŁo diferentes? {y}\n")
    
    # ComparaĂ§ĂŁo de magnitude (maior que)
    z = n1 > n2
    print(f"{n1} ĂŠ maior que {n2}? {z}\n")
    
    # ComparaĂ§Ăľes adicionais
    print(f"{n1} ĂŠ menor que {n2}? {n1 < n2}\n")
    print(f"{n1} ĂŠ maior ou igual a {n2}? {n1 >= n2}\n")
    print(f"{n1} ĂŠ menor ou igual a {n2}? {n1 <= n2}")
    
except ValueError:
    print("Erro: Por favor, digite apenas nĂşmeros inteiros vĂĄlidos.")
````

## 9. Operadores LĂłgicos AND, OR, NOT

### Operadores LĂłgicos em Python

Operadores lĂłgicos sĂŁo usados para combinar expressĂľes condicionais e **sempre retornam valores booleanos** (`True` ou `False`).

**Os trĂŞs operadores lĂłgicos:**

* **and** - retorna `True` apenas se **ambas** as condiĂ§Ăľes forem verdadeiras
* **or** - retorna `True` se **pelo menos uma** das condiĂ§Ăľes for verdadeira
* **not** - **inverte** o valor booleano (True vira False, False vira True)

````python
# Exemplo: Verificar se uma pessoa pode participar de um evento
# Requisitos: ter pelo menos 18 anos E ter pelo menos 1.75m de altura

idade = 15
altura = 1.75

# Operador AND: ambas as condiĂ§Ăľes precisam ser True
resultado = (idade >= 18) and (altura >= 1.75)

# Construindo e exibindo a mensagem
msg = "Pode participar do evento? " + str(resultado)
print(msg)
print(f"\nExplicaĂ§ĂŁo: idade ({idade}) >= 18? {idade >= 18}")
print(f"ExplicaĂ§ĂŁo: altura ({altura}) >= 1.75? {altura >= 1.75}")
print(f"AND: Ambas verdadeiras? {resultado}")
````

### Operador AND - Tabela Verdade

O operador `and` sĂł retorna `True` quando **ambas** as condiĂ§Ăľes sĂŁo verdadeiras:

| CondiĂ§ĂŁo A | CondiĂ§ĂŁo B | A and B |
| --- | --- | --- |
| True | True | **True** |
| True | False | False |
| False | True | False |
| False | False | False |

### Operador OR (OU)

O operador `or` retorna `True` se **pelo menos uma** das condiĂ§Ăľes for verdadeira. Ă usado quando queremos verificar se qualquer uma de vĂĄrias condiĂ§Ăľes ĂŠ satisfeita.

````python
# Exemplo: Sistema de alarme
# O alarme dispara se a porta OU a janela estiverem abertas
# "a" = aberto, "f" = fechado

porta = "a"
janela = "f"

# Operador OR: pelo menos uma condiĂ§ĂŁo precisa ser True
alarme = (porta == "a") or (janela == "a")

# Construindo e exibindo a mensagem
msg = "Alarme ativado? " + str(alarme)
print(msg)
print(f"\nExplicaĂ§ĂŁo: porta aberta? {porta == 'a'}")
print(f"ExplicaĂ§ĂŁo: janela aberta? {janela == 'a'}")
print(f"OR: Pelo menos uma aberta? {alarme}")
````

### Operador OR - Tabela Verdade

O operador `or` retorna `True` quando **pelo menos uma** condiĂ§ĂŁo ĂŠ verdadeira:

| CondiĂ§ĂŁo A | CondiĂ§ĂŁo B | A or B |
| --- | --- | --- |
| True | True | **True** |
| True | False | **True** |
| False | True | **True** |
| False | False | False |

### Operador NOT (NĂO)

O operador `not` **inverte** o valor booleano:
* `not True` retorna `False`
* `not False` retorna `True`

Ă usado para negar uma condiĂ§ĂŁo ou inverter o estado de uma variĂĄvel booleana.

````python
# Exemplo: Inverter o estado de uma variĂĄvel booleana
# Inicialmente a pessoa nĂŁo tem dinheiro, mas depois recebe

tem_dinheiro = False
print(f"Estado inicial: tem dinheiro? {tem_dinheiro}")

# Operador NOT: inverte o valor booleano
tem_dinheiro = not tem_dinheiro

# Exibindo o novo estado
msg = "Tem dinheiro? " + str(tem_dinheiro)
print(msg)
print(f"\nExplicaĂ§ĂŁo: NOT False = {not False}")
print(f"ExplicaĂ§ĂŁo: NOT True = {not True}")
````

### Operador NOT - Tabela Verdade

O operador `not` simplesmente **inverte** o valor:

| CondiĂ§ĂŁo A | not A |
| --- | --- |
| True | **False** |
| False | **True** |

### Combinando MĂşltiplos Operadores LĂłgicos

Podemos combinar AND, OR e NOT em expressĂľes mais complexas. Use parĂŞnteses para controlar a ordem de avaliaĂ§ĂŁo.

**Ordem de precedĂŞncia:**
1. **not** (maior precedĂŞncia)
2. **and**
3. **or** (menor precedĂŞncia)

````python
# Exemplo complexo: Acesso a um sistema
# Regra: (usuĂĄrio autenticado E tem permissĂŁo) OU (ĂŠ administrador)

usuario_autenticado = True
tem_permissao = False
e_administrador = False

# Combinando operadores lĂłgicos
acesso_permitido = (usuario_autenticado and tem_permissao) or e_administrador

print("=" * 50)
print("Sistema de Controle de Acesso")
print("=" * 50)
print(f"UsuĂĄrio autenticado? {usuario_autenticado}")
print(f"Tem permissĂŁo? {tem_permissao}")
print(f"Ă administrador? {e_administrador}")
print(f"\nAcesso permitido? {acesso_permitido}")

# Testando com NOT
print(f"\nAcesso NEGADO? {not acesso_permitido}")
````

## 10. Desvios Condicionais - if, elif, else

### Estruturas Condicionais: if, elif, else

As estruturas condicionais permitem que o programa tome **decisĂľes** e execute diferentes blocos de cĂłdigo baseados em condiĂ§Ăľes.

**Sintaxe bĂĄsica:**

```python
if condiĂ§ĂŁo:
    # cĂłdigo executado se a condiĂ§ĂŁo for True
elif outra_condiĂ§ĂŁo:
    # cĂłdigo executado se a primeira for False e esta for True
else:
    # cĂłdigo executado se todas as condiĂ§Ăľes anteriores forem False
```

**Regras importantes:**
* A indentaĂ§ĂŁo (espaĂ§os Ă  esquerda) ĂŠ **obrigatĂłria** e define o bloco de cĂłdigo
* O `elif` e `else` sĂŁo opcionais
* Pode haver mĂşltiplos `elif`, mas apenas um `else`
* Os dois-pontos (`:`) sĂŁo obrigatĂłrios apĂłs cada condiĂ§ĂŁo

### Sistema de AprovaĂ§ĂŁo de Aluno

Vamos criar um programa que determina se um aluno foi **aprovado**, estĂĄ em **recuperaĂ§ĂŁo** ou foi **reprovado** baseado em:

**CritĂŠrios de aprovaĂ§ĂŁo:**
* **Aprovado**: mĂŠdia >= 7 **E** menos de 10 faltas
* **RecuperaĂ§ĂŁo**: mĂŠdia >= 5 **E** menos de 10 faltas
* **Reprovado**: qualquer outra situaĂ§ĂŁo (mĂŠdia < 5 OU faltas >= 10)

````python
# InicializaĂ§ĂŁo das variĂĄveis
n1 = n2 = 0.0
media = 0.0

# Solicitando entrada do usuĂĄrio
try:
    n1 = float(input("Digite a primeira nota: "))
    n2 = float(input("Digite a segunda nota: "))
    faltas = int(input("Digite o nĂşmero de faltas: "))
    
    # Calculando a mĂŠdia aritmĂŠtica das duas notas
    media = (n1 + n2) / 2
    
    print(f"\nNotas: {n1} e {n2}")
    print(f"MĂŠdia calculada: {media:.2f}")
    print(f"NĂşmero de faltas: {faltas}")
    print("\n" + "=" * 50)
    
except ValueError:
    print("Erro: Por favor, digite valores numĂŠricos vĂĄlidos.")
````

````python
# Estrutura condicional para determinar a situaĂ§ĂŁo do aluno
# Usa operadores lĂłgicos (and) para combinar mĂşltiplas condiĂ§Ăľes

if (media >= 7) and (faltas < 10):
    # Primeira condiĂ§ĂŁo: mĂŠdia suficiente E poucas faltas
    print("â APROVADO")
    print("ParabĂŠns! VocĂŞ atingiu a mĂŠdia necessĂĄria e teve frequĂŞncia adequada.")
    
elif (media >= 5) and (faltas < 10):
    # Segunda condiĂ§ĂŁo: mĂŠdia intermediĂĄria E poucas faltas
    print("â  RECUPERAĂĂO")
    print("VocĂŞ precisarĂĄ fazer a prova de recuperaĂ§ĂŁo.")
    
else:
    # Qualquer outra situaĂ§ĂŁo: mĂŠdia insuficiente OU muitas faltas
    print("â REPROVADO")
    if faltas >= 10:
        print("Motivo: NĂşmero de faltas excedeu o limite permitido.")
    if media < 5:
        print("Motivo: MĂŠdia abaixo do mĂ­nimo necessĂĄrio.")

# Exibindo resumo final
print("\n" + "=" * 50)
print(f"Resumo: MĂŠdia = {media:.2f} | Faltas = {faltas}")
print("=" * 50)
````

### Fluxo de ExecuĂ§ĂŁo

A estrutura condicional ĂŠ avaliada **sequencialmente de cima para baixo**:

1. **if**: Avalia a primeira condiĂ§ĂŁo
   * Se `True`: executa o bloco e **pula** o restante
   * Se `False`: continua para o prĂłximo

2. **elif**: Avalia a segunda condiĂ§ĂŁo (sĂł se o `if` foi `False`)
   * Se `True`: executa o bloco e **pula** o restante
   * Se `False`: continua para o prĂłximo

3. **else**: Executado apenas se **todas** as condiĂ§Ăľes anteriores foram `False`

**Importante:** Apenas **um** bloco serĂĄ executado, mesmo que mĂşltiplas condiĂ§Ăľes sejam verdadeiras!

### F-Strings - FormataĂ§ĂŁo Moderna

As **f-strings** (introduzidas no Python 3.6+) sĂŁo a forma **mais moderna e legĂ­vel** de formatar strings.

**Sintaxe:**
```python
f"texto {variavel} mais texto {expressao}"
```

**Vantagens:**
* Mais **legĂ­vel** e **conciso**
* Permite **expressĂľes** dentro das chaves
* **Melhor performance** que `.format()`

````python
# F-strings: forma mais moderna de formataĂ§ĂŁo
nome = "FĂĄbio"
peso = 73.50

# f"..." indica uma f-string
# VariĂĄveis e expressĂľes podem ser usadas diretamente dentro de {}
print(f"OlĂĄ, meu nome ĂŠ {nome} e eu peso {peso} quilos.")

# F-strings permitem expressĂľes matemĂĄticas
a = 10
b = 5
resultado = f"A soma de {a} com {b} ĂŠ igual a {a + b}"
print(resultado)
````

### FormataĂ§ĂŁo de NĂşmeros em F-Strings

Podemos controlar a **formataĂ§ĂŁo de nĂşmeros** usando especificadores de formato:

* `{valor:.2f}` - Float com **2 casas decimais**
* `{valor:.0f}` - Float **sem casas decimais**
* `{valor:,.2f}` - Float com separador de milhares e 2 decimais
* `{valor:>10}` - Alinhamento Ă  direita (10 espaĂ§os)
* `{valor:<10}` - Alinhamento Ă  esquerda (10 espaĂ§os)

````python
# FormataĂ§ĂŁo de nĂşmeros decimais com f-strings
valor = 125.579637

# :.2f significa: formato float com 2 casas decimais
print(f"O valor ĂŠ {valor:.2f}")

# Exemplos adicionais de formataĂ§ĂŁo
print(f"Sem casas decimais: {valor:.0f}")
print(f"Com 4 casas decimais: {valor:.4f}")

# FormataĂ§ĂŁo com separador de milhares
valor_grande = 1234567.89
print(f"Com separador de milhares: {valor_grande:,.2f}")
````

### Caracteres de Escape

* `\'` - Aspas simples literal

## 13. LaĂ§os de RepetiĂ§ĂŁo While

### LaĂ§o de RepetiĂ§ĂŁo While

O laĂ§o `while` executa um bloco de cĂłdigo **repetidamente enquanto** uma condiĂ§ĂŁo for verdadeira.

**Sintaxe bĂĄsica:**
```python
while condiĂ§ĂŁo:
    # cĂłdigo executado enquanto a condiĂ§ĂŁo for True
    # atualizaĂ§ĂŁo da condiĂ§ĂŁo (importante para evitar loop infinito)
```

**Componentes essenciais:**
1. **CondiĂ§ĂŁo**: ExpressĂŁo booleana avaliada antes de cada iteraĂ§ĂŁo
2. **Corpo do loop**: CĂłdigo indentado que serĂĄ repetido
3. **AtualizaĂ§ĂŁo**: ModificaĂ§ĂŁo da variĂĄvel de controle para eventualmente tornar a condiĂ§ĂŁo False

**â ď¸ Cuidado:** Se a condiĂ§ĂŁo nunca se tornar False, o loop serĂĄ **infinito**!

````python
# LaĂ§o while bĂĄsico: contador de 1 a 10

# InicializaĂ§ĂŁo: variĂĄvel de controle do loop
num = 1

# Loop: enquanto num for menor ou igual a 10
while (num <= 10):
    print(num)  # Exibe o nĂşmero atual
    num += 1    # Incrementa o contador (essencial para evitar loop infinito)

# ApĂłs sair do loop (quando num > 10)
print("Finalizado!")
````

### Como o While Funciona

**Fluxo de execuĂ§ĂŁo:**
1. **Avalia a condiĂ§ĂŁo** antes de cada iteraĂ§ĂŁo
2. Se **True**: executa o corpo do loop e retorna ao passo 1
3. Se **False**: **sai do loop** e continua com o cĂłdigo apĂłs o while

**DiferenĂ§a entre while e if:**
* `if`: executa o bloco **uma vez** se a condiĂ§ĂŁo for True
* `while`: executa o bloco **repetidamente** enquanto a condiĂ§ĂŁo for True

### Operadores Compostos de AtribuiĂ§ĂŁo

Ă comum usar operadores compostos para atualizar variĂĄveis em loops:

* `num += 1` â equivalente a `num = num + 1` (incremento)
* `num -= 1` â equivalente a `num = num - 1` (decremento)
* `num *= 2` â equivalente a `num = num * 2` (multiplicaĂ§ĂŁo)
* `num /= 2` â equivalente a `num = num / 2` (divisĂŁo)

**Uso comum em loops:**
* `num += 1` - incrementa de 1 em 1
* `num += 2` - incrementa de 2 em 2
* `num *= 2` - dobra o valor a cada iteraĂ§ĂŁo

````python
# Exemplo 1: Contador decrescente
print("Contagem regressiva:")
contador = 10

while contador > 0:
    print(contador)
    contador -= 1  # Decremento: subtrai 1 a cada iteraĂ§ĂŁo

print("LanĂ§amento!\n")

# Exemplo 2: Incremento de 2 em 2
print("NĂşmeros pares de 2 a 10:")
par = 2

while par <= 10:
    print(par)
    par += 2  # Incrementa de 2 em 2

print("\nFinalizado!")
````

### O Comando `continue`

**DiferenĂ§a entre break e continue:**

## 15. Encadeamento de LaĂ§os de RepetiĂ§ĂŁo

### LaĂ§os de RepetiĂ§ĂŁo Aninhados (Nested Loops)

LaĂ§os aninhados sĂŁo laĂ§os **dentro de outros laĂ§os**. O laĂ§o interno ĂŠ executado completamente a **cada iteraĂ§ĂŁo** do laĂ§o externo.

**Estrutura bĂĄsica:**
```python
for externo in sequencia_externa:
    # CĂłdigo do laĂ§o externo
    for interno in sequencia_interna:
        # CĂłdigo do laĂ§o interno (executa N vezes para cada iteraĂ§ĂŁo externa)
    # Continua o laĂ§o externo
```

**Como funciona:**
* Para **cada** iteraĂ§ĂŁo do laĂ§o externo
* O laĂ§o interno executa **todas** as suas iteraĂ§Ăľes
* Depois volta para a prĂłxima iteraĂ§ĂŁo do laĂ§o externo

**AplicaĂ§Ăľes comuns:**
* Processar matrizes (tabelas 2D)
* Criar padrĂľes e desenhos
* Gerar combinaĂ§Ăľes
* Percorrer estruturas bidimensionais

````python
# Exemplo bĂĄsico: laĂ§o dentro de outro laĂ§o
# LaĂ§o externo: rodadas de 1 a 5
# LaĂ§o interno: contagem regressiva de 5 a 1

print("DEMONSTRAĂĂO: LAĂOS ANINHADOS")
print("=" * 50)

# LaĂ§o externo: controla as rodadas
for cont_ext in range(1, 6):
    print(f"\nđš Rodada {cont_ext}")
    print("-" * 30)
    
    # LaĂ§o interno: executa completamente a cada rodada
    # Conta de 5 atĂŠ 1 (range com step negativo)
    for cont_int in range(5, 0, -1):
        print(f"  Valor: {cont_int}")

# ApĂłs todos os laĂ§os terminarem
print("\n" + "=" * 50)
print("â Fim dos laĂ§os")
print(f"\nTotal de iteraĂ§Ăľes do laĂ§o externo: 5")
print(f"Total de iteraĂ§Ăľes do laĂ§o interno: 5 Ă 5 = 25")
````

### Fluxo de ExecuĂ§ĂŁo

**Exemplo com nĂşmeros pequenos:**

```python
    for j in range(1, 3):  # Interno: 1, 2
```

**SaĂ­da:**
```

```

**ObservaĂ§ĂŁo importante:**
* Se o laĂ§o externo tem **N** iteraĂ§Ăľes
* E o laĂ§o interno tem **M** iteraĂ§Ăľes
* O cĂłdigo dentro do laĂ§o interno executa **N Ă M** vezes

````python
# VisualizaĂ§ĂŁo clara: mostrando cada iteraĂ§ĂŁo dos laĂ§os aninhados

print("VISUALIZAĂĂO: LAĂOS ANINHADOS")
print("=" * 50)

contador_total = 0  # Conta o total de execuĂ§Ăľes do laĂ§o interno

# LaĂ§o externo: 3 iteraĂ§Ăľes
for externo in range(1, 4):
    print(f"\n>>> LAĂO EXTERNO: iteraĂ§ĂŁo {externo}")
    
    # LaĂ§o interno: 4 iteraĂ§Ăľes
    for interno in range(1, 5):
        contador_total += 1
        print(f"    LAĂO INTERNO: iteraĂ§ĂŁo {interno} (execuĂ§ĂŁo #{contador_total} no total)")

print("\n" + "=" * 50)
print(f"Resumo: LaĂ§o externo executou 3 vezes")
print(f"        LaĂ§o interno executou 4 vezes por cada externa = 12 vezes no total")
````

**Importante:** `break` e `continue` afetam apenas o laĂ§o **mais interno** onde estĂŁo:

**Break em laĂ§o interno:**
```python

    for interno in range(3):
        if interno == 2:
```

**Para sair de mĂşltiplos laĂ§os:**

### Desafio: Encontrar NĂşmeros Primos com LaĂ§os Aninhados

Use laĂ§os aninhados para encontrar todos os nĂşmeros primos entre 2 e 50.

**LĂłgica:**
* LaĂ§o externo: percorre cada nĂşmero de 2 a 50
* LaĂ§o interno: testa se o nĂşmero ĂŠ divisĂ­vel por algum valor
* Se nĂŁo for divisĂ­vel por nenhum (exceto 1 e ele mesmo), ĂŠ primo

````python
# Programa: Encontrar NĂşmeros Primos com LaĂ§os Aninhados

print("=" * 50)
print("NĂMEROS PRIMOS ENTRE 2 E 50")
print("=" * 50)

primos_encontrados = []  # Lista para armazenar os primos

# LaĂ§o externo: testa cada nĂşmero de 2 a 50
for numero in range(2, 51):
    eh_primo = True  # Assume que ĂŠ primo inicialmente
    
    # LaĂ§o interno: testa divisibilidade
    # Testa divisores de 2 atĂŠ a raiz quadrada do nĂşmero (otimizaĂ§ĂŁo)
    for divisor in range(2, int(numero ** 0.5) + 1):
        if numero % divisor == 0:
            # Encontrou um divisor, nĂŁo ĂŠ primo
            eh_primo = False
            break  # NĂŁo precisa testar mais divisores
    
    # Se passou por todos os testes, ĂŠ primo
    if eh_primo:
        primos_encontrados.append(numero)

# Exibindo os resultados
print("\nNĂşmeros primos encontrados:")
print("-" * 50)

# Exibe em linhas de 10 nĂşmeros
for i, primo in enumerate(primos_encontrados):
    print(f"{primo:3d}", end="  ")
    if (i + 1) % 10 == 0:  # Nova linha a cada 10 nĂşmeros
        print()

print("\n\n" + "=" * 50)
print(f"Total de nĂşmeros primos encontrados: {len(primos_encontrados)}")
print("=" * 50)
````
